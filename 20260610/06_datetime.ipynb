```python
from datetime import datetime as dt
today = dt.now()
print(today)

print('년: {}'.format(today.year))
print('월: {}'.format(today.month))
print('일: {}'.format(today.day))
print('시: {}'.format(today.hour))
print('분: {}'.format(today.minute))
print('초: {}'.format(today.second))
print('마이크로초: {}'.format(today.microsecond))
print('밀리초: {}'.format(today.microsecond // 1000))
# weekday() 메소드로 요일을 숫자로 얻어올 수 있다. => 월(0), 화(1), 수(2), 목(3), 금(4), 토(5), 일(6)
print('요일: {}'.format(today.weekday()))
week = ['월', '화', '수', '목', '금', '토', '일']
print('요일: {}요일'.format(week[today.weekday()]))
```

날짜, 시간 수정하기

```python
from datetime import datetime as dt
today = dt.now()
print(today)

# replace() 메소드로 날짜, 시간을 수정할 수 있다.
print(today.replace(year=2027))
print(today.replace(month=12))
print(today.replace(day=25))
print(today.replace(hour=17))
print(today.replace(minute=40))
print(today.replace(second=1))
print(today.replace(microsecond=111111))
```

날짜, 시간 데이터 만들기

```python
from dateutil.parser import parse
```

```python
string = '20260612' # 2026-6-12, 2026/6/12, 2026.6.12, 20260612
print(string, type(string))
date = parse(string)
print(date, type(date))

string = '12:50:01'
print(string, type(string))
time = parse(string)
print(time, type(time))

string = '2026-6-12 12:50:01'
print(string, type(string))
now = parse(string)
print(now, type(now))
```

두 날짜 데이터 사이의 차이 계산하기

```python
from datetime import datetime as dt
from dateutil.parser import parse

start = dt.now() # 오늘
end = parse('2026-12-3 17:40:1') # 종강일
print(start, end)

gap = end - start
print(gap)
print(gap.days) # 두 날짜 사이의 일 단위 차이를 얻어온다.
print(gap.seconds) # 두 날짜 사이의 날짜를 제외한 시간의 차이를 초단위로 얻어온다.
print(gap.microseconds)
print(gap.total_seconds())
```

날짜, 시간 출력서식 지정하기  
날짜데이터.strftime('출력서식')

날짜 서식  
%Y: 년도 4자리, %y: 년도 2자리  
%m: 월(숫자), %B: 월(영어 => 전체), %b: 월(영어 => 앞 3글자)  
%d: 일  
%A: 요일(영어 => 전체), %a: 요일(영어 => 앞 3글자)  

```python
from datetime import datetime as dt
today = dt.now()
print(today)

print(today.strftime('%Y-%m-%d'))
print(type(today.strftime('%Y-%m-%d')))
print(today.strftime('%Y/%m/%d'))
print(today.strftime('%Y.%m.%d'))
print(today.strftime('%Y-%B-%d %A'))
print(today.strftime('%Y-%b-%d %a'))
print(today.strftime('%a %b-%d %Y'))

# 날짜, 시간 서식에 한글을 넣었을때 유니코드 인코딩 문제로 UnicodeEncodeError가 발생될 수 있다.
print(today.strftime('%Y년 %m월 %d일'))
# 날짜, 시간 서식 지정시 UnicodeEncodeError가 발생되면 아래와 같이 사용하면 된다.
print(today.strftime('%Y년 %m월 %d일'.encode('unicode-escape').decode()).encode().decode('unicode-escape'))
```

시간 서식  
%H: 시간(24시각), %I: 시간(12시각)  
%M: 분  
%S: 초  
%f: 마이크로초  
%p: AM/PM

```python
from datetime import datetime as dt
today = dt.now()
print(today)

print(today.strftime('%H:%M:%S'))
print(today.strftime('%p %I:%M:%S.%f'))
print(today.strftime('%H시 %M분 %S초'))
print(today.strftime('%H시 %M분 %S초'.encode('unicode-escape').decode()).encode().decode('unicode-escape'))
```
