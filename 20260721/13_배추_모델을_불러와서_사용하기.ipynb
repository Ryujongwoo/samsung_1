```python
import warnings
warnings.filterwarnings("ignore")
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np 
import pandas as pd
```

디스크에 저장된 모델을 불러와서 사용하기위한 변수, 플레이스홀더, 수식을 정의한다.

```python
X = tf.placeholder(dtype=tf.float64, shape=[None, 4])
a = tf.Variable(tf.random_uniform([4, 1], dtype=tf.float64))
b = tf.Variable(tf.random_uniform([1], dtype=tf.float64))
y = tf.matmul(X, a) + b
```

배추 가격을 결정하는 4가지 변화 요인을 키보드로 입력받아 모델에 적용할 데이터를 만든다.

```python
# 3.2, -2.5, 7.1, 15.0
avgTemp = float(input('평균온도: '))
minTemp = float(input('최저온도: '))
maxTemp = float(input('최고온도: '))
rainFall = float(input('강수량: '))
newData = [[avgTemp, minTemp, maxTemp, rainFall]]
print(newData)
```

세션을 만들고 디스크에 저장된 모델을 불러와서 넣어준다.

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

saver = tf.train.Saver()
# Saver() 클래스 객체의 restore() 메소드로 디스크에 저장된 모델을 불러와서 세션에 넣어준다.
saver.restore(sess, './model/baechu.ckpt')
print('디스크에 저장된 모델을 불러왔습니다.')
```

불러온 학습 모델에 키보드로 입력한 데이터를 적용시켜 배추 가격을 예측한다.

```python
result = sess.run(y, feed_dict={X: newData})
print('{}'.format(result))
print('{:,.0f}'.format(result[0][0]))
```
