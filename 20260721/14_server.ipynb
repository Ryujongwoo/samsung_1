flask web server를 사용하기 위해서 flask 모듈을 설치하고 import 시킨다.

```python
# !pip install flask
from flask import Flask, render_template, request
```

Flask() 객체를 선언한다.

```python
app = Flask(__name__)
```

Flask의 기본 포트인 5000번 포트의 기본 경로('/')에 GET, POST 방식으로 요청을 받는다.

```python
@app.route('/', methods=['GET', 'POST'])
def index():
    # render_template() 함수를 이용해서 특정 페이지(index.html)를 브라우저에 표시한다.
    return render_template('index.html')
```

```python
if __name__ == '__main__':
    # flask web server를 실행한다.
    app.run(debug=True)
```
