linkedList에 저장할 데이터와 다음 데이터의 주소를 기억하는 클래스 => 노드

```python
class Node:
    def __init__(self, data=None):
        self.data = data # linkedList에 저장할 실제 데이터
        self.next = None # 다음 데이터가 저장된 위치(주소)
```

linkedList 자체를 의미하는 클래스

```python
class LinkedList:
    def __init__(self):
        self.head = None # LinkedList의 시작 위치
        self.count = 0 # LinkedList에 저장된 데이터 개수
        
    # 입력
    # LinkedList에 데이터를 입력하는 경우는 모두 3가지가 있다.
    # 1. LinkedList의 맨 뒤에 데이터를 추가하는 경우
    #    1-1. LinkedList가 비어있는 경우
    #    1-2. LinkedList가 비어있지 않는경우
    # 2. LinkedList의 맨 앞(head 바로 다음 위치)에 데이터를 삽입하는 경우
    # 3. LinkedList의 특정(맨 앞과 맨 뒤를 제외한) 위치에 데이터를 삽입하는 경우

    # LinkedList의 맨 뒤에 데이터를 추가하는 메소드
    def appendLast(self, data):
        # LinkedList의 맨 뒤에 추가할 데이터를 넘겨받아 Node 클래스 객체(LinkedList에 저장할 데이터)를 만든다.
        newNode = Node(data)
        # print('newNode: {}'.format(newNode))
        # LinkedList에 저장된 데이터 개수를 증가시킨다.
        self.count += 1
        
        # LinkedList가 비어있는 경우와 비어있지 않은 경우에 따라 LinkedList에 데이터를 추가하는 방법이 다르다.
        # LinkedList가 비어있나 물어봐서 비어있으면 head 바로 다음에 데이터를 추가하고 메소드를 종료한다.
        if self.head is None: # LinkedList가 비어있는가?
            # LinkedList가 비어있으므로 head 바로 다음에 데이터를 넣어준다. => head에 newData가 메모리에 생성된 주소를 넣어준다.
            self.head = newNode
            # print('self.head: {}'.format(self.head))
            # 데이터를 LinkedList의 head 바로 다음에 추가했으므로 메소드를 return 시켜서 종료한다.
            return
        # ===== if
        
        # LinkedList가 비어있지 않다면 LinkedList의 시작 위치(self.head)부터 마지막 위치를 탐색해서 이동한 후 데이터를 추가한다.
        # LinkedList의 시작 위치를 저장한다.
        start = self.head
        # start 부터 시작해서 LinkedList에 저장된 마지막 데이터로 이동한다. => 마지막 데이터는 next에 None이 들어있다.
        # start.data는 현재 데이터를 의미하고 start.next는 다음 데이터가 저장된 주소를 의미한다.
        # start.next에 저장된 값이 None이 아니면 True, None이면 False가 된다.
        while start.next is not None: # LinkedList에 저장된 다음 데이터가 있는가?
            # 다음 데이터로 접근한다.
            start = start.next
        # ===== while
        # 다음 데이터가 없으면 새 데이터를 추가한다.
        start.next = newNode
    # ===== appendLast()
    
    # LinkedList의 맨 앞(head 바로 다음 위치)에 데이터를 삽입하는 메소드
    def insertFirst(self, data):
        newNode = Node(data)
        self.count += 1
        
        # LinkedList의 맨 뒤에 데이터를 추가하는 경우를 제외한 나머지 경우(삽입)는 데이터를 입력할 때 반드시 순서를 지켜야 한다.
        # 삽입할 데이터(newNode)가 다음 데이터를 기억하는 next에 head에 저장된 값을 넣어준다.
        newNode.next = self.head
        # head에는 새로 삽입되는 데이터의 주소를 넣어준다.
        self.head = newNode
    # ===== insertFirst()
    
    # LinkedList의 특정(맨 앞과 맨 뒤를 제외한) 위치에 데이터를 삽입하는 메소드
    def insertIndex(self, index, data):
        # 데이터가 삽입될 인덱스가 올바른가 검사한다. => 올바른 인덱스가 아니면 오류 메시지 출력하고 메소드를 종료한다.
        # if index < 1 or index >= self.count:
        if index not in [i for i in range(1, self.count)]:
            print(f'{index}번째 인덱스는 {data} 데이터가 삽입될 위치로 올바르지 않습니다.')
            return
        # ===== if
        
        # 데이터를 삽입할 인덱스가 올바로 지정된 경우면 index 번째 위치에 데이터를 삽입한다.
        newNode = Node(data)
        self.count += 1
        
        # 데이터가 삽입될 바로 전(index - 1 번째)위치의 데이터를 찾는다. => 반복을 index - 1번 시킨다.
        start = self.head
        for i in range(index - 1):
            start = start.next
        # ===== for
        
        # 삽입할 데이터(newNode)의 다음 데이터 주소를 기억하는 next에 데이터가 삽입될 바로 전위치(start) 데이터의 next에 저장된 값을 넣어준다.
        newNode.next = start.next
        # 데이터가 삽입될 바로 전위치(start) 데이터의 next에 삽입할 데이터(newNode)의 주소를 넣어준다.
        start.next = newNode
    # ===== insertIndex()

    # 데이터 보기
    # LinkedList에 저장된 모든 데이터를 출력한다.
    def listPrint(self):
        # LinkedList의 시작 위치를 저장한다.
        start = self.head
        # LinkedList가 비어있나 비어있지않나 판단해서 LinkedList의 데이터를 출력한다.
        # if self.count <= 0:
        if self.head is None:
            print('LinkedList에 저장된 데이터가 없습니다. - listPrint()')
        else:
            print(f'LinkedList에 저장된 데이터가 {self.count}개 있습니다.', end=' => ')
            # LinkedList에 저장된 데이터의 개수만큼 반복하며 저장된 데이터를 출력한다.
            for i in range(self.count):
                print(start.data, end=' ')
                # 데이터 1건을 출력했으므로 다음 데이터로 접근한다.
                start = start.next
            # ===== for
            print()
        # ===== if
    # ===== listPrint()

    # 제거
    # LinkedList에 저장된 데이터를 찾아서 제거하는 메소드
    def remove(self, data):
        # 삭제할 데이터를 LinkedList에서 찾아야하므로 시작 위치를 지정한다.
        start = self.head
        if self.head is None:
            print('LinkedList에 저장된 데이터가 없습니다. - remove()')
        else:
            # LinkedList에 저장된 데이터가 있으므로 데이터를 찾아 제거한다.
            # 제거할 데이터가 0번째(head 바로 다음) 위치의 데이터일 경우
            if start.data == data: # 0번째 위치의 데이터가 제거할 데이터인가?
                # 0번째 위치한 데이터의 next를 head에 넣어준다.
                self.head = start.next
                # LinkedList에서 데이터를 제거했으므로 데이터 개수를 감소시킨다.
                self.count -= 1
                return
            # ===== if
            
            # 제거할 데이터가 없거나 1번째 위치 이후의 데이터일 경우
            while start is not None:
                if start.data == data: # 제거할 데이터를 찾았는가?
                    break # 삭제할 데이터를 찾았으면 삭제할 데이터를 찾는 반복을 탈출한다.
                # ===== if
                prev = start # 다음 데이터로 이동하기 전에 삭제할 위치 바로 전 데이터를 저장한다.
                start = start.next # 다음 데이터로 이동한다.
            # ===== while
            
            # while 명령이 정상적으로 종료되서 start가 None인 경우는 제거하려는 데이터가 LinkedList에 없음을 의미한다.
            if start == None: # 삭제할 데이터가 LinkedList에 존재하지 않는가?
                print(f'{data}는(은) LinkedList에 존재하지 않는 데이터입니다.')
                return
            # ===== if
            
            # while 명령이 break를 만나서 종료된 경우는 제거하려는 데이터가 LinkedList에 있음을 의미한다.
            # 삭제하려는 데이터 이전 데이터의 next에 삭제하려는 데이터의 next를 넣어준다.
            prev.next = start.next
            self.count -= 1
```

```python
print('LinkedList를 만든다.')
linkedList = LinkedList()
linkedList.listPrint()
linkedList.remove('손오공')
print('=' * 80)

print('LinkedList의 맨 뒤(head 바로 다음)에 데이터를 추가한다.')
linkedList.appendLast('홍길동')
linkedList.listPrint()
print('=' * 80)

print('LinkedList의 맨 뒤에 데이터를 추가한다.')
linkedList.appendLast('임꺽정')
linkedList.listPrint()
linkedList.appendLast('장길산')
linkedList.listPrint()
linkedList.appendLast('일지매')
linkedList.listPrint()
print('=' * 80)

print('LinkedList의 맨 앞(head 바로 다음)에 데이터를 삽입한다.')
linkedList.insertFirst('손오공')
linkedList.listPrint()
print('=' * 80)

print('LinkedList의 맨 앞과 맨 뒤를 제외한 위치에 데이터를 삽입한다.')
linkedList.insertIndex(0, '저팔계')
linkedList.insertIndex(5, '사오정')
linkedList.insertIndex(3, '삼장법사')
linkedList.listPrint()
print('=' * 80)

print('LinkedList의 0번째 위치에 데이터를 제거한다.')
linkedList.remove('손오공')
linkedList.listPrint()
print('=' * 80)

print('LinkedList의 1번째 위치 이후의 데이터를 제거한다.')
linkedList.remove('장길산')
linkedList.listPrint()
print('=' * 80)

print('LinkedList에 없는 데이터를 제거하려고 시도한다.')
linkedList.remove('피콜로')
linkedList.listPrint()
print('=' * 80)
```
