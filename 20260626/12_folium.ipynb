folium 라이브러리를 사용하면 지도를 이용한 시각화를 할 수 있다.  
folium 라이브러리를 설치하고 import 한다.

```python
# !pip install folium
import folium
```

folium 라이브러리의 Map 클래스로 location 속성과 zoom_start 속성을 지정해서 지도를 만든다.  
location 속성으로 위도와 경도를 지정해서 지도의 중심을 설정하고 zoom_start 속성으로 지도의 배율을 설정한다.

```python
# folium 라이브러리를 사용해서 지도를 만든다.
foliumMap = folium.Map(location=[37.44508047, 127.13906206], zoom_start=17)

# folium 라이브러리의 Marker 클래스로 location 속성을 지정해서 마커를 만들고 add_to() 메소드로 지도에 추가한다.
# popup 속성으로 마커를 클릭했을 때 표시할 문자열을 추가할 수 있다.
# folium.Marker(location=[37.442561275669, 127.136588869764], popup='스타벅스 성남 태평점').add_to(foliumMap)

# 마커를 클릭했을 때 표시되는 문자열이 나타나는 말풍선 크기가 마음에 들지 않으면 folium 라이브러리의 Popup 클래스로 말풍선에 나타낼 문자열과 최대 가로 크기를
# 지정할 수 있다.
popup = folium.Popup('스타벅스 성남 태평점', max_width=300)
folium.Marker(location=[37.442561275669, 127.136588869764], popup=popup).add_to(foliumMap)
popup = folium.Popup('스타벅스 가천대학교점', max_width=300)
folium.Marker(location=[37.44941, 127.1272], popup=popup).add_to(foliumMap)

# save() 메소드로 folium 라이브러리를 사용해서 작성한 지도를 html 파일로 저장할 수 있다.
foliumMap.save('./output/foliumMap.html')
# folium 라이브러리를 사용해서 만든 지도를 출력한다.
foliumMap
```
