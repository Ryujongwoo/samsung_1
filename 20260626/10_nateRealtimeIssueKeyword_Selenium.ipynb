selenium 4.6 버전 이후로 업데이트되면서, 크롬 버전을 확인하고 드라이버를 다운받으 수동으로 경로를 지정하는 번거로운 과정이 사라졌다.

driver = webdriver.Chrome() => 현재 사용방법

```python
# selenium을 사용해서 가상 크롬을 실행한다.
driver = webdriver.Chrome()
# 가상 크롬 드라이버에 get() 메소드의 인수로 크롤링할 타겟 사이트를 지정한다.
driver.get('https://www.nate.com')
# 가상 크롬에 실행된 사이트의 html 정보를 읽어온다.
html = driver.page_source
# print(type(html))
# print(html)
soup = BeautifulSoup(html, 'html.parser')
# print(soup)

ranks = soup.select('#olLiveIssueKeyword .num_rank')
issues = soup.findAll('a', {'class', 'ik'})

for i in range(len(issues)):
    rank = ranks[i].text
    issue = issues[i].text.strip().split('\n')[0]
    upDown = issues[i].text.strip().split('\n')[1]
    print('{:>2s}위 {}'.format(rank, issue), end=' ')
    if upDown[:2] == '상승':
        print(f'[↑{upDown[-1]}]')
    elif upDown[:2] == '하락':
        print(f'[↓{upDown[-1]}]')
    elif upDown[:2] == '동일':
        print('[-]')
    else:
        print('[new]')
```
