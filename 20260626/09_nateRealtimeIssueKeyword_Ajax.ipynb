```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
# BeautifulSoup은 html 문서를 파싱할때 사용하기때문에 json 형태로 넘어오는 데이터에는 사용하지 않는다.
# from bs4 import BeautifulSoup
# json 형태를 갖춘 문자열을 파이썬에서 처리할 수 있도록 리스트나 딕셔너리 형태로 변경하기위해서 import 한다.
import json
```

네이트의 실시간 이슈 키워드가 웹 페이지에 전부 혹은 일부가 포함되어있지않고 Ajax를 이용해서 실시간으로 받아오는 방식이 사용된다.

```python
targetSite = 'https://www.nate.com/js/data/jsonLiveKeywordDataV1.js?v=202606261030'
request = requests.get(targetSite)
print(request)
# 파이썬은 한글 인코딩이 유니코드(utf-8)로 되어있기때문에 'euc-kr' 방식으로 인코딩된 사이트를 크롤링할 때 한글이 깨지면 아래와 같이 인코딩 방식을 지정한다.
# request.encoding = 'euc-kr'
```

```python
issues = request.text
print(issues)
print(type(issues))
```

json 형태의 문자열로 얻어온 실시간 이슈 데이터를 파이썬이 처리할 수 있는 데이터 타입으로 변환해서 처리한다.

json 모듈의 loads() 메소드는 json 형태의 문자열이 {}를 포함하면 딕셔너리로 {}를 포함하지 않으면 리스트로 변환한다.

```python
ranks = json.loads(issues)
print(ranks)
print(type(ranks))
```

requests 모듈을 이용해서 서버에서 응답받은 객체에서 json() 메소드를 실행해도 json 모듈의 loads() 메소드를 실행한 것과 같은 결과를 얻을 수 있다.

```python
ranks = request.json()
print(ranks)
print(type(ranks))
```

```python
for rank in ranks:
    print('{:>2s}위 {}'.format(rank[0], rank[1]), end=' ')
    if rank[2] == '+':
        print(f'[↑{rank[3]}]')
    elif rank[2] == '-':
        print(f'[↓{rank[3]}]')
    elif rank[2] == 'n':
        print('[new]')
    else:
        print('[-]')
```
