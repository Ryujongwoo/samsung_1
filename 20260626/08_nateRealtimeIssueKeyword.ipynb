```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
```

```python
targetSite = 'https://www.nate.com/'
headers = {'User-agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64; Trident/7.0; MSAppHost/2.0; rv:11.0) like Gecko'}
request = requests.get(targetSite, headers=headers)
# print(request)
html = request.text
soup = BeautifulSoup(html, 'html.parser')
# print(soup)

# 순위
# ranks = soup.findAll('span', {'class', 'num_rank'})
ranks = soup.select('#olLiveIssueKeyword .num_rank')
# print(len(ranks))
# for rank in ranks:
    # print(rank.text)
    
# 실시간 이슈 키워드
issues = soup.findAll('span', {'class', 'txt_rank'})
# issues = soup.select('#olLiveIssueKeyword .txt_rank')
# print(len(issues))
# for issue in issues:
    # print(issue.text)
    
# 실시간 이유 키워드와 상승/하락 폭
issues = soup.findAll('a', {'class', 'ik'})
# print(len(issues))
# for issue in issues:
    # print(issue.text.strip().split('\n'))
    # print(issue.text.strip().split('\n')[0], issue.text.strip().split('\n')[1])
    
for i in range(len(issues)):
    rank = ranks[i].text
    issue = issues[i].text.strip().split('\n')[0]
    upDown = issues[i].text.strip().split('\n')[1]
    # print('{:>2s}위 {}'.format(rank, issue))
    print('{:>2s}위 {}'.format(rank, issue), end=' ')
    if upDown[:2] == '상승':
        print(f'[↑{upDown[-1]}]')
    elif upDown[:2] == '하락':
        print(f'[↓{upDown[-1]}]')
    elif upDown[:2] == '동일':
        print('[-]')
    else:
        print('[new]')
```
