```python
import random
```

```python
# 로또 1등 번호를 기억할 빈 set을 만든다.
lotto = set()

# 1등 번호
# 중복되는 번호가 나올수있기 때문에 난수 발생시키는 작업을 몇 번 반복해야하는지 알 수 없으므로 while 반복을 무한루프를 돌려서 처리한다.
while True:
    # 1 ~ 45 사이의 난수를 정수로 발생시킨다.
    number = random.randint(1, 45)
    # print(number)
    # 리스트는 중복을 확인하는 if가 필요했지만 set은 중복되는 데이터를 허용하지 않기때문에 중복을 확인하는 if가 필요없다.
    lotto.add(number)
    # set에 중복되지않는 숫자가 6개 저장되면 무한루프를 탈출시킨다.
    if len(lotto) == 6:
        break
    # ===== if
# ===== while
print('1등 번호: {}'.format(lotto))

# 보너스 번호
while True:
    bonus = random.randint(1, 45)
    if bonus not in lotto:
        break
    # ===== if
# ===== while
print('1등 번호: {}, 보너스 번호: {}'.format(lotto, bonus))
```
