리스트와 함께 사용하는 for

```python
a = ['Life', 'is', 'too', 'Good']
print(a)
print('=' * 80)

for i in a:
    print(i)
print('=' * 80)

for i in range(len(a)):
    print(a[i])
print('=' * 80)

for i in a[1:3]:
    print(i)
```

딕셔너리와 함께 사용한 for

```python
a = {'apple': 1000, 'banana': 2000, 'melon': 10000, 'mango': 5000}
print(a)
print(a.keys())
print(a.values())
print(a.items())
```

```python
# 딕셔너리 이름만 쓰면 딕셔너리이름.key()와 같은 의미로 사용된다.
for key in a:
    print(key)
print('=' * 80)

for key in a.keys():
    print(key)
print('=' * 80)

for value in a.values():
    print(value)
print('=' * 80)

for item in a.items():
    print(item)
```

key를 이용해서 value 얻어오기

```python
for key in a.keys():
    print(a[key]) # 딕셔너리이름[key]
print('=' * 80)

for key in a.keys():
    print(a.get(key)) # 딕셔너리이름.get(key)
```

item을 이용해서 key, value 얻어오기

```python
for item in a.items():
    print(item)
print('=' * 80)

for item in a.items():
    print('{}는(은) {:,}원 입니다.'.format(item[0], item[1]))
print('=' * 80)

for key, value in a.items():
    print('{}는(은) {:,}원 입니다.'.format(key, value))
```

enumerate() 함수는 인수로 지정된 리스트, 튜플, set, 딕셔너리의 index와 데이터를 쌍으로 묶은 튜플을 리턴한다.

```python
a = ['Life', 'is', 'too', 'Good']
print(enumerate(a))
print(list(enumerate(a)))

for index, data in enumerate(a):
    print(index, data)
```

zip() 함수는 서로 다른 데이터가 저장된 객체를 반복 가능한 형태로 묶는다.

```python
a = ['java', 'jsp', 'python']
b = [100, 95, 98]

print(zip(a, b))
print(list(zip(a, b)))

for i in zip(a, b):
    print(i)
```

```python
java = [100, 100, 100]
jsp = [90, 80, 70]
python = [1, 2, 3]

total = []
for i in range(len(java)):
    total.append(java[i] + jsp[i] + python[i])
print(total)

total = []
for a, b, c in zip(java, jsp, python):
    total.append(a + b + c)
print(total)
```
