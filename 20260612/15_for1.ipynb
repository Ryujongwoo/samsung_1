range() 함수

range() 함수는 지정한 초기치, 최종치, 증가치에 따른 숫자 리스트를 만든다.  
range() 함수는 초기치부터 증가치만큼 증가하며 최종치를 넘어가지 않는 숫자 리스트를 만든다.

[] 안의 내용은 필요에 따라서 생략할 수 있다.  
range([초기치, ]최종치[, 증가치])  
초기치와 증가치는 생략할 수 있고 초기치는 생략시 0을 증가치는 생략시 1을 기본값으로 사용한다.  
즉, range(0, 10, 1)와 range(10)는 같은 표현이다.

```python
print(range(0, 10, 1))
# range() 함수가 실행된 결과를 리스트로 확인하려면 list() 생성자 함수를 사용해서 확인해야 한다.
print(list(range(0, 10, 1)))
print(list(range(10)))
print(list(range(0, 10, 2)))
print(list(range(1, 10, 2)))
print(list(range(9, -1, -1)))
```

```python
print(list(range(5)))
for i in range(5):
    print(i)
for i in [0, 1, 2, 3, 4]:
    print(i)
```

```python
i = 0
total = 0
while i < 10:
    i += 1
    total += i
# ===== while
print(total)
```

```python
total = 0
for i in range(1, 11, 1):
    total += i
# ===== for
print(total)
```

주사위를 10번 굴려서 출현한 눈의 개수를 센다. => 변수 사용

```python
import random
```

```python
# 1부터 6사이의 주사위 눈의 개수를 기억할 변수 6개를 선언하고 0으로 초기화시킨다.
number_1 = 0; number_2 = 0; number_3 = 0; number_4 = 0; number_5 = 0; number_6 = 0
# 주사위를 10번 굴린다.
for i in range(10):
    dice = random.randint(1, 6)
    print(dice)
    # 주사위를 굴려서 발생한 눈의 개수를 증가시킨다.
    if dice == 1:
        number_1 += 1
    elif dice == 2:
        number_2 += 1
    elif dice == 3:
        number_3 += 1
    elif dice == 4:
        number_4 += 1
    elif dice == 5:
        number_5 += 1
    else:
        number_6 += 1
    # ===== if
# ===== for

# 주사위 눈의 개수를 출력한다.
print('1의 개수: {}'.format(number_1))
print('2의 개수: {}'.format(number_2))
print('3의 개수: {}'.format(number_3))
print('4의 개수: {}'.format(number_4))
print('5의 개수: {}'.format(number_5))
print('6의 개수: {}'.format(number_6))
```

주사위를 10번 굴려서 출현한 눈의 개수를 센다. => 리스트 사용

```python
# 1. 주사위 눈의 개수를 기억할 크기가 6인 0으로 초기화된 리스트를 만든다.
# number = [0, 0, 0, 0, 0, 0]
# number = [0] * 6
# 2. 빈 리스트를 선언하고 append() 메소드로 0을 넣어준다.
# number = []
# for i in range(6):
    # number.append(0)
# 3. 컴프리핸션을 사용해서 리스트를 만든다.
number = [0 for i in range(6)]
print(number)

for i in range(10):
    dice = random.randint(1, 6)
    print(dice)
    # 주사위를 굴려서 발생한 눈의 개수를 증가시킨다.
    if dice == 1:
        number[0] += 1
    elif dice == 2:
        number[1] += 1
    elif dice == 3:
        number[2] += 1
    elif dice == 4:
        number[3] += 1
    elif dice == 5:
        number[4] += 1
    else:
        number[5] += 1
    # ===== if
# ===== for

# 주사위 눈의 개수를 출력한다.
print('1의 개수: {}'.format(number[0]))
print('2의 개수: {}'.format(number[1]))
print('3의 개수: {}'.format(number[2]))
print('4의 개수: {}'.format(number[3]))
print('5의 개수: {}'.format(number[4]))
print('6의 개수: {}'.format(number[5]))
```

```python
number = [0 for i in range(6)]
for i in range(10):
    dice = random.randint(1, 6)
    # print(dice)
    number[dice - 1] += 1
    # ===== if
# ===== for

for i in range(6):
    print('{}의 개수: {}'.format(i + 1, number[i]))
```
