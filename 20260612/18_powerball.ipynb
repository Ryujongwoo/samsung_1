```python
import random
import time
```

```python
lotto = [i for i in range(1, 46)]
print(lotto)
# sample(리스트, 개수) 메소드는 데이터 중에서 무작위로 지정한 개수만큼 뽑아낸다.
lotto = random.sample(lotto, 7)
print(lotto)

print('1등 번호: ', end='')
for i in range(6):
    time.sleep(1)
    print('{:02d}'.format(lotto[i]), end=', ')
# ===== for
time.sleep(1)
print('보너스 번호: {:02d}'.format(lotto[6]))
```

```python
# seed() 메소드를 이용하면 매번 같은 난수 배열을 얻을 수 있다.
random.seed(3)
lotto = [i for i in range(1, 46)]
print(random.sample(lotto, 6))
```

```python
# powerball: 1 ~ 69 => 5개, 1 ~ 26 => 1개
powerball = [i for i in range(1, 70)]
powerball = random.sample(powerball, 5)
print('흰공: ', end='')
for i in powerball:
    # time.sleep(1)
    print('{:02d}'.format(i), end=', ')
# ===== for
# time.sleep(1)
print('파워볼: {:02d}'.format(random.randint(1, 26)))
```
