```python
import random
import time
```

```python
%%time
# %%time 매직 명령은 jupyter nbclassic 한 셀이 실행되는 시간을 출력한다.

# 로또 추첨기로 사용할 리스트를 만들고 1 ~ 45로 초기화시킨다.
# lotto = []
# for i in range(45):
    # lotto.append(i + 1)
# lotto = [i + 1 for i in range(45)]
lotto = [i for i in range(1, 46)]
# print(lotto)

# 섞기전 상태를 출력한다. => 1줄에 10개씩 출력한다.
for i in range(len(lotto)):
    print('{:02d}'.format(lotto[i]), end=' ')
    if (i + 1) % 10 == 0:
        print()
    # ===== if
# ===== for
print('\n' + '=' * 30 + ' 섞기전')

# 섞는다.
'''
# lotto[0]와(고정) lotto[1] ~ lotto[44] 중에서 랜덤한 위치의 값을 선택해서 두 기억장소의 값을 교환한다.
for i in range(1000000):
    # 1 ~ 44 중에서 랜덤한 값을 발생시킨다.
    r = random.randint(1, 44)
    # lotto[0]와 lotto[r]에 저장된 값을 교환한다.
    # temp = lotto[0]
    # lotto[0] = lotto[r]
    # lotto[r] = temp
    lotto[0], lotto[r] = lotto[r], lotto[0]
# ===== for
'''
# shuffle(리스트) 메소드로 데이터를 섞을 수 있다. => 반복 처리보다 속도가 월등히 빠르다.
random.shuffle(lotto)

# 섞은후 상태를 출력한다. => 1줄에 10개씩 출력한다.
for i in range(len(lotto)):
    print('{:02d}'.format(lotto[i]), end=' ')
    if (i + 1) % 10 == 0:
        print()
    # ===== if
# ===== for
print('\n' + '=' * 30 + ' 섞은후')

# 앞에서 6개가 1등 번호이고 6번째 인덱스의 값이 보너스 번호이다.
print('1등 번호: ', end='')
for i in range(6):
    time.sleep(1)
    print('{:02d}'.format(lotto[i]), end=', ')
# ===== for
time.sleep(1)
print('보너스 번호: {:02d}'.format(lotto[6]))
```
