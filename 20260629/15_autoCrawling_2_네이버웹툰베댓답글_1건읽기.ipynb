```python
import warnings
warnings.filterwarnings(action='ignore')
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
# selenium 4 버전의 'By'를 사용하기 위해 import 한다.
from selenium.webdriver.common.by import By
import time
```

find_element_by_class_name('') => find_element(By.CLASS_NAME, '')  
find_element_by_id('') => find_element(By.ID, '')  
find_element_by_name('') => find_element(By.NAME, '')  
find_element_by_link_text('') => find_element(By.LINK_TEXT, '')  
find_element_by_partial_link_text('') => find_element(By.PARTIAL_LINK_TEXT, '')  
find_element_by_tag_name('') => find_element(By.TAG_NAME, '')  
find_element_by_css_selector('') => find_element(By.CSS_SELECTOR, '')  
find_element_by_xpath('') => find_element(By.XPATH, '')  

```python
option = Options()
option.add_argument('--window-size=1800,1000')
driver = webdriver.Chrome(options=option)
driver.get('https://comic.naver.com/webtoon/detail?titleId=812354&no=237&week=thu')
driver.implicitly_wait(5)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
reviews = soup.select('p._text_8n8z9_16')
print(len(reviews))
print('{:2d}. {}'.format(3, reviews[2].text))
print('=' * 100)

# 답글 버튼을 찾아서 클릭한다.
# driver.find_element_by_xpath('//*[@id="wcc_root"]/section/ul/li[1]/div[2]/div[1]/button').click() # selenium 3 버전에서 사용하던 방법
driver.find_element(By.XPATH, '//*[@id="wcc_root"]/section/ul/li[1]/div[2]/div[1]/button').click()
# driver.implicitly_wait(5)
time.sleep(1)

'''
while True:
    try:
        driver.find_element(By.XPATH, '//*[@id="wcc_root"]/section/ul/li[1]/div[2]/div[0]/button').click()
        driver.implicitly_wait(5)
    except:
        print('에러')
        break
'''

# 답글 버튼을 찾아서 클릭하면 가상 크롬의 화면이 변경되었으므로 다시 읽어온다.
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
replys = soup.select('div.Ug1Ok p._text_8n8z9_16')
print(len(replys))
for i in range(len(replys)):
    print('{:2d}. {}'.format(i + 1, replys[i].text.split('\n')))

# 답글 접기 버튼을 찾아서 클릭한다.
driver.find_element(By.XPATH, '//*[@id="wcc_root"]/section/ul/li[1]/div[3]/button').click()
time.sleep(1)
```
