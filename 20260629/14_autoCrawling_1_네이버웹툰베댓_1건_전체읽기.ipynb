```python
import warnings
warnings.filterwarnings(action='ignore')
from bs4 import BeautifulSoup
from selenium import webdriver
# selenium option을 사용하기 위해 import 한다.
from selenium.webdriver.chrome.options import Options
import time
```

```python
option = Options()
option.add_argument('--window-size=1800,1000') # 가상 크롬의 크기를 지정한다.

driver = webdriver.Chrome(options=option)
driver.get('https://comic.naver.com/webtoon/detail?titleId=812354&no=237&week=thu')
# 크롤링할 사이트가 가상 크롬에 최초 로딩되기까지 대기시간을 지정한다.
driver.implicitly_wait(5) # 지정한 시간이 모두 경과되지 않았더라도 크롤링할 페이지가 다 로딩되면 기다리지 않고 다음으로 넘어간다.
# 이미 가상 크롬에 로딩된 페이지의 내용이 변경될 때 대기시간을 지정한다.
# time.sleep(5) # 가상 크롬에 크롤링할 페이지가 다 로딩되었더라고 지정된 시간만큼 기다린다.


html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
review = soup.select_one('#wcc_root > section > ul > li:nth-child(1) > div:nth-child(1) > div._root_8n8z9_1.tcKGz > div > p')
# print(type(review))
print(review.text)
print('=' * 100)

reviews = soup.select('p._text_8n8z9_16')
# print(type(reviews))
# print(reviews.text) # ResultSet(결과를 모아놓은) 객체는 text 속성을 사용할 수 없으므로 에러가 발생된다.
for i in range(len(reviews)):
    # print(type(reviews[i]))
    print('{:2d}. {}'.format(i + 1, reviews[i].text))
```
