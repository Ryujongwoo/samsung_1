numpy 데이터 타입

numpy는 다음과 같은 데이터 타입을 지원한다. => numpy 배열을 만들때 dtype 속성으로 데이터 타입을 지정할 수 있다.  
np.int: 정수(고정 소수점) 타입  
np.float: 실수(부동 소수점) 타입  
np.bool: 불린(논리값) 타입  
np.complex: 복소수 타입  
np.object: 파이썬 객체 타입  
np.string: 문자열 타입  
np.unicode: 유니코드 타입  

numpy는 다차원 배열을 지원하고 numpy 배열 구조는 shape으로 표현한다.  
shape은 배열의 구조를 튜플 자료형을 사용해서 정의한다.  
예를들면, 28 * 28 컬러 사진은 폭이 28픽셀 높이가 28픽셀, 각 픽셀은 3개의 채널(RGB) 구조를 가진다. 따라서 컬러 사진의 데이터 구조는 shape(28, 28, 3)인 3차원 배열이 된다.  
다차원 배열은 입체적인 구조를 가지며 데이터 차원은 여러 갈래의 데이터 방향(axis)을 가진다.  
3차원 배열의 데이터 방향은 면은 axis=0, 행은 axis=1, 열은 axis=2로 지정한다.

numpy 배열 객체 정보 출력용으로 사용할 함수 선언

```python
def pprint(arr):
    print(f'type: {type(arr)}')
    print(f'shape: {arr.shape}, dimension: {arr.ndim}, dtype: {arr.dtype}')
    print('numpy 배열의 데이터\n', arr, sep='')
```

파이썬 리스트로 numpy 배열을 생성할 수 있다.

```python
# 파이썬 1차원 리스트로 numpy 배열 만들기
arr = [1, 2, 3]
print(arr)
print(type(arr))
print('=' * 100)
# numpy의 array() 메소드는 인수로 리스트와 데이터 타입을 넘겨받아 numpy 배열을 만든다.
# 데이터 타입은 생략 가능하고 기본값으로 정수는 np.int64가 사용된다.
# 정수형 데이터의 dtype은 np.int8(-256 ~ 255), np.int16(-32768 ~ 32767), np.int32(-2147483648 ~ 2147483647), np.int64를 사용할 수 있다.
a = np.array(arr)
pprint(a)
print('=' * 100)
b = np.array(arr, dtype=np.int64)
pprint(b)
print('=' * 100)
c = np.array(arr, dtype=np.int32)
pprint(c)
```

```python
arr = [1, 2., 3]
print(arr)
print(type(arr))
print('=' * 100)
# 데이터 타입은 생략 가능하고 기본값으로 실수는 np.float64가 사용된다.
# 실수형 데이터의 dtype은 np.float16, np.float32, np.int64를 사용할 수 있다.
a = np.array(arr)
pprint(a)
print('=' * 100)
b = np.array(arr, dtype=np.float64)
pprint(b)
print('=' * 100)
c = np.array(arr, dtype=np.float16)
pprint(c)
print('=' * 100)
```

```python
# 파이썬 2차원 리스트로 numpy 배열 만들기
arr = [[1, 2, 3], [4, 5, 6]]
print(arr)
print(type(arr))
print('=' * 100)
a = np.array(arr, dtype=np.float64)
pprint(a)
```

```python
# 파이썬 3차원 리스트로 numpy 배열 만들기
arr = [[[1, 2, 3], [4, 5, 6]], [[11, 22, 33], [44, 55, 66]]]
print(arr)
print(type(arr))
print('=' * 100)
a = np.array(arr, dtype=np.float64)
pprint(a)
```

numpy는 원하는 shape으로 배열을 생성하고 각 요소들을 특정 값으로 초기화시키는 메소드가 제공된다.

```python
# zeros() 메소드는 지정한 shape 만큼 배열을 생성하고 모든 요소를 0으로 초기화시킨다. => 영행렬
# zeros(shape[, dtype, order]) # dtype, order는 생략 가능하다.
# order: 2차원 이상의 배열에서 배열에 데이터가 채워지는 방식을 지정한다.
# 행 우선 방식: 대부분의 프로그래밍 언어에서 채택, c/c++, java, c#, python, ...
# 열 우선 방식: fortran
a = np.zeros(shape=(3, 4))
pprint(a)
print('=' * 100)
b = np.zeros(shape=(3, 4), dtype=np.int32) # order 생략시 행 우선 방식을 사용한다.
pprint(b)
print('=' * 100)
c = np.zeros(shape=(3, 4), dtype=np.int32, order='f') # order='f' 속성을 지정하면 열 우선 방식이 사용된다.
pprint(c)
```

```python
# ones() 메소드는 지정한 shape 만큼 배열을 생성하고 모든 요소를 1로 초기화시킨다.
# ones(shape[, dtype, order])
a = np.ones(shape=(3, 4))
pprint(a)
```

```python
# full() 메소드는 지정한 shape 만큼 배열을 생성하고 모든 요소를 특정값으로 초기화시킨다.
# full(shape, fill_value[, dtype, order])
a = np.full(shape=(3, 4), fill_value=7)
pprint(a)
```

```python
# eye() 메소드는 지정한 shape 만큼 크기를 가지는 단위행렬(주 대각 요소 1이고 나머지 요소는 0인 정방행렬)을 만든다.
# eye(n[, dtype])
a = np.eye(5)
pprint(a)
```

```python
# empty() 메소드는 지정한 shape 만큼 배열을 생성하고 초기화시키지 않는다. => 쓰레기가 저장된다.
# empty(shape[, dtype, order])
a = np.empty(shape=(3, 4))
pprint(a)
```

numpy는 인수로 지정한 배열과 shape이 같은 배열을 만드는 *_like() 메소드를 제공한다.  
zeros_like(), ones_like(), full_like(), empty_like()

```python
a = [[1, 2, 3], [4, 5, 6]]
print(a)
print('=' * 100)
b = np.zeros_like(a)
print(b)
print('=' * 100)
c = np.ones_like(a)
print(c)
print('=' * 100)
d = np.full_like(a, fill_value=9)
print(d)
print('=' * 100)
e = np.empty_like(a)
print(e)
```

데이터 시각화를 위해서 matplotlib 라이브러리를 설치하고 import 한다.

```python
# !pip install matplotlib
import matplotlib.pyplot as plt
```

데이터 생성 메소드

```python
# linspace() 메소드는 start 부터 stop 사이의 범위를 num개의 균일한 간격으로 데이터를 생성하고 배열을 만든다.
# linspace(start, stop, num)
a = np.linspace(0, 1, 5)
pprint(a)

plt.figure(figsize=(4, 3))
plt.plot(a, 'ro')
plt.show()
```

```python
# arange() 메소드는 start 부터 stop이 되기전까지 step 간격으로 데이터를 생성하고 배열을 만든다.
# arange([start, ]stop[step, dtype])
# linspace() 메소즈는 요소의 개수를 지정해서 배열을 만들지만 arange() 메소드는 요소의 개수가 아닌 데이터의 간격을 기준으로 배열을 만든다.
# start를 생략하면 0이 step을 생략하면 1이 기본값으로 사용된다.
# arange(10), arange(0, 10), arange(0, 10, 1)은 모두 같은 기능이 실행된다.
a = np.arange(1, 10, 2)
pprint(a)

plt.figure(figsize=(4, 3))
plt.plot(a, 'r*')
plt.show()
```

```python
# logspace() 메소드는 로그 스케일 start 부터 end 사이의 범위에서 num개를 균일한 간격으로 데이터를 생성하고 배열을 만든다.
# logspace(start, end, num[, base]) # base를 생략하면 10이 기본값으로 사용된다.
a = np.logspace(1, 10, 100, base=2)
pprint(a)

plt.figure(figsize=(4, 3))
plt.plot(a, 'r')
plt.show()
```

난수 기반 배열 생성

numpy 난수 발생 및 배열을 생성하는 random이라는 모듈을 제공한다.

```python
# normal() 메소드는 정규분포를 따르는 난수 데이터를 생성하고 배열을 만든다.
# normal(평균, 표준편차, 난수의 개수)
# 평균이 0이고 표준편차가 1인(표준 정규 분포) 정규 분포를 따르는 난수를 발생시킨다.
a = np.random.normal(0, 1, 10)
pprint(a)

plt.figure(figsize=(4, 3))
plt.hist(a)
plt.show()
```

```python
a = np.random.normal(3, 2.5, 1000000)
pprint(a)

plt.figure(figsize=(4, 3))
plt.hist(a, bins=500) # bins 옵션으로 히스토그램 막대(구간)의 개수를 지정할 수 있다.
plt.show()
```

```python
# rand() 메소드는 0 ~ 1 사이의 균등분포를 따르는 난수 데이터(실수)를 생성하고 배열을 만든다.
# rand(난수의 개수)
a = np.random.rand(10)
pprint(a)

plt.figure(figsize=(4, 3))
plt.hist(a)
plt.show()
```

```python
a = np.random.rand(1000000)
pprint(a)

plt.figure(figsize=(4, 3))
plt.hist(a, bins=500)
plt.show()
```

```python
# randint() 메소드는 low ~ high 사이의 균등분포를 따르는 난수 데이터(정수)를 생성하고 배열을 만든다.
# randint(low, high, 난수의 개수)
a = np.random.randint(1, 46, 6)
pprint(a)

plt.figure(figsize=(4, 3))
plt.hist(a)
plt.show()
```

```python
a = np.random.randint(-100, 100, 1000000)
pprint(a)

plt.figure(figsize=(4, 3))
plt.hist(a, bins=100)
plt.show()
```

```python
# random() 메소드는 0 ~ 1 사이의 균등분포를 따르는 난수 데이터(실수)를 생성하고 배열을 만든다.
# random(low, high, 난수의 개수)
a = np.random.random(10)
pprint(a)

plt.figure(figsize=(4, 3))
plt.hist(a)
plt.show()
```

```python
a = np.random.random(1000000)
pprint(a)

plt.figure(figsize=(4, 3))
plt.hist(a, bins=400)
plt.show()
```

약속된 난수

난수 발생 메소드는 특정 시작 숫자로부터 난수처럼 보이는 수열을 만드는 알고리즘의 결과이다.  
따라서 시작점을 설정하면 난수 발생을 재연할 수 있다. => 매번 같은 배열의 난수가 발생되게 할 수 있다.

```python
# seed() 메소드는 매번 같은 배열의 난수를 얻을수있게 시작 숫자를 고정한다.
np.random.seed(100)
print(np.random.normal(0, 1, 5)) # 평균과 표준편차를 지정한 정규분포를 따르는 난수
print(np.random.randn(5)) # 표준 정규분포(평균이 0, 표준편차 1)를 따르는 난수
print(np.random.rand(5)) # 0 ~ 1 사이의 균등분포를 따르는 난수
print(np.random.randint(1, 10, 5)) # 두 정수 사이의 균등분포를 따르는 난수
print(np.random.random(5)) # 0 ~ 1 사이의 균등분포를 따르는 난수
```

```python
# rand() 메소드와 random() 메소드는 1차원 난수를 발생시킬때는 똑같이 사용한다.
print(np.random.rand(5))
print(np.random.random(5))
print('=' * 100)
# 다차원 난수를 발생시켜야 할 경우 rand() 메소드는 shape만 적어주면 되지만 random() 메소드는 shape을 반드시 튜플로 적어줘야 한다.
print(np.random.rand(3, 4))
print(np.random.random((3, 4)))
```

numpy 배열 상태 검사

numpy는 배열의 상태를 검사하는 다음과 같은 방법을 제공한다.

```python
arr = np.random.rand(2, 3, 6)
pprint(arr)
```

```python
# 파이썬의 type() 함수를 사용하면 numpy 배열의 테이터 타입을 확인할 수 있다.
print(f'배열의 타입: {type(arr)}')
```

```python
# shape 속성으로 numpy 배열의 shape을 확인할 수 있다.
print(f'배열의 shape: {arr.shape}')
```

```python
# 파이썬의 len() 함수를 인덱싱과 적절히 섞어 사용하면 다양한 배열의 길이를 확인할 수 있다.
print(f'배열의 길이(면): {len(arr)}')
print(f'배열의 길이(행): {len(arr[0])}')
print(f'배열의 길이(열): {len(arr[0][0])}')
print('=' * 100)
# shape 속성과 인덱싱을 적절히 섞어 사용하면 다양한 배열의 길이를 확인할 수 있다.
print(f'배열의 길이(면): {arr.shape[0]}')
print(f'배열의 길이(행): {arr.shape[1]}')
print(f'배열의 길이(열): {arr.shape[2]}')
```

```python
# ndim 속성으로 numpy 배열의 차원을 확인할 수 있다.
print(f'배열의 차원: {arr.ndim}')
```

```python
# dtype 속성으로 numpy 배열의 데이터 타입을 확인할 수 있다.
print(f'배열의 데이터 타입: {arr.dtype}')
print(f'배열의 데이터 타입: {arr.dtype.name}')
```

```python
# astype() 메소드로 numpy 배열의 데이터 타입을 변경할 수 있다.
arr = arr.astype(np.int32)
pprint(arr)
```
