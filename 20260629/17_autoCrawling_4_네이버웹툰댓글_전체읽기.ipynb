```python
import warnings
warnings.filterwarnings(action='ignore')
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
```

```python
option = Options()
option.add_argument('--window-size=1800,1000')
driver = webdriver.Chrome(options=option)
driver.get('https://comic.naver.com/webtoon/detail?titleId=812354&no=237&week=thu')
driver.implicitly_wait(5)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
# 전체 댓글 버튼을 찾아 클릭한다.
driver.find_element(By.XPATH, '//*[@id="wcc_root"]/section/div[4]/button[2]').click()
time.sleep(1)

# 더 보기 버튼을 찾아 더 보기 버튼이 없을때까지 클릭한다.
while True:
    try:
        driver.find_element(By.XPATH, '//*[@id="wcc_root"]/section/button').click()
        time.sleep(1)
    except:
        break
        
# 전체 댓글을 읽는다.
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
reviews = soup.select('p._text_8n8z9_16')
print(len(reviews))

for i in range(len(reviews)):
    print('{:3d}번째 댓글: {}'.format(len(reviews) - i, ' '.join(reviews[i].text.split('\n'))))
```

```python
file = open('./output/reviews.txt', 'w', -1, 'utf-8')
for review in reviews:
    file.write(' '.join(review.text.split('\n')) + '\n')
file.close()
print('reviews.txt 파일로 저장완료')
```
