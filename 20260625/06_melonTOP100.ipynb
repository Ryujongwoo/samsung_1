```python
import warnings
warnings.filterwarnings(action='ignore')
import requests
from bs4 import BeautifulSoup
from datetime import datetime as dt
```

```python
targetSite = 'https://www.melon.com/chart/index.htm'
request = requests.get(targetSite)
print(request)
# <Response [406]>: 허용되지 않음, 요청한 페이지가 요청한 콘텐츠의 특성으로 응답할 수 없다.
```

Trident User Agent

https://explore.whatismybrowser.com/useragents/explore/layout_engine_name/trident/

```python
targetSite = 'https://www.melon.com/chart/index.htm'
headers = {'User-agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64; Trident/7.0; MSAppHost/2.0; rv:11.0) like Gecko'}
# <Response [406]>이 발생되면 User-agent' 정보를 get() 메소드의 headers 속성에 추가한다.
request = requests.get(targetSite, headers=headers)
# print(request)
html = request.text
soup = BeautifulSoup(html, 'html.parser')
# print(soup)
```

노래 제목 크롤링

```python
titles = soup.findAll('div', {'class': 'ellipsis rank01'})
# print(len(titles))
# print(titles)

for i in range(len(titles)):
    # print('{:3d}위 {}'.format(i + 1, titles[i].text.strip()))
    # print('{:3d}위 {}'.format(i + 1, titles[i].text.split('\n')[1]))
    print('{:3d}위 {}'.format(i + 1, titles[i].text.strip().split('\n')[0]))
```

아티스트 크롤링

```python
artists = soup.findAll('span', {'class': 'checkEllipsis'})
# print(len(artists))
# print(artists)

for i in range(len(artists)):
    print('{:3d}위 {}'.format(i + 1, artists[i].text.strip().split(',')[0]))
    # print('{:3d}위 {}'.format(i + 1, artists[i].text.split('\n')[0])) # 적절 X
    # print('{:3d}위 {}'.format(i + 1, artists[i].text.strip().split('\n')[0])) # 적절 X
```

```python
print(dt.now())
for i in range(len(titles)):
    artist = artists[i].text.strip().split('\n')[0]
    title = titles[i].text.strip().split('\n')[0]
    print('{:3d}위 {} - {}'.format(i + 1, artist, title))
```

```python
with open('./output/melonTOP100.txt', 'w', -1, 'utf-8') as file:
    file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
    for i in range(len(titles)):
        artist = artists[i].text.strip().split('\n')[0]
        title = titles[i].text.strip().split('\n')[0]
        file.write('{:3d}위 {} - {}\n'.format(i + 1, artist, title))
print('melonTOP100.txt 파일로 쓰기 완료')
```
