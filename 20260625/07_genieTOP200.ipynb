4개 페이지를 크롤링한 결과를 하나의 리스트에 모아준다.

```python
genie = []
headers = {'User-agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64; Trident/7.0; MSAppHost/2.0; rv:11.0) like Gecko'}
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20260625&hh=15&rtm=Y&pg='

for i in range(1, 5):
    targetSite = f'{url}{i}'
    request = requests.get(targetSite, headers=headers)
    time.sleep(1)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    
    for j in range(len(titles)):
        genie.append(
            (artists[j + 5].text.strip().split('\n')[0], titles[j].text.strip().split('\n')[0])
        )
    # ===== for
# ===== for
print(len(genie))
```

```python
print(dt.now())
for i in range(len(genie)):
    print('{:3d}위 {} - {}'.format(i + 1, genie[i][0], genie[i][1]))
```

```python
with open('./output/genieTOP200.txt', 'w', -1, 'utf-8') as file:
    file.write(dt.now().strftime('%Y-%m-%d %H:%M:%S') + '\n')
    for i in range(len(genie)):
        file.write('{:3d}위 {} - {}\n'.format(i + 1, genie[i][0], genie[i][1]))
print('genieTOP200.txt 파일로 쓰기 완료')
```
