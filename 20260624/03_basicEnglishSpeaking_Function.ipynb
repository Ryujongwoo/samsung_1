```python
import requests
from bs4 import BeautifulSoup
```

```python
def getSubject():
    targetSite = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'
    request = requests.get(targetSite)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    
    subjects = [] # лМАнЩФ м£Љм†Ь
    contentLinks = [] # лМАнЩФ м£Љм†ЬмЧР лФ∞л•Є мДЄлґА лМАнЩФ лВімЪ©мЭШ urlмЭД м†АмЮ•нХ† лєИ л¶ђмК§нКЄ
    
    divs = soup.findAll('div', {'class': 'tcb-flex-col'})
    for div in divs:
        a = div.findAll('a')
        for text in a:
            subjects.append(text.text) # лМАнЩФ м£Љм†Ь мґФк∞А
            contentLinks.append(text.get('href')) # мДЄлґА лМАнЩФ лВімЪ©мЭШ url мґФк∞А
        # ===== for
    # ===== for
    
    # лМАнЩФ м£Љм†ЬмЩА лМАнЩФ м£Љм†ЬмЧР лФ∞л•Є мДЄлґА лМАнЩФ лВімЪ©мЭШ urlмЭі м†АмЮ•лРЬ л¶ђмК§нКЄл•Љ л¶ђнДімЛЬнВ®лЛ§.
    # л¶ђнДілРШлКФ лН∞мЭінД∞к∞А 2к∞Ьк∞А мХДлЛИлЭЉ subjectsмЩА contentLinksл•Љ нКЬнФМл°Ь лђґмЦімДЬ л¶ђнДімЛЬнВ®лЛ§.
    return subjects, contentLinks
```

```python
subjects, contentLinks = getSubject()
```

```python
for i in range(len(subjects)):
    print('{:2d}. {}: {}'.format(i + 1, subjects[i], contentLinks[i]))
```
