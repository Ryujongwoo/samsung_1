BinaryTree 클래스를 만든다.

```python
class BinaryTree:
    def __init__(self, data=None):
        self.left = None # 왼쪽 자식 노드의 주소를 기억한다.
        self.data = data # 데이터
        self.right = None # 오른쪽 자식 노드의 주소를 기억한다.
        
    # 이진 트리에 데이터를 추가하는 메소드
    def append(self, data):
        print(f'이진 트리에 추가하려는 데이터 {data}의 부모 노드는 {self.data}입니다.')
        
        # 이진 트리에 추가할 데이터와 부모 노드의 데이터를 비교해서 이진 트리에 데이터가 추가되는 위치가 결정된다.
        # 이진 트리에 추가할 데이터가 부모 노드의 데이터보다 작으면 부모 노드의 왼쪽으로 추가한다.
        # 이진 트리에 추가할 데이터가 부모 노드의 데이터보다 크면 부모 노드의 오른쪽으로 추가한다.
        
        # 이진 트리에 추가하려는 데이터가 이진 트리에 존재하는가 확인한다.
        if data == self.data:
            print(f'{data}는(은) 이진 트리에 존재하는 데이터입니다.')
        # 이진 트리에 추가하려는 데이터가 부모 노드의 데이터보다 작은가 확인한다.
        elif data < self.data:
            # 부모 노드가 크기때문에 부모 노드의 왼쪽으로 데이터를 추가한다.
            print('부모 노드가 크기때문에 부모 노드의 왼쪽으로 데이터를 추가한다.')
            # 부모 노드의 왼쪽 링크(self.left)가 비어(None)있어야 데이터를 추가할 수 있다.
            if self.left is None:
                # 부모 노드의 왼쪽 링크가 비어있으므로 데이터를 추가한다.
                print(f'부모({self.data}) 왼쪽에 데이터 {data} 추가 가능')
                # 새 노드를 추가해야하므로 이진 트리에 추가할 데이터를 BinaryTree 클래스 객체를 생성해서 부모 노드의 비어있는 왼쪽 링크에 BinaryTree 클래스
                # 객체의 주소를 넣어준다.
                self.left = BinaryTree(data)
                print(f'부모({self.data}) 왼쪽에 데이터 {data} 추가 완료')
            else:
                # 부모 노드의 왼쪽 링크가 비어있지않으므로 데이터를 추가할 수 없다.
                print(f'부모({self.data}) 왼쪽 링크가 비어있지않으므로 데이터 {data} 추가 불가능')
                # 부모 노드의 왼쪽 링크가 비어있지 않기 때문에 부모 노드의 왼쪽 링크가 참조하는 노드에서 append() 메소드를 실행한다.
                print(f'부모({self.data}) 왼쪽에 {data}를(을) 추가할 수 없으므로 부모 왼쪽 노드({self.left.data})에서 append() 메소드를 실행한다.')
                self.left.append(data)
            # ===== if
        else:
            # 부모 노드가 작기때문에 부모 노드의 오른쪽으로 데이터를 추가한다.
            print('부모 노드가 작기때문에 부모 노드의 오른쪽으로 데이터를 추가한다.')
            # 부모 노드의 오른쪽 링크(self.right)가 비어(None)있어야 데이터를 추가할 수 있다.
            if self.right is None:
                # 부모 노드의 오른쪽 링크가 비어있으므로 데이터를 추가한다.
                print(f'부모({self.data}) 오른쪽에 데이터 {data} 추가 가능')
                # 새 노드를 추가해야하므로 이진 트리에 추가할 데이터를 BinaryTree 클래스 객체를 생성해서 부모 노드의 비어있는 오른쪽 링크에 BinaryTree 클래스
                # 객체의 주소를 넣어준다.
                self.right = BinaryTree(data)
                print(f'부모({self.data}) 오른쪽에 데이터 {data} 추가 완료')
            else:
                # 부모 노드의 오른쪽 링크가 비어있지않으므로 데이터를 추가할 수 없다.
                print(f'부모({self.data}) 오른쪽 링크가 비어있지않으므로 데이터 {data} 추가 불가능')
                # 부모 노드의 오른쪽 링크가 비어있지 않기 때문에 부모 노드의 오른쪽 링크가 참조하는 노드에서 append() 메소드를 실행한다.
                print(f'부모({self.data}) 오른쪽에 {data}를(을) 추가할 수 없으므로 부모 오른쪽 노드({self.right.data})에서 append() 메소드를 실행한다.')
                self.right.append(data)
            # ===== if
        # ===== if
    # ===== append()
    
    # 이진 트리의 운행(중복되는 노드없이 모든 노드를 방문하는 것)법은 부모 노드를 언제 방분하느냐에 따라서 3가지가 있다.
    # inorder: 왼쪽 자식 => 부모 => 오른쪽 자식
    # preorder: 부모 => 왼쪽 자식 => 오른쪽 자식
    # postorder: 왼쪽 자식 => 오른쪽 자식 => 부모
    
    # 이진 트리를 inorder 방식으로 운행하는 메소드
    # 방문 순서: 3 => 6 => 10 => 12 => 15 => 20 => 25
    def inorder(self):
        print(f'{self.data}의 inorder() 메소드 시작')
        # 왼쪽 자식 노드가 있나 검사해서 있으면 왼쪽 자식 노드에서 inorder() 메소드를 실행한다.
        if self.left is not None:
            self.left.inorder()
        # 방문한 노드의 데이터를 출력한다.
        print(f'방문한 노드의 데이터: {self.data}')
        # 오른쪽 자식 노드가 있나 검사해서 있으면 오른쪽 자식 노드에서 inorder() 메소드를 실행한다.
        if self.right is not None:
            self.right.inorder()
        print(f'{self.data}의 inorder() 메소드 종료')
    # ===== inorder()
    
    # 이진 트리를 preorder 방식으로 운행하는 메소드
    # 방문 순서: 12 => 6 => 3 => 10 => 20 => 15 => 25
    def preorder(self):
        print(f'{self.data}의 preorder() 메소드 시작')
        # 방문한 노드의 데이터를 출력한다.
        print(f'방문한 노드의 데이터: {self.data}')
        # 왼쪽 자식 노드가 있나 검사해서 있으면 왼쪽 자식 노드에서 preorder() 메소드를 실행한다.
        if self.left is not None:
            self.left.preorder()
        # 오른쪽 자식 노드가 있나 검사해서 있으면 오른쪽 자식 노드에서 preorder() 메소드를 실행한다.
        if self.right is not None:
            self.right.preorder()
        print(f'{self.data}의 preorder() 메소드 종료')
    # ===== preorder()
    
    # # 이진 트리를 postorder 방식으로 운행하는 메소드
    # 방문 순서: 3 => 10 => 6 => 15 => 25 => 20 => 12
    def postorder(self):
        print(f'{self.data}의 postorder() 메소드 시작')
        # 왼쪽 자식 노드가 있나 검사해서 있으면 왼쪽 자식 노드에서 postorder() 메소드를 실행한다.
        if self.left is not None:
            self.left.postorder()
        # 오른쪽 자식 노드가 있나 검사해서 있으면 오른쪽 자식 노드에서 postorder() 메소드를 실행한다.
        if self.right is not None:
            self.right.postorder()
        # 방문한 노드의 데이터를 출력한다.
        print(f'방문한 노드의 데이터: {self.data}')
        print(f'{self.data}의 postorder() 메소드 종료')
    # ===== postorder
```

```python
# 이진 트리의 최상위(root) 노드를 만든다.
root = BinaryTree(12)
# 이진 트리에 데이터를 추가한다.
root.append(12)
print('=' * 100)
root.append(6)
print('=' * 100)
root.append(3)
print('=' * 100)
root.append(20)
print('=' * 100)
root.append(25)
print('=' * 100)
root.append(10)
print('=' * 100)
root.append(15)
print('=' * 100)
```
