pm => 개인별 지급 화폐 매수  
tm => 전체 지급 화폐 매수  
n => 이름  
p => 출장비  
m => 화폐 권종(50000, 10000, 5000, 1000, 500, 100, 50, 10, 5, 1)  
sw => 화폐 권종 변환에 사용하는 스위치 변수  
t => 입력받은 출장비(p)를 출력하려면 입력받은 출장비를 유지하기 위해서 화폐 권종 계산에 사용하는 변수

```python
pm = [0] * 10 # 개인별 지급 화폐 매수를 기억할 리스트를 만든다.
tm = [0 for _ in range(10)] # 전체 지급 화폐 매수를 기억할 리스트를 만든다.
# print(pm, tm)

# 이름이 'quit'가 입력될 때까지 반복하며 화폐 권종별 매수를 계산한다.
while True:
    n, p = input('이름과 출장비를 입력하세요: ').split()
    if n.upper() == 'QUIT':
        break
    # ===== if
    
    t = int(p) # 나중에 입력받은 출장비를 출력하기 위해 출장비의 사본을 만든다.
    m = 50000 # 화폐 권종별 매수를 계산에 사용할 변수를 초기화시킨다.
    sw = 1 # 화폐 권종 변경에 사용하는 스위치 변수
    
    # 화폐 권종 개수만큼 반복하며 개인별 출장비 화폐 권종별 매수를 계산한다.
    for k in range(10):
        pm[k] = t // m # 개인별 화폐 권종별 매수를 계산한다.
        t %= m # 다음 화폐 권종 계산하기 위해 권종별 화폐 매수를 계산하고 남은 나머지 금액을 계산한다.
        # 다음 화폐 권종을 계산한다.
        if sw == 1:
            m //= 5
            sw = 0
        else:
            m //= 2
            sw = 1
        # ===== if
    # ===== for
    
    # 이름과 출장비 및 개인별 지급 화폐 권종의 매수를 출력한다.
    print(n, p, end=' ')
    for k in range(10):
        print(pm[k], end=' ')
        tm[k] += pm[k] # 전체 권종별 화폐 매수를 계산한다.
    # ===== for
    print()
# ===== while
# 전체 권종별 화폐 매수를 출력한다.
print(tm)
```

이름과 출장비를 넘겨받아 초기화하고 개인별 지급 화폐 권종의 매수를 계산해서 기억하는 클래스

```python
class Money:
    def __init__(self, name, money):
        self.name = name
        self.money = int(money)
        # 개인별 지급 화폐 권종의 매수를 계산한다.
        self.pm = [0 for _ in range(10)]
        t = self.money
        m = 50000
        sw = 1
        for k in range(10):
            self.pm[k] = t // m
            t %= m
            if sw == 1:
                m //= 5
                sw = 0
            else:
                m //= 2
                sw = 1
            # ===== if
        # ===== for
        
    def __str__(self):
        return f'{self.name} {self.money} {self.pm}'
```

```python
money = Money('강감찬', '537620')
print(money)
print(money.name)
print(money.money)
print(money.pm)
```

```python
m = []
while True:
    n, p = input('이름과 출장비를 입력하세요: ').split()
    if n.upper() == 'QUIT':
        break
    # ===== if
    money = Money(n, p)
    m.append(money)
# ===== while
print(m)
```

```python
for i in m:
    print(i)
```

```python
tm = [0 for _ in range(10)]
print('{:^103s}'.format('출장비 지급 일람표'))
print('=' * 103)
print('{:^10s}    출장비  오만원    만원  오천원    천원  오백원    백원  오십원    십원    오원    일원'.format('이름'))
print('=' * 103)

for i in m:
    print('{:^10s} {:8,d}'.format(i.name, i.money), end='')
    for j in range(10):
        print('{:8d}'.format(i.pm[j]), end='')
        tm[j] += i.pm[j]
    # ===== for j
    print()
# ===== for i
print('=' * 103)
print('{:^16s}'.format('전체화폐매수'), end='')
for j in range(10):
    print('{:8d}'.format(tm[j]), end='')
# ===== for j
print('\n' + '=' * 103)
```
