BeautifulSoup: requests 라이브러리를 사용해서 얻어온 웹 문서 파싱(분석)한다.

```python
# !pip install beautifulsoup4
from bs4 import BeautifulSoup
```

```python
targetSite = 'http://www.dowellcomputer.com/main.jsp'
```

requests 모듈의 get() 메소드로 타겟 사이트의 정보를 읽어온다.

```python
request = requests.get(targetSite)
print(request)
# <Response [200]>: 서버가 요청을 제대로 처리했다는 뜻으로 서버가 요청한 페이지를 제공했다는 의미이다.
```

타겟 사이트에서 얻어온 정보 중에서 text된 html 문서만 얻어온다.

```python
html = request.text
print(type(html))
print(html)
```

bs4 모듈의 BeautifulSoup 클래스로 타겟 사이트에서 얻어온 html 문서(문자열)을 html parser를 사용해서 파싱(분석)할 수 있도록 객체를 만든다.

```python
soup = BeautifulSoup(html, 'html.parser')
print(type(soup))
print(soup)
```

html 태그와 html 태그에 지정한 속성을 이용한 크롤링

findAll('태그이름'[, {'속성이름': '속성값', ...}]) => 속성없이 태그 단위로 크롤링할 때는 {} 부분을 생략한다.

```python
images = soup.findAll('img')
print(type(images))
print(len(images))
print(images)
print(images[0])
print(images[1])
print(images[2])
for image in images:
    print(image)
```

```python
# 최근 강의 제목은 <b> 태그 내부에 있다.
titles = soup.findAll('b')
print(titles)
print('=' * 100)
for title in titles:
    print(title.text)
```

```python
# <td> 태그 중에서 class 속성의 속성값이 'middle'인 요소만 얻어온다.
titles = soup.findAll('td', {'class': 'middle'})
print(titles)
print('=' * 100)
for title in titles:
    print(title.text.strip())
```

CSS 선택자를 이용한 크롤링

select('CSS 선택자')

```python
# 최근 강의 제목은 <td> 태그의 자식 요소인 <a> 태그 내부에 있다.
# >: 자식 선택자, 공백: 자손 선택자
titles = soup.select('td > a')
print(titles)
print('=' * 100)
for title in titles:
    print(title.text.strip())
```

```python
titles = soup.select('td a')
print(titles)
print('=' * 100)
for title in titles:
    # get('속성이름'): 인수로 지정한 속성의 속성값을 얻어온다.
    # print(title.get('href'))
    # if title.get('href').find('study') >= 0:
        # print(title.text)
    if title.text.find('활용') >= 0:
        print(title.text)
```
