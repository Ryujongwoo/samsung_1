class

class는 데이터와 데이터를 처리하는 함수(클래스 내부에서 만든 함수를 메소드라 부른다.)를 가지고있는 자료형이다.

```python
'''
형식

class 클래스이름: # 클래스 이름은 다른 변수와 구분하기 위해 일반적으로 첫 문자를 대문자로 작성한다.

    # 생성자(초기자) 함수를 만든다. => __init__(self):
    # 생성자(constructor)란 객체(클래스로 만드는 변수)개 생성될 때 자동으로 실행되는 함수이다.
    # 클래스 내부에서 선언하는 모든 함수의 첫 번째 인수는 무조건 self를 사용한다.
    # self라는 변수로 클래스의 객체가 메모리에 생성된 주소가 넘어온다. => self로 넘어온 주소로 객체를 식별한다.
    # 생성자는 class 내부의 모든 함수에서 사용하는 변수(멤버 변수)를 초기화시키는 목적으로 사용한다.
    def __init__(self[, 인수, ...]):
        # 멤버 변수 초기화
        # 변수 이름 앞에 'self.'이 붙어있는 변수를 멤버 변수라 부르고 class 전체에서 사용할 수 있다.
        # 변수 이름 앞에 'self.'이 붙어있지않은 변수는 지역 변수라 부르고 변수가 선언된 함수 내부에서만 사용할 수 있다.
        
    # 클래스에 저장된 데이터를 출력하려면 __str__(self) 함수는 java의 toString() 메소드와 같은 기능을 실행한다.
    def __str__(self):
        # return 뒤에 클래스 객체를 출력할 때 출려하려는 내용을 반드시 문자열로 return 시키다.
        return '출력할 내용'
'''
pass
```

```python
class Score:
    # Score class의 객체가 생성될 때 이름과 3과목의 점수를 넘겨받아서 멤버 변수를 초기화시키고 총점과 평균을 계산한다.
    # 생성자의 인수로 이름과 3과목의 점수가 넘어오지 않을 경우 이름은 '무명씨', 점수는 0으로 초기화시킨다.
    def __init__(self, name='무명씨', python=0, numpy=0, pandas=0):
        # print('Score class의 객체가 메모리에 생성될 때 자동으로 실행된다.')
        # print('Score class의 객체가 메모리에 생성된 주소: {}'.format(self))
        # 멤버 변수 초기화
        self.name = name
        self.python = python
        self.numpy = numpy
        self.pandas = pandas
        # 총점과 평균을 계산한다.
        self.total = python + numpy + pandas
        self.mean = self.total / 3
        # print('{}님의 총점은 {}점이고 평균은 {:.2f}점 입니다.'.format(name, self.total, self.mean))
    # ===== __init__()
    
    # 이름과 총점을 출력하는 함수
    def getTotal(self):
        print('{}님의 총점은 {}점 입니다.'.format(self.name, self.total))
    
    # 이름과 평균을 출력하는 함수
    def getMean(self):
        print('{}님의 평균은 {:.2f}점 입니다.'.format(self.name, self.mean))
    
    def __str__(self):
        # return 'name: {}, python: {}, numpy: {}, pandas: {}, total: {}, mean: {:.2f}'.format(self.name, self.pandas, self.numpy, self.pandas, self.total, self.mean)
    
        # 파이썬은 1줄에 1문장 코딩이 원칙이지만 1줄에 2개 이상의 문장을 코딩하려면 문장과 문장을 ';'으로 구분한다.
        # 문자열 1문장이 너무 길어서 2줄 이상으로 나눠 코딩하려면 줄을 변경하는 부분에 '\'를 입력하고 다음줄에 코딩하면 같은 문장으로 인식한다.
        # '\'뒤에 공백이 있으면 에러가 발생된다.
        '''
        return 'name: {}, python: {}, numpy: {}, pandas: {}, total: {}, ' \ 
                'mean: {:.2f}'.format(self.name, self.pandas, self.numpy, self.pandas, self.total, self.mean)
        '''
        
        # () 내부에서는 줄을 몇 번을 바꾸더라고 1문장으로 인식된다.
        return 'name: {}, python: {}, numpy: {}, pandas: {}, total: {}, mean: {:.2f}'.format(
            self.name, 
            self.pandas, 
            self.numpy, 
            self.pandas, 
            self.total, 
            self.mean
        )
```

class 객체(변수) 선언 방법

객체(변수)이름 = 클래스이름([생성자로 전달할 데이터, ...])  
class 객체를 선언하면 class 객체가 메모리에 할당된 주소가 생성자의 첫 번째 인수인 self로 전달된다.

```python
score1 = Score('홍길동', 100, 100, 99)
print(score1)
```

class 객체의 멤버 변수(앞에 'self.'을 붙인 변수)에 접근하는 방법

객체이름.멤버변수이름

```python
print('{}님의 총점은 {}점이고 평균은 {:.2f}점 입니다.'.format(score1.name, score1.total, score1.mean))
```

class 객체의 멤버 함수(메소드) 실행 방법

객체이름.메소드이름()

```python
score1.getTotal()
score1.getMean()
```

```python
score2 = Score()
print(score2)
print('{}님의 총점은 {}점이고 평균은 {:.2f}점 입니다.'.format(score2.name, score2.total, score2.mean))
score2.getTotal()
score2.getMean()
```
