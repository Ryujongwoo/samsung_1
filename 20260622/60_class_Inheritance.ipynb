부모(상위, 슈퍼, 기반) 클래스

```python
class Parent:
    def __init__(self, name='무명씨', gender=False):
        print('부모 클래스의 생성자 실행')
        self.name = name
        self.gender = gender
        
    def __str__(self):
        return '{}({})'.format(self.name, '남' if self.gender else '여')
```

```python
# 부모로 사용할 클래스가 잘 생성되었나 확인한다.
parent1 = Parent()
print(parent1)
parent2 = Parent('이몽룡', True)
print(parent2)
parent3 = Parent('성춘향', False)
print(parent3)
```

클래스 상속

상속이란 부모 클래스에서 정의한 모든 멤버 변수와 함수를 자식 클래스에서 정의하지 않고 정의한 것처럼 사용할 수 있게 해주는 기능을 말한다.

상속의 형식

class 자식클래스이름(부모클래스이름)

자식(하위, 서브, 파생) 클래스  
부모 클래스(Parent)를 상속받아 자식(Child) 클래스를 만든다.

```python
class Child(Parent):
    # Parent 클래스로부터 __init__(self, name='무명씨', gender=False), __str__(self), self.name, self.gender를 상속받았으므로 별도의 선언없이 사용할 수 있다.
    # 부모 클래스에서 상속받은 함수의 기능만으로는 자식 클래스에서 사용하기에 불충분하기 때문에 자식 클래스에서 재정의해서 사용한다.
    # 함수 재정의, override
    def __init__(self, name='무명씨', gender=False, age=0, nickname='없음'):
        # 부모 클래스에서 상속받은 멤버 변수는 부모 클래스의 생성자를 호출해서 초기화시킨다.
        # super()는 부모 클래스를 의미한다.
        super().__init__(name, gender) # 부모 클래스의 생성자를 호출한다.
        print('자식 클래스의 생성자 실행')
        # self.name = name
        # self.gender = gender
        self.age = age
        self.nickname = nickname
        
    def __str__(self):
        return '{}({}) - {}, {}'.format(self.name, '남' if self.gender else '여', self.age, self.nickname)
```

```python
child = Child('홍길동', True, 20, '도적넘')
print(child)
print(child.name)
print(child.gender)
print(child.age)
print(child.nickname)
```
