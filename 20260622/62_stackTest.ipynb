소스프로그램이 위치한 폴더 바로 아래 폴더(lib)에 위치한 stackModule 모듈에서 Stack 클래스만 import 한다.

```python
from lib.stackModule import Stack
```

```python
stack = Stack()
stack.view()
```

```python
menu = 0
while menu != 4:
    while True:
        print('=' * 30)
        print(' 1.push  2.pop  3.view  4.종료 ')
        print('=' * 30)
        menu = int(input('원하는 메뉴를 입력하고 enter를 누르세요: '))
        if menu in [1, 2, 3, 4]:
            break
        else:
            print('메뉴는 1 ~ 4 사이의 값만 입력해야 합니다.')
        # ===== if
    # ===== while
    
    if menu == 1:
        data = input('스택에 저장할 데이터를 입력하세요: ')
        stack.push(data)
    elif menu == 2:
        stack.pop()
    elif menu == 3:
        stack.view()
    # ===== if
# ===== while
print('프로그램을 종료합니다. 바이바이~~~~~')
```
