```python
import random
# 1 ~ 9사이의 숫자가 채워진 리스트를 만든다.
baseball = [i for i in range(1, 10)]
print(baseball)
from datetime import datetime as dt
```

```python
# 레벨을 입력받는다.
level = int(input('레벨을 입력하세요: '))
# 입력받은 레벨만큼 baseball 리스트에 저장된 숫자를 섞어서 뽑아낸다.
# random.shuffle(baseball)
# print(baseball)
# cpu = baseball[:level]
cpu = random.sample(baseball, level)
# print(cpu)

# 몇 번만에 맞출지 모르기때문에 무한루프를 돌려서 처리한다.
while True:
    # 새 반복이 시작될때마다 스트라이크 개수와 볼의 개수를 세는 변수를 0으로 초기화시킨다.
    s = 0 # 스트라이크 개수
    b = 0 # 볼 개수
    # 컴퓨터가 랜덤하게 발생시킨 숫자를 맞추기위해서 유저에게 입력을 요구한다.
    user = input('맞춰봐: ')
    
    # 레벨만큼 반복하며 스트라이크 개수와 볼의 개수를 센다.
    for i in range(level):
        # cpu[i]와 user[i]에 저장된 값의 데이터 타입이 다르기때문에 str() 또는 int() 함수로 데이터 타입을 맞춰서 비교한다.
        # 스트라이크 => cpu[i]와 user[i]에 저장된 값이 같을 경우
        if cpu[i] == int(user[i]): # str(cpu[i]) == user[i]와 같이 비교해도 된다.
            s += 1
        # 볼 => user[i] 값이 cpu에 포함되어 있을 경우
        elif int(user[i]) in cpu:
            b += 1
    # ===== for
        
    # 스트라이크 개수와 레벨이 같으면 맞춘것이므로 무한루프를 탈출한다.
    if s == level:
        break
    # 맞추지못했으므로 힌트를 출력한다.
    print(f'{s} 스트라이크, {b} 볼')
# ===== while
print('맞췄다.')
```

```python
level = int(input('레벨을 입력하세요: '))
cpu = random.sample(baseball, level)
print(cpu)

# 몇 번만에 맞췄나 기억할 변수를 선언하고 0으로 초기화시킨다.
count = 0
# baseball 게임 시작 시간을 저장한다.
start = dt.now()
# print(start)

while True:
    s = 0
    b = 0
    user = input('맞춰봐: ')
    # 입력받을때마다 몇 번만에 맞췄나 기억하는 변수를 1증가시킨다.
    count += 1
    
    for i in range(level):
        if cpu[i] == int(user[i]):
            s += 1
        elif int(user[i]) in cpu:
            b += 1
    # ===== for
        
    if s == level:
        break
    print(f'{s} 스트라이크, {b} 볼')
# ===== while

# baseball 게임 시작 종료을 저장한다.
end = dt.now()
# print(end)
print(f'{count}번 시도해서 {(end - start).total_seconds()}초 만에 맞췄다.')
```
