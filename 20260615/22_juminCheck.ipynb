주민등록번호 유효성 검사

주민등록번호의 각 자리 숫자와 가중치(234567892345)를 곱한 합계를 계산한다.  
위의 결과를 11로 나눈 나머지를 계산해서 11에서 뺀다.  
11에서 나머지를 뺀 결과가 2자리 숫자가 나오면 10의 자리는 버리고 1의 자리만 취한다.  
1의 자리가 주민등록번호의 마지막 자리와 같으면 정상적인 주민등록번호이고 다르면 오류가있는 주민등록번호이다.

<img src="juminCheck.png" width="1100" align="left" />

```python
jumin = input('주민등록번호 13자리를 "-"없이 입력하세요: ')

# 주민등록번호의 각 자리 숫자와 가중치(234567892345)를 곱한 합계를 계산한다.
check = '234567892345'
total = 0
for i in range(12):
    total += int(jumin[i]) * int(check[i])
# ===== for
print(total)

# 주민등록번호와 가중치를 곱한 결과의 합계를 11로 나눈 나머지를 11에서 뺀다.
# 11에서 나머지를 뺀 결과가 2자리면 10의 자리는 버리고 1의 자리만 취한다.
result = (11 - total % 11) % 10
print(result)

# 위의 연산 결과와 주민등록번호의 마지막 자리를 비교한다.
if str(result) == jumin[-1]:
    print('정상')
else:
    print('누구냐?? 넌...')
print('정상' if str(result) == jumin[-1] else '누구냐?? 넌...')
```

인덱스가 8보다 작은 경우 가중치는 무조건 인덱스보다 2가 크고 인덱스가 8이상인 경우 가중치는 인덱스보다 무조건 6이 작다.

<img src="juminCheck2.png" width="900" align="left" />

```python
jumin = input('주민등록번호 13자리를 "-"없이 입력하세요: ')
# check = '234567892345'
total = 0
for i in range(12):
    # total += int(jumin[i]) * int(check[i])
    total += int(jumin[i]) * (i + 2 if i < 8 else i - 6)
# ===== for
result = (11 - total % 11) % 10
print('정상' if str(result) == jumin[-1] else '누구냐?? 넌...')
```

가중치는 무조건 인덱스를 8로 나눈 나머지 보다 2가 크다.

<img src="juminCheck3.png" width="900" align="left" />

```python
jumin = input('주민등록번호 13자리를 "-"없이 입력하세요: ')
total = 0
for i in range(12):
    # total += int(jumin[i]) * (i + 2 if i < 8 else i - 6)
    total += int(jumin[i]) * (i % 8 + 2)
# ===== for
result = (11 - total % 11) % 10
print('정상' if str(result) == jumin[-1] else '누구냐?? 넌...')
```
