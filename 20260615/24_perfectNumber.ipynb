완전수

완전수란 자기 자신을 제외한 약수의 합이 자기 자신과 같은 숫자를 말한다.  
6의 약수는 1, 2, 3, 6이고 여기서 자기 자신인 6을 제외한 약수 1, 2, 3의 합계가 자기 자신인 6이 되므로 6은 완전수이다.

4 ~ 10000 사이의 완전수를 계산해서 출력하고 완전수의 개수를 출력하는 프로그램

```python
lm = 0 # 4 ~ 10000 사이의 완전수의 개수를 세는 변수
# 바깥 반복은 4 ~ 10000 사이의 완전수를 판별해야하므로 완전수인가 판별할 수(n)를 4부터 10000까지 1씩 증가시킨다.
for n in range(4, 10001):
    # 완전수인가 판별할 숫자가 바뀔때마다 약수의 합을 다시 계산해야하므로 n이 바뀔때마다 초기화시킨다.
    # n이 바뀔때마다 초기화 시키지않으면 이전 숫자의 약수 합계에 현재 숫자의 약수가 누적된다.
    total = 0 # 완전수인가 판별할 수의 자신을 제외한 약수의 합계를 기억하는 변수
    # 어떤 숫자를 나눠서 떨어뜨릴 수 있는 가장 큰 숫자는 자신을 제외하면 자신의 절반을 넘어가지 않는다.
    # 완전수는 자신을 제외한 약수의 합이 자신과 같으면 되므로 자신으로 나눌 필요가 없다.
    k = n // 2
    # 안쪽 반복은 n이 완전수인가 판별하기위해서 자신을 제외한 약수의 합계를 계산한다.
    for j in range(1, k + 1):
        # 완전수인가 판별할 수의 약수를 계산하기위해서 1 ~ k 사이의 숫자로 나눈 나머지를 계산한다.
        r = n % j
        # n을 j로 나눈 나머지가 0이면 j는 n의 약수이므로 약수의 합계를 계산한다.
        if r == 0: # j가 n의 약수인가?
            total += j # j가 n의 약수이므로 약수의 합계를 계산한다.
        # ===== if
    # ===== for j
    # n이 완전수이가 판단해서 완전수면 n을 출력하고 완전수의 개수를 1증가시킨다.
    if n == total: # 완전수인가?
        print('{:4d}'.format(n))
        lm += 1 # 완전수의 개수를 증가시킨다.
    # ===== if
# ===== for n
print(f'완전수의 개수는 {lm}개 입니다.')
```
