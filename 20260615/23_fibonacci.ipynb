```python
a = 1 # 피보나치 수열의 첫 번째 항 초기화, n-2 번째 항의 값
b = 1 # 피보나치 수열의 두 번째 항 초기화, n-1 번째 항의 값
y = 2 # 피보나치 수열의 첫 번째 항과 두 번째 항의 합계, n번째 항까지 피보나치 수열의 합계를 기억하는 변수

# 피보나치 수열의 합계를 계산할 항(n)을 입력받는다.
n = int(input('피보나치 수열의 합계를 계산할 항을 입력하세요: '))

# 첫 번째 항과 두 번째 항은 이미 합계가 계산되었다는 가정하에 작업을 시작하므로 세 번째 항부터 n번째 항까지 반복하며 합계를 계산한다.
for k in range(3, n + 1):
    c = a + b # k번째 항의 피보나치 수를 계산한다.
    y += c # k번째 항까지 피보나치 수열의 합계를 계산한다.
    # a = b # 다음 항의 값을 계산하기 위해서 현재 반복의 n-1 번째 항의 값(b)을 다음 반복의 n-2 번째 항의 값(a)에 넣어준다.
    # b = c # 다음 항의 값을 계산하기 위해서 현재 반복의 n 번째 항의 값(c)을 다음 반복의 n-1 번째 항의 값(b)에 넣어준다.
    a, b = b, c
# ===== for
print(f'피보나치 수열 {n}번째 항 까지의 합계: {y}')
```

```python
# 피보나치 수열의 1번째, 2번째 항의 값으로 초기화된 리스트를 만든다.
f = [1, 1]
n = int(input('피보나치 수열의 합계를 계산할 항을 입력하세요: '))
for k in range(3, n + 1):
    # k=3일때 f[0] + f[1]
    # k=4일때 f[1] + f[2]
    # k=5일때 f[2] + f[3]
    f.append(f[k - 3] + f[k - 2])
# ===== for
print(f'피보나치 수열의 {n}번째 항 까지의 합계: {sum(f)} = {f}')
```

```python
# 피보나치 수열의 8번째 항까지의 합계: 54 = 1 + 1 + 2 + 3 + 5 + 8 + 13 + 21
print(f'피보나치 수열의 {n}번째 항 까지의 합계: {sum(f)} = ', end='')
for i in range(len(f) - 1):
    print(f[i], end=' + ')
print(f[-1])
```

```python
print(f'피보나치 수열의 {n}번째 항 까지의 합계: {sum(f)} = ', end='')
for i in range(len(f)):
    if i > 0:
        print(' + ', end='')
    print(f[i], end='')
```

```python
p = [str(i) for i in f]
print(p)
print(f'피보나치 수열의 {n}번째 항 까지의 합계: {sum(f)} = {" + ".join(p)}', end='')
```

```python
print(f'피보나치 수열의 {n}번째 항 까지의 합계: {sum(f)} = ', end='')
for i in range(len(f)):
    print(f[i], end=' + ' if i < len(f) - 1 else '')
```
