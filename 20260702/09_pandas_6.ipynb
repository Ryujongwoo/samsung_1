자료형 변경하기

pandas의 category 자료형은 R의 factor와 같은 범주형 데이터를 의미한다. => 연산을 위한 데이터가 아니고 분류를 위한 데이터이다.

```python
# total_bill 열의 데이터 타입은 pandas 2.x까지 object로 사용되었다.
print(tips_df.dtypes)
```

astype() 메소드를 사용해서 자료형을 변경할 수 있다.  
total_bill 열의 데이터를 float64 타입에서 str 타입으로 변환한다.

```python
tips_df['total_bill'] = tips_df['total_bill'].astype(str)
tips_df
```

```python
print(tips_df.info())
```

잘못 입력한 문자열 처리하기

```python
tips_df_miss = tips_df.head(10)
tips_df_miss
```

```python
tips_df_miss.loc[[1, 3, 5, 7], 'total_bill'] = 'missing'
# tips_df_miss.iloc[[1, 3, 5, 7], 0] = 'missing'
tips_df_miss
```

```python
# tips_df_miss 데이터프레임의 total_bill 열을 astype() 메소드를 사용해서 실수로 변환하려하면 'missing'이라는 문자열 때문에 에러가 발생된다.
tips_df_miss['total_bill'] = tips_df_miss['total_bill'].astype(float)
```

```python
# to_numeric() 메소드를 사용해도 'missing'이라는 문자열 때문에 에러가 발생된다.
# tips_df_miss['total_bill'] = pd.to_numeric(tips_df_miss['total_bill'])

# to_numeric() 메소드의 errors 속성을 사용하면 어느정도 오류 제어를 할 수 있다.
# errors 속성을 생략하면 'raise'가 기본값으로 사용되며 숫자로 변환할 수 없는 값이 있으면 에러를 발생시킨다.
tips_df_miss['total_bill'] = pd.to_numeric(tips_df_miss['total_bill'], errors='raise')
```

```python
# pandas 3.x 버전부터 errors 속성값으로 'ignore'를 지원하지 않는다. => 에러가 발생되면 명령을 무시한다. => 실행하지 않는다.
# errors 속성값으로 'coerce'를 지정하면 오류가 발생되는 데이터를 누락값으로 변경한다.
tips_df_miss['total_bill'] = pd.to_numeric(tips_df_miss['total_bill'], errors='coerce')
tips_df_miss
```

숫자로 변경할 수 없는 데이터를 누락값으로 변경한 후 07_pandas_4 파일의 누락값 처리하기를 이어 실행하면 된다.
