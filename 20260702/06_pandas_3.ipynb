분석하기 좋은 데이터

분석하기 좋은 데이터는 데이터 집합을 분석하기 좋은 형태로 만들어 놓은 것을 말하고 데이터 분석 작업에서 데이터 정리는 아주 중요한 작업으로 실제 데이터 분석작업의 70% 이상을 차지하고 있는 작업이 데이터 정리(전처리) 작업이다.  
분석하기 좋은 데이터는 데이터를 분석 목적에 맞는 데이터를 모아서(수집) 새로운 표(table, 데이터프레임)를 만들어야 하고 측정한 값(데이터 1건)은 행으로 변수(특성, 속성)들을 열로 고성하며 이 모든 조건을 만족하는 데이터를 깔끔한 데이터(good data)라고 부른다.

```python
df1 = pd.read_csv('./data/concat_1.csv')
df1
```

```python
df2 = pd.read_csv('./data/concat_2.csv')
df2
```

```python
df3 = pd.read_csv('./data/concat_3.csv')
df3
```

concat() 메소드는 데이터프레임을 연결한다.  
연결하려는 데이터프레임을 리스트에 담아서 concat() 메소드에 전달하면 연결한 데이터프레임을 리턴한다.  
concat([데이터프레임1, 데이터프레임2, ...][, axis])  
axis 속성을 생략하면 0이 기본값이고 위에서 아래 방향으로 연결하고 axis=1로 지정하면 왼쪽에서 오른쪽 방향으로 연결한다.

```python
row_concat = pd.concat([df1, df2, df3]) # pd.concat([df1, df2, df3], axis=0), 아래로 붙인다.
row_concat
```

```python
col_concat = pd.concat([df1, df2, df3], axis=1) # 옆으로 붙인다.
col_concat
```

데이터프레임에 데이터 1건을 추가하기

```python
row_series = pd.Series(['n1', 'n2', 'n3', 'n4']) # 시리즈
row_series
```

```python
pd.concat([df1, row_df])
```

DataFrame()의 인수로 ['n1', 'n2', 'n3', 'n4']와 같이 1차원 리스트를 전달하면 데이터 하나하나를 개별 데이터로 취급해서 열방향 데이터프레임이 생성된다.  
[['n1', 'n2', 'n3', 'n4']]와 같이 2차원 리스트로 전달하면 1차원 리스트 전체를 1개의 데이터로 취급해서 행 방향 데이터프레임이 생성된다.

```python
row_df = pd.DataFrame([['n1', 'n2', 'n3', 'n4']]) # 행 방향 데이터프레임
row_df
```

```python
pd.concat([df1, row_df])
```

DataFrame()으로 데이터프레임을 생성할 때 열 인덱스를 지정하지 않으면 pandas가 자동으로 0, 1, 2, 3, ...와 같은 방식으로 열 인덱스를 지정하기 때문에 columns 속성으로 열 인덱스를 전달한다.

```python
# row_df = pd.DataFrame([['n1', 'n2', 'n3', 'n4']], columns=['A', 'B', 'C', 'D'])
row_df = pd.DataFrame([['n1', 'n2', 'n3', 'n4']], columns=df1.columns)
row_df
```

```python
pd.concat([df1, row_df])
```

concat() 메소드로 데이터프레임을 연결하면 발생되는 중복되는 인덱스를 무시하고 인덱스를 다시 부여하려면 ignore_index 속성의 속성값을 True로 지정하면 인덱스가 0부터 1씩 증가하는 값으로 다시 부여된다.

```python
row_concat = pd.concat([df1, df2, df3], ignore_index=True)
row_concat
```

```python
col_concat = pd.concat([df1, df2, df3], axis=1, ignore_index=True)
col_concat
```

```python
print(df1.columns)
df2.columns = ['E', 'F', 'G', 'H']
print(df2.columns)
df3.columns = ['A', 'C', 'F', 'H']
print(df3.columns)
```

위와 같이 데이터프레임의 열 인덱스를 다시 부여하고 concat() 메소드로 연결하면 데이터프레임에 없는 열 인덱스의 데이터는 NaN(누락값, 결측치)으로 처리된다.

```python
pd.concat([df1, df2, df3], ignore_index=True)
```

concat() 메소드의 join 속성의 기본값은 'outer'이고 공통으로 데이터가 들어있는(NaN이 없는) 열에 신경쓰지 않고 연결한다. => NaN이 대량 발생된다.

```python
pd.concat([df1, df2, df3], ignore_index=True, join='outer')
```

concat() 메소드의 join 속성값을 'inner'로 지정하면 공통으로 데이터가 들어있는(NaN이 없는) 열만 연결한다.

```python
pd.concat([df1, df2, df3], ignore_index=True, join='inner')
```

```python
pd.concat([df1, df3], ignore_index=True, join='inner')
```

```python
pd.concat([df2, df3], ignore_index=True, join='inner')
```

```python
pd.concat([df1, df2, df3], axis=1, ignore_index=True, join='outer')
```

merge() 메소드는 2개의 데이터프레임을 병합한다.  
기본값으로 내부 조인을 실행하며 merge() 메소드를 실행한 데이터프레임이 left로 지정되고 merge() 메소드의 인수로 지정한 데이터프레임이 right로 지정된다.  
left_on 속성과 right_on 속성은 각각 왼쪽과 오른쪽의 데이터프레임에서 일치해야 할 열 이름을 지정하고 how 속성으로 두 데이터프레임의 병합 방법을 지정한다.  
왼쪽데이터프레임.merge(오른쪽데이터프레임, left_on, right_on[, how])

survey_person.csv => 조사한 사람  
survey_site.csv => 조사 위치  
survey_survey.csv => 기상 정보  
survey_visited.csv => 조사한 날짜

```python
person = pd.read_csv('./data/survey_person.csv')
person
```

```python
site = pd.read_csv('./data/survey_site.csv')
site
```

```python
survey = pd.read_csv('./data/survey_survey.csv')
survey
```

```python
visited = pd.read_csv('./data/survey_visited.csv')
visited
```

site 데이터프레임과 visited 데이터프레임의 merge

```python
print(site.columns)
print(visited.columns)
```

how 속성을 두 데이터프레임을 병합하는 방법을 지정한다. 기본값은 how='inner' 이다.  
how 속성을 생략하거나 'inner'(내부 조인)로 지정하면 두 데이터프레임에 공통적으로 포함된 값이 존재하는 데이터만 병합된다.

```python
# site.merge(visited, left_on='name', right_on='site')
site.merge(visited, left_on='name', right_on='site', how='inner')
```

```python
site.merge(visited, left_on='name', right_on='site', how='outer')
```

```python
site.merge(visited, left_on='name', right_on='site', how='left')
```

```python
site.merge(visited, left_on='name', right_on='site', how='right')
```

person 데이터프레임과 survey 데이터프레임의 merge

```python
print(person.columns)
print(survey.columns)
```

```python
p2s = person.merge(survey, left_on='ident', right_on='person', how='inner')
p2s
```

how 속성을 'outer'(외부 조인)로 지정하면 두 데이터프레임의 모든 데이터가 병합된다.

```python
p2s = person.merge(survey, left_on='ident', right_on='person', how='outer')
p2s
```

how 속성을 how='left'(왼쪽 외부 조인)로 지정하면 왼쪽 데이터프레임의 모든 데이터와 오른쪽 데이터프레임에 있는 공통적으로 포함된 데이터가 병합된다.

```python
p2s = person.merge(survey, left_on='ident', right_on='person', how='left')
p2s
```

how 속성을 how='right'(오른쪽 외부 조인)로 지정하면 오른쪽 데이터프레임의 모든 데이터와 왼쪽 데이터프레임에 있는 공통적으로 포함된 데이터가 병합된다.

```python
p2s = person.merge(survey, left_on='ident', right_on='person', how='right')
p2s
```

visited 데이터프레임과 survey 데이터프레임의 merge

```python
print(visited.columns)
print(survey.columns)
```

```python
v2s = visited.merge(survey, left_on='ident', right_on='taken', how='inner')
v2s
```

p2s 데이터프레임과 v2s 데이터프레임의 merge

```python
print(p2s.columns)
print(v2s.columns)
```

left_on 속성과 right_on 속성에 리스트로 묶어서 여러개의 열 인덱스를 전달할 수 있다.  
merge() 메소드를 실행하는 두 데이터프레임에 공통된 열 인덱스가 있으면 왼쪽 데이터프레임의 열 인덱스에는 '_x'라는 접미사가 자동으로 추가되고 오른쪽 데이터프레임의 열 인덱스에는 '_y'라는 접미사가 자동으로 추가된다.

```python
p2s.merge(v2s, left_on=['ident', 'taken', 'quant', 'reading'], right_on=['person', 'taken', 'quant', 'reading'], how='inner')
```
