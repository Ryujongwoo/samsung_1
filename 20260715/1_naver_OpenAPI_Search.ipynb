https://developers.naver.com/docs/serviceapi/search/blog/blog.md#python

```python
import os
import sys
import urllib.request
client_id = "PikZgd6cgsqrlllmiO_s" # 네이버 개발자 센터에서 발급받은 Client ID 값을 입력한다.
client_secret = "8ec8ohtkx7"# 네이버 개발자 센터에서 발급받은 Client Secret 값을 입력한다.
encText = urllib.parse.quote("홍길동")
url = "https://openapi.naver.com/v1/search/blog?query=" + encText # JSON 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id", client_id)
request.add_header("X-Naver-Client-Secret", client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

블로그 검색

```python
searchText = input('검색어를 입력하세요: ')
# 요청 url
# https://openapi.naver.com/v1/search/blog.xml => xml 형태로 검색 결과를 리턴한다.
# https://openapi.naver.com/v1/search/blog.json => json 형태로 검색 결과를 리턴한다.
# 파라미터: url 뒤에 '?'붙이고 입력한다.
# query(필수): 검색어, UTF-8로 인코딩되어야 한다.
# display(선택): 한 번에 표시할 검색 결과 개수(기본값: 10, 최댓값: 100)를 지정한다.
# start(선택): 검색 시작 위치(기본값: 1, 최댓값: 1000)를 지정한다.
# sort(선택): 검색 결과 정렬 방법, 기본값은 'sim'이고 정확도순으로 내림차순 정렬, 'date'는 작성일의 내림차순 정렬
targetSite = 'https://openapi.naver.com/v1/search/blog.json?query=' + searchText
# print(targetSite)

response = requests.get(targetSite, headers=headers) # 네이버 검색 API에게 요청한다.
# print(response)
# print(type(response))
# print(response.text)
# print(type(response.text))
searchResult = response.json() # 응답받은 json 형태의 문자열을 파이썬에서 처리할 수 있도록 리스트나 딕셔너리로 변환한다.
# print(type(searchResult))
# print(searchResult.keys())

items = searchResult['items']
# print(items)
for item in items:
    print(item['link'], end='\n' + '=' * 100 + '\n')
```

오타 변환  
오타 변환은 문장단위로 실행하지 않고 단어 단위로 실행하고 오타가 발견되면 올바른 단어를 리턴하고 오타가 발견되지 않으면 아무것도 리턴하지 않는다.

```python
searchText = input('오타를 수정하려는 문장을 입력하세요: ').split()
result = []
print(searchText)

for text in searchText:
    targetSite = 'https://openapi.naver.com/v1/search/errata.json?query=' + text
    response = requests.get(targetSite, headers=headers)
    searchResult = response.json()
    # print(searchResult['errata'])
    errata = searchResult['errata']
    if errata == '':
        # print(text, end=' ')
        result.append(text)
    else:
        # print(errata, end=' ')
        result.append(errata)
print()

print(' '.join(result))
```
