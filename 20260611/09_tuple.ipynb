튜플 만들기

튜플은 리스트와 거의 비슷하지만 약간의 차이가 있다.  
리스트는 []로 데이터와 데이터를 감싸고 튜플은 ()로 감싼다.  
리스트는 데이터의 수정 및 삭제가 가능하지만 튜플은 불가능하다.

```python
# 빈 튜플 만들기
a = ()
print(a)
print(type(a))

# 생성자 함수를 사용해서 만들기
b = tuple()
print(b)
print(type(b))
```

```python
# 데이터가 1개인 리스트
a = [1]
print(a)
print(type(a))

# 데이터가 1개인 튜플
# 데이터가 1개인 튜플을 만들러는 경우 ()만 사용하면 ()를 튜플 정의로 인식하는 것이 아니고 일반 연산식의 우선순위를 지정하는 ()로 인식한다.
b = (1)
print(b)
print(type(b))

# 데이터가 1개인 튜플을 만들려면 ()안의 데이터 뒤에 반드시 ','를 입력해야 한다.
c = (1,)
print(c)
print(type(c))
```

```python
# 초기치를 가지는 튜플
a = (1, 2, 3)
print(a)
b = ('a', 'b', 'c')
print(b)
c = (1, 'a', 2, 'b')
print(c)
d = ('a', 'b', 'c', (1, 2, 3))
print(d)
e = ('a', 'b', 'c', [1, 2, 3])
print(e)
f = ['a', 'b', 'c', (1, 2, 3)]
print(f)
```

튜플은 리스트와 같이 인덱싱과 슬라이싱이 가능하다.

```python
a = ('a', 'b', ('ab', 'cd'))
print(a)
print(a[0])
print(a[-1])
print(a[-1][0])
print(a[0:2])
print(a[:2])
print(a[1:3])
print(a[1:])
print(a[:])
print(a[::-1])
```

튜플은 리스트와 같이 덧셈(연결) 연산과 곱셈(반복) 연산이 가능하다.

```python
a = (1, 2, 3)
b = (1, 2, 3, 4)
print(a + b)
print(a * 3)
```

튜플은 리트와 달리 데이터의 수정 및 삭제를 할 수 없으므로 에러가 발생된다.

```python
a = (1, 2, 3, 4, 5)
a[0] = 999
```

```python
a = (1, 2, 3, 4, 5)
del a[0]
```

```python
# del 명령을 사용해서 튜플 자체를 삭제하는 것은 가능하다.
a = (1, 2, 3, 4, 5)
print(a)
del a
print(a)
```

튜플을 만들 때 ()를 사용하지 않고 데이터를 ','로 구분해서 나열하면 튜플로 인식된다.

```python
a = 1, 2, 3, 4, 5
print(a)
print(type(a))
```
