set 만들기

set은 수학의 집합과 비슷한 형태로 중복되는 데이터의 입력을 허용하지 않고 입력하는 데이터의 순서와 실제로 저장되는 데이터의 순서가 일치하지 않아서 인덱싱과 슬라이싱을 사용할 수 없다.  
딕셔너리처럼 {}를 사용해서 데이터를 감싸고 key는 존재하지 않고 value만 존재한다.

```python
# 빈 set 만들기
a = {} # 빈 set은 {}를 사용해서 만들수 없다. => 딕셔너리로 인식된다.
print(type(a))

# 빈 set을 만들때 {}만 사용하면 딕셔너리가 생성되기 때문에 반드시 생성자 함수를 사용해서 만든다.
b = set()
print(type(b))
```

```python
# 초기치를 가지는 set
a = {1, 2, 3, 4}
print(a)
print(type(a))
```

```python
# set은 중복되는 데이터의 입력을 허용하지 않는다.
a = {1, 5, 1, 1, 1, 1, 3, 5, 7, 7}
print(a)
```

```python
# 생성자 함수의 인수로 리스트나 튜플을 넘겨서 set을 만들 수 있다.
a = set([1, 2, 3, 4, 1])
print(a)

b = set((1, 2, 3, 4, 1))
print(b)

b = set(1, 2, 3, 4, 1) # 에러, set의 생성자 함수로 튜플을 넘길때는 반드시 ()로 묶어서 사용해야 한다.
```

```python
# add() 메소드로 데이터를 set에 추가할 수 있다. => 1개씩 추가한다.
a = set()
print(a)
a.add('홍길동')
print(a)
a.add('임꺽정')
print(a)
a.add('장길산')
print(a)
a.add('일지매')
print(a)
```

```python
# update() 메소드로 데이터를 set에 추가할 수 있다. => 일괄 추가한다.
a = {'일지매', '장길산', '홍길동', '임꺽정'}
print(a)

b = ['손오공', '저팔계', '사오정', '삼장법사']
print(b)
a.update(b)
print(a)

c = '피콜로', '마인부우', '베지터', '프리저'
print(c)
a.update(c)
print(a)
```

```python
# remove() 메소드로 인수로 지정한 데이터를 set에서 제거할 수 있다.
a = {'일지매', '장길산', '홍길동', '임꺽정'}
print(a)

a.remove('장길산')
print(a)
a.remove('장길산') # set에 존재하지 않는 데이터를 삭제하려 했으므로 에러가 발생된다.
print(a)
```

```python
# discard() 메소드로 인수로 지정한 데이터를 set에서 제거할 수 있다.
a = {'일지매', '장길산', '홍길동', '임꺽정'}
print(a)

a.discard('장길산')
print(a)
a.discard('장길산') # set에 존재하지 않는 데이터를 삭제하려 하면 메소드가 무시된다.
print(a)
```

```python
# clear() 메소드로 set에 저장된 모든 데이터를 제거할 수 있다.
a = {'일지매', '장길산', '홍길동', '임꺽정'}
print(a)

a.clear()
print(a)

# del 명령을 사용해서 set 자체를 제거할 수 있다.
del a
print(a)
```

집합 연산

```python
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

# 합집합
print(a.union(b))
print(a | b)

# 교집합
print(a.intersection(b))
print(a & b)

# 차집합
print(a.difference(b))
print(a - b)

# 대칭 차집합 => 합집합 - 교집합
print(a.symmetric_difference(b))
print(a ^ b)
```
