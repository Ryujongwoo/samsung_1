while

while은 조건이 참인동안 반복하고 조건이 거짓이 되는 순간 반복을 종료한다.  
while의 최초 진입 조건이 거짓일 경우 while 블록을 한 번도 실행하지 않는다.  
while은 반복을 하기는 해야하는데 몇 번 반복해야하는지 모를때 사용하고 for는 반복횟수를 알고있을때 사용한다.

while 조건식:  
&nbsp;&nbsp;&nbsp;&nbsp;조건식이 참인 동안 실행할 문장  
&nbsp;&nbsp;&nbsp;&nbsp;...

```python
# 1부터 10까지의 합계
i = 0 # 1에서 10까지 1씩 증가할 변수
total = 0 # 1에서 10까지의 합계를 기억할 변수

while i < 10:
    i += 1
    total += i
# ===== while
print(total)
```

```python
i = 0
total = 0

# 의도적으로 무한루프를 발생시키려면 조건을 항상 참이되도록 하면된다.
# 의도적으로 발생시킨 무한루프는 반드시 break 명령을 이용해서 무한루프를 탈출시키는 처리를 해야한다.
while True:
    i += 1
    total += i
    if i == 10:
        break
    # ===== if
# ===== while
print(total)
```

```python
# break 명령은 반복 범위의 남아있는 문장들을 실행하지 않고 반복을 탈출한다.
# continue 명령은 반복 범위의 남아있는 문장들을 실행하지 않고 다음 반복을 실행한다.
i = 0
total = 0
while i < 10:
    i += 1
    if i % 2 == 0:
        continue
    total += i
# ===== while
print(total)
```

```python
menu = 0
while menu != 5:
    while True:
        print('=' * 40)
        print(' 1.입력  2.보기  3.수정  4.삭제  5.종료 ')
        print('=' * 40)
        menu = int(input('원하는 메뉴를 입력하고 enter 키를 누르세요: '))

        # 메뉴가 1 ~ 5 사이의 값이 입력되었다면 무한루프를 탈출시킨다.
        # if 1 <= menu <= 5:
        # if menu >= 1 and menu <= 5:
        if menu in [1, 2, 3, 4, 5]:
            break
        else:
            print('메뉴는 1 ~ 5 사이의 값만 입력해야 합니다.')
        # ===== if
    # ===== while

    # 여기까지 넘어왔다면 menu에는 1 ~ 5 사이의 정수가 입력된 것이다.
    if menu == 1:
        print('입력 작업 실행')
    elif menu == 2:
        print('보기 작업 실행')
    elif menu == 3:
        print('수정 작업 실행')
    elif menu == 4:
        print('삭제 작업 실행')
    # ===== if
# ===== while
print('프로그램을 종료합니다. 바이바이~~~~~')
```
