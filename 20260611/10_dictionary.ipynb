딕셔너리 만들기

딕셔너리는 key와 value가 1:1로 대응되는 관계를 나타내는 자료형으로 key와 key에 할당되는 데이터(value)가 한 쌍이 되는 자료형이다.  
딕셔너리는 {}로 감싸주고 {key: value, key: value, ...}와 같은 형태를 가진다.  
일반적으로 key는 string 타입을 많이 사용하고 value에는 저장할 데이터를 적어준다.

```python
# 빈 딕셔너리 만들기
a = {}
print(a)
print(type(a))

# 빈 딕셔너리는 생성자 함수를 사용해서 만들수도 있다.
b = dict()
print(b)
print(type(b))
```

```python
# 초기치를 가지는 딕셔너리
a = {'name': '홍길동', 'phone': '010-1234-5678', 'age': 20}
print(a)
```

딕셔너리에 데이터 추가 및 수정하기

```python
a = {'name': '홍길동', 'phone': '010-1234-5678', 'age': 20}
print(a)

# 딕셔너리에 존재하지 않는 key에 데이터를 할당하면 데이터가 추가된다.
# 딕셔너리에 존재하는 key에 데이터를 할당하면 데이터가 수정된다.
# 딕셔너리이름[key] = value
a['gender'] = '남자' # 추가
print(a)
a['gender'] = '여자' # 수정
print(a)
```

딕셔너리에 저장된 데이터 제거하기

```python
a = {'name': '홍길동', 'phone': '010-1234-5678', 'age': 20, 'gender': '여자'}
print(a)

# del 딕셔너리이름[key]
del a['gender']
print(a)

# clear() 메소드로 딕셔너리에 저장된 모든 데이터를 제거할 수 있다.
a.clear()
print(a)

# del 명령을 딕셔너리 자체를 제거할 수 있다.
del a
print(a)
```

딕셔너리에 저장된 데이터 얻어오기

```python
a = {'name': '홍길동', 'phone': '010-1234-5678', 'age': 20, 'gender': '여자'}
print(a)

# 딕셔너리이름.get(key)
print(a.get('age'))
print(a.get('gender'))
# get() 메소드는 딕셔너리에 존재하지 않는 key를 지정하면 None이 리턴된다.
print(a.get('email'))
# get() 메소드는 딕셔너리에 존재하지 않는 key를 지정했을 때 기본값을 2번째 인수로 지정한다.
print(a.get('email', '존재하지 않는 key 입니다.'))

# 딕셔너리이름[key]
print(a['age'])
print(a['gender'])
print(a['email'])
```

딕셔너리 전용 메소드

```python
a = {'name': '홍길동', 'phone': '010-1234-5678', 'age': 20, 'gender': '여자'}
print(a)

# keys() 메소드는 딕셔너리에 저장된 key 목록만 얻어온다.
print(a.keys())
# values() 메소드는 딕셔너리에 저장된 value 목록만 얻어온다.
print(a.values())
# items() 메소드는 딕셔너리에 저장된 key와 value를 쌍으로 묶은 튜플을 얻어온다.
print(a.items())
```
