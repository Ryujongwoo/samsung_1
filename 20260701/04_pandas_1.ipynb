iloc를 사용하면 0 ~ 1703사이의 행 번호를 이용해서 행 단위 데이터를 추출할 수 있고 1704와 같이 행 번호의 범위를 벗어나는 행 번호를 사용하면 에러가 발생된다.

```python
print(gapminder_df.iloc[0])
print('=' * 100)
print(gapminder_df.iloc[1703])
# print(gapminder_df.loc[1704]) # 존재하지 않는 행 번호의 값을 얻어오라 했기때문에 에러가 발생된다.
```

```python
# iloc는 행 번호를 지정할 때 음수를 사용할 수 있고 -1은 1703행, -2는 1702행, ... 과 같이 행 번호가 지정된다.
print(gapminder_df.iloc[-1])
```

```python
# iloc를 이용해서 여러 인덱스를 추출하려면 리스트에 추출하려는 인덱스를 담아서 iloc에 넘겨주면 된다.
print(gapminder_df.iloc[[1, 11, 111, 1111]])
```

loc와 iloc를 자유자재로 사용하려면 추출할 데이터의 행과 열을 지정하는 방법을 알아야 한다.  
데이터프레임.loc[[행], [열]], 데이터프레임.iloc[[행], [열]] 방법을 사용해야 하며 슬라이싱이나 range() 함수를 함께 사용하는 방법을 알아야 한다.

슬라이싱

```python
# 모든 행(':')의 데이터에 대해서 country, year, pop 열을 추출한다.
subset = gapminder_df[['country', 'year', 'pop']]
subset.head(3)
```

iloc의 'i'는 'integer'를 의미한다고 보면 된다. 따라서 loc를 사용할때는 숫자를 사용하면 안되고 iloc를 사용할때는 문자를 사용하면 안된다.

```python
subset = gapminder_df.loc[:, ['country', 'year', 'pop']]
# subset = gapminder_df.iloc[:, ['country', 'year', 'pop']] # iloc를 사용했는데 문자를 넘겨서 에러가 발생된다.
subset.head(3)
```

```python
# subset = gapminder_df.loc[:, [0, 2, 4]] # loc를 사용했는데 숫자를 넘겨서 에러가 발생된다.
subset = gapminder_df.iloc[:, [0, 2, 4]]
subset.head(3)
```

기초 통계 계산하기

groupby() 메소드의 인수로 그룹화하려는 열 이름을 지정해서 그룹화 할 수 있다.

```python
# year 열로 그룹화하고 lifeExp 열만 추출해서 평균(mean)을 계산한다.
print(gapminder_df.groupby('year')) # year 열로 그룹화
print(gapminder_df.groupby('year')['lifeExp']) # lifeExp 열만 추출
print(gapminder_df.groupby('year').lifeExp) # lifeExp 열만 추출
print(gapminder_df.groupby('year')['lifeExp'].mean())
print(gapminder_df.groupby('year').lifeExp.mean())
```

```python
# continent 열로 그룹화한 후 year 열로 그룹화하고 lifeExp 열만 추출해서 평균(mean)을 계산한다.
print(gapminder_df.groupby(['continent', 'year']).lifeExp.mean())
```
