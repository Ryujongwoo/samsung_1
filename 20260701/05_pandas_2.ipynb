pandas의 데이터프레임과 시리즈

데이터프레임과 시리즈는 파이썬의 리스트나 딕셔너리와 비슷하지만 데이터를 다루는데 더 특화되어 있다.  
리스트나 딕셔너리는 많은 양의 데이터를 저장하거나 조작할 수 있는 제공되는 메소드가 많치 않은 반면에 pandas의 데이터프레임과 시리즈는 많은 양의 데이터를 처리할 수 있을 뿐만아니라 스프레드시트 계열(엘셀 등)의 프로그램을 사용하는 것 처럼 행과 열 단위로 데이터를 조작할 수 있는 다양한 메소드와 속성을 제공한다.

시리즈 만들기

시리즈를 만들기 위해서는 리스트 또는 튜플을 Series() 클래스의 인수로 전달하면 된다.

```python
s = pd.Series(['banana', 40])
print(type(s))
print(s)
```

```python
s = pd.Series(('홍길동', '임꺽정'))
print(type(s))
print(s)
print(s.loc[0])
print(s.iloc[1])
```

시리즈를 생성할때 문자열을 인덱스로 지정할 수 있다.  
문자열을 인덱스로 지정하려면 index 속성으로 인덱스로 사용할 문자열을 리스트에 담아서 전달하면 된다.

```python
s = pd.Series(('홍길동', '임꺽정'), index=['name1', 'name2'])
print(s)
print(s.index)
print(s.loc['name1'])
print(s.iloc[0])
```

데이터프레임 만들기

데이터프레임을 만들기 위해서는 딕셔너리 DataFrame() 클래스의 인수로 전달하면 된다.  
딕셔너리의 key가 데이터프레임의 열 이름이 되고 딕셔너리의 value(반드시 리스트 형태)는 데이터가 된다.

```python
data = {
    'name': ['홍길동', '임꺽정'],
    'job': ['도둑넘', '도적놈'],
    'born': ['1569-04-20', '1875-06-10'],
    'died': ['1620-08-25', '1950-01-15'],
    'age': [51, 75]
}
print(type(data))
print(data)
df = pd.DataFrame(data)
print(type(df))
df
```

```python
data = {
    'job': ['도둑넘', '도적놈'],
    'born': ['1569-04-20', '1875-06-10'],
    'died': ['1620-08-25', '1950-01-15'],
    'age': [51, 75]
}
# 데이터프레임을 만들때 문자열을 인덱스로 지정할 수 있다.
# 문자열을 인덱스로 지정하려면 index 속성으로 인덱스로 사용할 문자열을 리스트에 담아서 전달하면 된다.
# columns 속성으로 데이터프레임으로 만들려는 딕셔너리의 key 중에서 데이터프레임으로 만들 key를 지정할 수 있다.
df = pd.DataFrame(data, index=['홍길동', '임꺽정'], columns=['job', 'age'])
df
```

시리즈 다루기

```python
data = {
    'job': ['도둑넘', '도적놈'],
    'born': ['1569-04-20', '1875-06-10'],
    'died': ['1620-08-25', '1950-01-15'],
    'age': [51, 75]
}
df = pd.DataFrame(data, index=['홍길동', '임꺽정'])
df
```

```python
# 데이터프레임에서 시리즈를 선택하려면 loc 속성에 인덱스(열 또는 행 이름)를 전달하거나 iloc 속성에 열 또는 행의 일련번호를 전달한다.
row = df.iloc[0]
print(type(row))
print(row)
print('=' * 100)
row = df.loc['홍길동']
print(type(row))
print(row)
```

```python
col = df.iloc[:, 0]
print(type(col))
print(col)
print('=' * 100)
col = df.loc[:, 'job']
print(type(col))
print(col)
```

index, values 속성과 keys() 메소드

```python
print(row)
print(row.index) # 시리즈의 인덱스를 얻어온다.
print(row.keys()) # keys() 메소드도 index 속성과 같이 시리즈의 인덱스를 얻어온다.
print(row.values) # 시리즈의 데이터를 얻어온다.
print(type(row.values))
```

```python
print(col)
print(col.index)
print(col.keys())
print(col.values)
print(type(col.values))
```

브로드캐스팅

시리즈나 데이터프레임에서 여러 데이터에 대해서 한 번에 연산하는 것을 브로드캐스팅이라 한다. 다대일 연산  
여러개의 값을 가지는 데이터 벡터라하고 단순 크기를 나타내는 데이터 1개를 스칼라라 부른다.

scientists_df 데이터프레임의 Age 시리즈에서 평균 나이(59.125)보다 나이가 많은 데이터만 추출한다.

```python
bool_values = ages > ages.mean() # 브로드캐스팅
print(bool_values)
```

```python
print(ages[bool_values]) # 불린 인덱싱
```

```python
print(ages[[False, True, True, True, False, False, False, True]])
```

```python
print(ages[ages > ages.mean()]) # 브로드캐스팅을 이용한 불린 인덱싱
```

```python
for age in ages:
    print('{:4d}'.format(age), end=' ')
```

```python
# 같은 길이의 벡터로 연산을 하면 결과값으로 같은 길이의 벡터가 출력된다.
for age in ages + ages:
    print('{:4d}'.format(age), end=' ')
print()
for age in ages * ages:
    print('{:4d}'.format(age), end=' ')
```

```python
# 벡터에 스칼라 값을 연산하면 각각의 벡터값에 스칼라 값이 연산된 결과가 출력된다.
for age in ages + 100:
    print('{:4d}'.format(age), end=' ')
print()
for age in ages * 2:
    print('{:4d}'.format(age), end=' ')
```

```python
# 길이가 서로 다른 벡터를 연산하면 같은 인덱스를 가지는 값만 계산된다.
print(pd.Series([1, 100]))
# pd.Series([1, 100])의 0, 1번째 인덱스만 계산되고 나머지는 계산할 수 없으므로 누락값(결측치, NaN, nan, NA, na)으로 처리된다.
print(ages + pd.Series([1, 100]))
```

시리즈 정렬

sort_index() 메소드는 인덱스를 정렬하고 sort_values() 메소드는 데이터를 정렬한다.  
sort_index() 메소드와 sort_values() 메소드는 별도의 설정을 하지않으면 오름차순으로 정렬되지만 내림차순으로 정렬하려면 ascending 속성값을 False로 지정한다.

```python
print(ages)
print(ages.sort_index()) # 오름차순 정렬, ages.sort_index(ascending=True)
print(ages.sort_index(ascending=False)) # 내림차순 정렬
```

```python
print(ages)
print(ages.sort_values()) # 오름차순 정렬, ages.sort_values(ascending=True)
print(ages.sort_values(ascending=False)) # 내림차순 정렬
```

```python
# 벡터와 벡터의 연산은 같은 인덱스를 가지는 값끼리 수행된다.
for age in ages + ages.sort_index(ascending=False):
    print('{:4d}'.format(age), end=' ')
```

날짜를 문자열 형태로 저장한 데이터는 날짜 및 시간에 관련된 작업을 할 수 있도록 데이터 타입을 변경해야 한다.  
to_datetime() 메소드는 str(문자열) 타입의 데이터를 datetime(날짜 및 시간) 타입으로 변경한다.

```python
born_datetime = pd.to_datetime(scientists_df.Born)
print(born_datetime)
died_datetime = pd.to_datetime(scientists_df.Died)
print(died_datetime)
```

```python
scientists_df.Born = pd.to_datetime(scientists_df.Born)
scientists_df.Died = pd.to_datetime(scientists_df.Died)
```

파생 변수(새로운 열, 계산에 의한 변수) 추가하기 => 데이터프레임의 한 열을 변수라고도 부른다.

```python
# str 타입을 datetime 타입으로 변경한 born_datetime, died_datetime를 데이터프레임에 파생 변수로 추가한다.
print(scientists_df.shape)
# 데이터프레임에 존재하는 열에 데이터를 넣으면 기존 데이터가 수정되고 존재하지 않는 열에 데이터를 넣으면 데이터가 추가된다.
# 데이터프레임에 파생 변수 추가는 데이터프레임이름['파생변수이름'] = 데이터 형태만 가능하다.
scientists_df['born_dt'] = born_datetime
# scientists_df.get('born_dt') = born_datetime # 사용 불가능
# scientists_df.born_dt = born_datetime # 사용 불가능
print(scientists_df.shape)
scientists_df['died_dt'] = died_datetime
print(scientists_df.shape)
```

```python
scientists_df['age_days'] = scientists_df.died_dt - scientists_df.born_dt
print(scientists_df.shape)
scientists_df
```
