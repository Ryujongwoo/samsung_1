모든 프로그래밍 언어는 함수의 실행 결과를 return 뒤에 딱 1개만 붙여서 return 시킬 수 있다. => 파이썬도 예외는 아니다.

```python
def swap(a, b):
    temp = a
    a = b
    b = temp
    # 데이터를 ','로 구분해서 나열하면 ()로 묶지않아도 튜플로 인식한다.
    # return을 ()를 사용하지 않은 튜플 형태로 return 시키면 return 값이 2개 이상인 것 같은 효과를 낸다.
    return a, b # (a, b) 형태의 튜플을 리턴시킨다.
```

```python
a = 3
b = 4
print(f'a = {a}, b = {b}')
[a, b] = [b, a]
print(f'a = {a}, b = {b}')
(a, b) = (b, a)
print(f'a = {a}, b = {b}')
a, b = b, a
print(f'a = {a}, b = {b}')
result = swap(a, b)
print(type(result))
print(f'result[0] = {result[0]}, result[1] = {result[1]}')
a, b = swap(a, b)
print(f'a = {a}, b = {b}')
```
