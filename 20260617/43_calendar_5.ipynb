같은 module을 여러 프로젝트에서 사용해야 할 경우 각각의 프로젝트에 같은 module을 복사해서 사용할 경우 유지보수시 문제점이 발생될 수 있다.  
import 할 module이 현재 사용중인 컴퓨터 어딘가에 있을 경우 sys module을 import 하고 sys.path의 append() 메소드로 module이 위치한 기본 경로를 추가해서 사용한다.

```python
import sys
# module 경로 추가하기
sys.path.append('C:\myModule') # '\' 때문에 에러가 발생되면 '\'를 '\\'나 '/'로 수정한다.
```

```python
year, month = map(int, input('달력을 출력할 년과 월을 입력하세요: ').split())
print('=' * 28)
print('         {:4d}년{:2d}월'.format(year, month))
print('=' * 28)
print(' 일  월  화  수  목  금  토 ')
print('=' * 28)
for i in range(cm.weekDay(year, month, 1)):
    print('    ', end='')
for i in range(1, cm.lastDay(year, month) + 1):
    print(' {:2d} '.format(i), end='')
    if cm.weekDay(year, month, i) == 6 and i != cm.lastDay(year, month):
        print()
print('\n' + '=' * 28)
```
