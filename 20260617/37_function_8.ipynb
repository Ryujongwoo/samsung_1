함수 내부의 변수와 함수 외부의 변수는 이름이 같더라도 독립된 기억공간을 가진다.

```python
def func9(var1):
    var1 += 1
    print(f'func9() 함수 내부의 변수 var1: {var1}')
```

```python
var1 = 1
func9(var1)
print(f'func9() 함수 외부의 변수 var1: {var1}')
```

global 예약어를 사용하면 함수 내부에서 함수 외부의 변수를 사용할 수 있다.

```python
def func10():
    # C/C++ 언어의 call by reference(참조에 의한 호출) 방식처럼 사용된다. => side effect(부작용) 현상이 발생될 수 있으니 주의해서 사용해야 한다.
    # global로 사용할 변수는 함수의 인수로 선언하면 안된다.
    # global로 선언된 변수는 함수 내부에서 값을 변경하면 함수 외부의 값도 같이 변경된다.
    global var2 # 함수 외부에서 선언된 변수를 사용하게다고 선언한다.
    var2 += 1
    print(f'func10() 함수 내부의 변수 var2: {var2}')
```

```python
var2 = 999
func10()
print(f'func10() 함수 외부의 변수 var2: {var2}')
```
