 넘겨받는 인수를 딕셔너리 형태로 받는 함수 => 딕셔너리 변환 인자는 '**변수이름' 형태로 만든다.

```python
def func8(**args):
    print(args)
    print(type(args))
```

```python
func8()
```

'**변수이름' 형태로 인수를 사용하는 경우 함수에는 key와 key에 할당할 value를 'key=value' 형태로 전달한다.

```python
# 'key=value' 형태로 데이터를 전달하지 않아서 에러가 발생된다.
func8(1000)
```

```python
func8(apple=1000)
```

```python
func8(apple=1000, banana=2000, water_melon=20000)
```
