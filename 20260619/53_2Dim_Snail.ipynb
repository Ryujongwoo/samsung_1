```python
def pprint(data):
    for i in range(len(data)):
        for j in range(len(data[0])):
            print('{:3d}'.format(data[i][j]), end=' ')
        print()
```

```python
a = [[0] * 5 for _ in range(5)]
n = 0  # 1 ~ 25까지 1씩 증가하는 변수
s = 1  # 행과 열이 1또는 -1씩 증가 및 감소 처리에 사용하는 변수
i = 0  # 행
j = -1 # 열
k = 5  # 숫자를 채우는 반복문의 반복횟수 제어에 사용하는 변수

while True:
    # 행 방향으로 숫자를 채운다.
    for p in range(1, k + 1):
        n += 1
        j += s
        a[i][j] = n
    # ===== for
    # k값을 1감소 시킨다.
    k -= 1
    # 무한루프 탈출조건
    if k == 0:
        break
    # 열 방향으로 숫자를 채운다.
    for p in range(1, k + 1):
        n += 1
        i += s
        a[i][j] = n
    # ===== for
    # 행과 열에 숫자가 채워지는 방향을 변경한다.
    s *= -1
# ===== while
pprint(a)
```

3이상인 홀수를 입력받아 달팽이 배열을 출력한다.

```python
while True:
    m = int(input('3이상인 홀수를 입력하세요: '))
    if m >= 3 and m % 2 == 1:
        break
    # ===== if
    print('3이상인 홀수를 넣으라고 말했잖아.....')
# ===== while

a = [[0] * m for _ in range(m)]
n = 0; s = 1; i = 0; j = -1; k = m

while n < m ** 2:
    for p in range(1, k + 1):
        n += 1
        j += s
        a[i][j] = n
    # ===== for
    k -= 1
    for p in range(1, k + 1):
        n += 1
        i += s
        a[i][j] = n
    # ===== for
    s *= -1
# ===== while
pprint(a)
```

중심부터 바깥쪽으로 채워지는 달팽이 배열 만들기

```python
a = [[0] * 5 for _ in range(5)]
# 2차원 리스트의 정가운데 위치에 1이 채워져있다는 가정을 하고 시작한다.
i = 2; j = 2; n = 1; a[i][j] = n; s = 1
# 안쪽에서 바깥쪽으로 숫자를 채우기때문에 반복 횟수를 제어하는 변수는 1로 시작한다.
k = 1
# 마지막 숫자까지 채웠나 확인하는 변수
flag = False

while True:
    # k += 1 # 3시 방향으로 숫자를 채운다.
    for p in range(1, k):
        n += 1; j += s
        if n > 25:
            flag = True
            break
        # ===== if
        a[i][j] = n
    # ===== for
    # 위의 반복문이 정상적으로 종료되면 flag 변수값이 False이므로 다음 반복을 진행한다.
    # 위의 반복문이 break에 의해 종료되면 flag 변수값이 True이므로 프로그램을 종료한다.
    if flag:
        break
    for p in range(1, k):
        n += 1; i += s
        a[i][j] = n
    # ===== for
    s *= -1
    k += 1 # 9시 방향으로 숫자를 채운다.
# ===== while
pprint(a)
```

```python
while True:
    m = int(input('3이상인 홀수를 입력하세요: '))
    if m >= 3 and m % 2 == 1:
        break
    # ===== if
    print('3이상인 홀수를 넣으라고 말했잖아.....')
# ===== while

a = [[0] * m for _ in range(m)]
i = m // 2; j = m // 2 # 2차원 리스트의 정가운데 위치를 계산한다.
n = 1; a[i][j] = n; s = 1; k = 1; flag = False

while True:
    k += 1 # 3시 방향으로 숫자를 채운다.
    for p in range(1, k):
        n += 1; j += s
        if n > m ** 2:
            flag = True
            break
        # ===== if
        a[i][j] = n
    # ===== for
    if flag:
        break
    for p in range(1, k):
        n += 1; i += s
        a[i][j] = n
    # ===== for
    s *= -1
    # k += 1 # 9시 방향으로 숫자를 채운다.
# ===== while
pprint(a)
```
