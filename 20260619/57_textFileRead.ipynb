텍스트 파일을 읽기용으로 open 할 때는 디스크에 파일이 없을 수 있으므로 try ~ except를 사용해서 오류 처리를 한다.

```python
try:
    file = open('./data/data4.txt', 'r')
    # 텍스트 파일에 저장된 데이터가 몇 줄인지 알 수 없으므로 무한루프를 돌려서 읽는다.
    while True:
        char = file.read() # 텍스트 파일에 저장된 데이터를 문자 단위로 읽는다.
        if char == '': # 텍스트 파일에서 읽어들인 문자가 없으면
            break
        print(char)
    # ===== while
    # file.close() # 'r' 모드로 open 한 경우 close() 메소드를 실행하지않아도 상관없다.
except:
    print('파일이 디스크에 없습니다.')
```

```python
try:
    file = open('./data/data4.txt', 'r')
    lines = [] # 텍스트 파일의 데이터를 읽어서 저장할 빈 리스트를 선언한다.
    while True:
        line = file.readline() # 텍스트 파일에 저장된 데이터를 줄 단위로 읽어온다.
        if not line: # 텍스트 파일에서 읽어들인 줄이 없으면
            break
        # print(line, end='')
        # readline() 메소드로 데이터를 읽어들인 후 strip() 메소드로 문자열 앞, 뒤의 불필요한 공백을 제거한다.
        # print(line.strip())
        line = line.strip()
        if len(line) != 0:
            lines.append(line)
    # ===== while
    print('data4.txt 파일에서 읽기 완료')
except:
    print('파일이 디스크에 없습니다.')
print(lines)
```

```python
try:
    with open('./data/data4.txt', 'r') as file: 
        lines = []
        while True:
            line = file.readline()
            if not line:
                break
            line = line.strip()
            if len(line) != 0:
                lines.append(line)
        # ===== while
        print('data4.txt 파일에서 읽기 완료')
except:
    print('파일이 디스크에 없습니다.')
print(lines)
```

```python
try:
    file = open('./data/data4.txt', 'r')
    # readlines() 메소드는 텍스트 파일 전체를 한꺼번에 읽어서 리스트로 리턴한다.
    lines = file.readlines()
    print(type(lines))
    print('data4.txt 파일에서 읽기 완료')
except:
    print('파일이 디스크에 없습니다.')
print(lines)
```
