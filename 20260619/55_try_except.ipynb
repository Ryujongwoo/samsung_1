오류 처리

프로그램을 작성할 때 잘못동작되는 것을 방지하기 위해서 오류가 발생된다. => 오류가 발생되면 프로그램이 멈춘다.  
try ~ except를 사용하면 오류가 발생되었을때 별도의 처리를 하거나 오류를 무시하고 프로그램을 실행할 수 있다.  
오류가 발생될것으로 예상되는 문장을 try 블록에 코딩하고 try 블록을 실행하다가 오류가 발생되면 try 블록에 남아있는 코드를 실행하지않고 except 블록에 코딩된 내용을 실행한다.  
오류 발생 여부와 관계없이 무조건 실행해야 할 문장이 있다면 finally 블록에 코딩한다.

=================================================================================================================

try: # 필수  
&nbsp;&nbsp;&nbsp;&nbsp;오류가 발생될것으로 예상되는 문장  
&nbsp;&nbsp;&nbsp;&nbsp;...  
except [오류 이름 [as 변수 이름]]: # 필수 => 오류 이름과 변수 이름은 생략할 수 있다.  
&nbsp;&nbsp;&nbsp;&nbsp;오류가 발생되면 실행할 문장  
&nbsp;&nbsp;&nbsp;&nbsp;...  
finally: # 선택 => 필요할 경우 사용한다.  
&nbsp;&nbsp;&nbsp;&nbsp;오류 발생 여부와 관계없이 실행할 문장  
&nbsp;&nbsp;&nbsp;&nbsp;...

```python
try:
    print(4 / 1)
    print(4 / 0) # ZeroDivisionError 발생
    print(4 / 4) # 무시
except: # except 뒤에 오류 이름을 생략하면 모든 오류를 처리한다.
    # 오류 발생시 오류를 무시하려면 pass를 사용한다.
    # pass
    print('나눗셈은 0으로 할 수 없습니다.')
print('꺄~~~~~')
```

```python
data = [1, 2, 3]
try:
    print(data[2])
    print(data[3]) # IndexError 발생
    print(data[-1])
except:
    print('리스트의 인덱스 지정이 올바르지 않습니다')
print('꺄22~~~~~')
```

```python
data = [1, 2, 3]
string = '안녕~~~~~'
try:
    # print(4 / 0) # ZeroDivisionError 발생
    # print(data[3]) # IndexError 발생
    print(string)
    # print('%d' % string) # TypeError
# except 뒤에 오류 이름을 쓰면 지정된 오류만 처리한다.
except ZeroDivisionError:
    print('나눗셈은 0으로 할 수 없습니다.')
# except 뒤에 오류 이름을 쓰고 as 다음에 변수를 지정하면 오류 메시지가 변수에 저장된다.
except IndexError as e:
    print('리스트의 인덱스 지정이 올바르지 않습니다')
    print(e)
except:
    # except 뒤에 오류 이름을 지정하지 않으면 모든 오류를 처리하기때문 except 뒤에 코딩한 다른 except 문장들이 실행될 수 없기 때문에 SyntaxError가 발생된다.
    # except만 사용하는 경우 반드시 맨 마지막에 코딩해야 한다.
    # ZeroDivisionError, IndexError를 제외한 모든 오류가 이곳에 코딩된 문장을 실행한다.
    print('데이터 타입이 올바르지 않습니다.')
finally:
    # 오류 발생 여부와 관계없이 무조건 실행해야 할 문장이 있다면 finally 블록에 코딩한다.
    print('오류 발생 여부와 관계없이 무조건 실행해야 할 문장이 있다면 finally 블록에 코딩한다.')
print('꺄333~~~~~')
```

```python
# 숫자 데이터와 문자 데이터가 섞여서 입력되는 데이터를 정렬한다. 999가 입력되면 정렬을 시작한다.
number = [] # 숫자를 기억할 빈 리스트
alpha = []  # 문자열을 기억할 빈 리스트
while True:
    var = input('정렬할 데이터를 입력하세요(999가 입력되면 정렬을 시작합니다.): ')
    if var == '999':
        break
    # 입력받은 데이터가 '0' ~ '9' 사이의 문자로만 구성된 경우와 그렇치 않은 경우를 각각 number, alpha 리스트에 저장한다.
    try:
        number.append(int(var))
    except:
        alpha.append(var)
# ===== while
number.sort()
alpha.sort()
print(number + alpha)
```
