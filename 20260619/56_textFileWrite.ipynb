텍스트 파일로 출력하거나 텍스트 파일에서 읽어오는 경우 반드시 파일을 open 해야 한다.

파일 open

변수 = open('파일 이름', '파일 열기 모드')  
파일 열기 모드 => w(write, 쓰기(덮어쓰기)), a(append, 쓰기(추가)), r(read, 읽기)  
w는 지정한 파일이 없으면 파일을 만든 후 저장하고 파일이 있으면 내용을 지우고 저장한다. => 덮어쓰기  
a는 지정한 파일이 없으면 파일을 만든 후 저장하고 파일이 있으면 기존 내용 뒤에 추가해서 저장한다. => 추가  
r는 지정한 파일이 없으면 에러가 발생된다.  
w, a 모드로 파일을 열어서 작업했으면 반드시 파일을 close() 메소드를 사용해서 닫아야 정상적으로 저장된다.  
r 모드로 열어서 작업했으면 파일을 닫지 않아도 상관없다.

```python
# open() 함수에 파일 이름만 쓰면 현재 작업중인 프로그램이 위치한 경로에 위치한 파일이라는 의미이다.
# open('data.txt', 'w')를 사용해서 파일을 open 할 때, 'cp949'라는 메시지가 보이며 한글 인코딩 문제로 에러가 발생되면 open('data.txt', 'w', True, 'utf-8')
# 형식을 사용해서 open 시킨다.
file = open('data.txt', 'w') # data.txt 파일을 출력(덮어쓰기)용으로 open 한다.
for i in range(10):
    # 화면에 출력할때는 print() 함수를 사용하고 텍스트 파일로 출력할때는 write() 메소드를 사용한다.
    # 출력할 때 끝에 개행문자(\n)를 붙여야 텍스트 파일로 저장될 때 줄이 바뀐다.
    file.write('꺄아~~~~~' + '\n')
file.close()
print('data.txt 파일에 쓰기 완료')
```

경로 지정방법  

절대경로 지정방법과 상대경로 지정방법이 있다.  
절대경로는 디스크 드라이브의 최상위(root) 폴더(디렉토리)부터 파일이 저장된 폴더까지의 경로를 의미한다.  
상대경로는 현재 화면에 표시되는 소스프로그램 파일이 위치한 폴더부터 파일이 저장된 폴더까지의 경로를 의미한다.  

현재 작업중인 소스프로그램이 위치한 폴더가 아닌 다른 폴더에 파일을 저장하려면 반드시 파일이 저장된 경로를 지정해야 한다.  
D:\samsung_1\python_basic\workspace\data\data2.txt => 절대경로  
위와 같이 경로를 지정하면 이스케이프 시퀀스로 잘못 인식해서 오류가 발생될 수 있기때문에 '\'를 '\\' 또는 '/'로 수정한다.  
D:\\\\samsung_1\\\\python_basic\\\\workspace\\\\data\\\\data2.txt => 절대경로  
D:/samsung_1/python_basic/workspace/data/data2.txt => 절대경로  

'.' => 현재 작업중인 소스프로그램이 저장된 폴더를 의미한다.
'..' => 현재 작업중인 소스프로그램이 저장된 폴더의 한 단계 상위 폴더를 의미한다.  
.\data\data2.txt => 상대경로  
.\\\\data\\\\data2.txt => 상대경로  
./data/data2.txt => 상대경로

```python
file = open('./data/data2.txt', 'w')
for i in range(10):
    file.write('꺄아~~~~~' + '\n')
file.close()
print('data2.txt 파일에 쓰기 완료')
```

키보드로 입력한 문자열을 텍스트 파일에 저장한다.  
'quit'가 입력되면 저장을 종료한다. => 'quit'가 대문자로 입력되던 소문자로 입력되던 섞여서 입력되던 종료한다.

```python
file = open('./data/data3.txt', 'w')
while True:
    string = input('>>> ')
    if string.lower() == 'quit':
        break
    # ===== if
    file.write(string + '\n')
# ===== while
file.close()
print('data3.txt 파일에 쓰기 완료')
```

with 구문을 사용해서 파일을 open 하면 with 블록의 모든 내용이 실행되고 난 후 파일을 자동으로 닫아준다.

```python
with open('./data/data4.txt', 'w') as file: 
    while True:
        string = input('>>> ')
        if string.lower() == 'quit':
            break
        # ===== if
        file.write(string + '\n')
    # ===== while
    # file.close() # with 구문을 사용해서 파일을 open 하면 close() 메소드를 사용하지 않는다.
print('data4.txt 파일에 쓰기 완료')
```

```python
with open('./data/data4.txt', 'a') as file: 
    while True:
        string = input('>>> ')
        if string.lower() == 'quit':
            break
        # ===== if
        file.write(string + '\n')
    # ===== while
print('data4.txt 파일에 쓰기 완료')
```
