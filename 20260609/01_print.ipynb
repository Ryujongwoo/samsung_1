'%'와 서식 문자를 사용해 출력하기

서식 문자  
d(정수), f(실수), s(문자열)를 사용해서 출력 서식을 지정할 수 있고 출력 서식 문자를 제외한 나머지 문자들은 그대로 출력된다.

출력 서식의 형식  
%[-][0][n][.m]서식문자 => [] 안의 내용은 필요에 따라 생략이 가능하다.  
n: 출력할 전체 자릿수, 전체 자리수가 지정되면 데이터가 오른쪽으로 정렬되서 출력된다.  
0: 출력할 전체 자릿수가 지정된 경우 남는 자리에 '0'이 채워져서 출력된다. => 문자열 데이터에서는 무시된다.  
-: 출력할 전체 자릿수가 지정된 경우 데이터를 왼쪽으로 정렬해서 출력한다.  
.m: 실수를 출력할 경우 출력할 소수점 아래 자리수를 지정한다.

```python
print(100)
print(1)
print(10000)
```

print('출력 서식' % 출력할 데이터) => 출력할 데이터가 1개일 경우

```python
print('%5d' % 100)
print('%5d' % 1)
print('%5d' % 10000)
print('%05d' % 1)
print('%-5d' % 1)
```

```python
print('%f' % 123.45)
print('%6.2f' % 123.45)
print('%6.2f' % 123.456) # 소수점 아래 숫자는 잘리는 자리에서 반올림해 출력한다. => 실제 데이터는 변하지 않는다.
print('%6.2f' % 123.4)   # 남는 자리는 0을 채워 출력한다.
print('%5.0f' % 59.5)
```

```python
print('abc')
print('a')
print('abcde')
```

print('출력 서식' % (출력할 데이터1, 출력할 데이터2, ...)) => 출력할 데이터가 2개 이상일 경우  
서식을 지정해서 출력할 데이터가 2건 이상이라면 출력할 데이터들을 반드시 ()로 묶어준다.

```python
print('%d + %d = %d' % (5, 3, 5 + 3)) # 덧셈
print('%d - %d = %d' % (5, 3, 5 - 3)) # 뺄셈
print('%d * %d = %d' % (5, 3, 5 * 3)) # 곱셈
# 파이썬은 '/' 연산자로 나눗셈을 하면 소수점 아래까지 계산된다.
print(5 / 3) # 실수 나눗셈
print('%d / %d = %d' % (5, 3, 5 / 3)) # 연산 결과는 실수지만 출력 서식을 %d(정수)로 지정되서 정수만 출력된다.
print('%d / %d = %f' % (5, 3, 5 / 3))
# 파이썬은 '//' 연산자로 나눗셈을 하면 결과는 정수가 출력된다. => 몫만 계산된다.
print('%d // %d = %d' % (5, 3, 5 // 3))
print('%d // %d = %f' % (5, 3, 5 // 3))
print(5 % 3) # 나머지
# 출력 서식을 지정할 때 '%' 문자 자체를 출력하려면 2개를 연속해서 써야한다.
print('%d %% %d = %d' % (5, 3, 5 % 3))
print('%d의 %d승 = %d' % (2, 10, 2 ** 10)) # 거듭제곱
print('루트 %d = %f' % (2, 2 ** 0.5)) # 제곱근
```

산술 연산자 우선순위: () => ** => *, /, //, % => +, -

{}와 format() 메소드를 사용한 출력 서식 지정

'%'로 서식을 지정할때는 %d, %f, %s 위치에 데이터가 대입되어 출력되었지만 {}를 사용하면 {}에 데이터가 대입되서 출력된다.

```python
print('{} + {} = {}'.format(5, 3, 5 + 3))
print('{} - {} = {}'.format(5, 3, 5 - 3))
print('{} * {} = {}'.format(5, 3, 5 * 3))
print('{} / {} = {}'.format(5, 3, 5 / 3))
print('{} // {} = {}'.format(5, 3, 5 // 3))
print('{} % {} = {}'.format(5, 3, 5 % 3))
print('{}의 {}승 = {}'.format(2, 10, 2 ** 10))
print('루트 {} = {}'.format(2, 2 ** 0.5))
```

format() 메소드와 출력할 데이터의 인덱스를 사용해 출력하기 => 인덱스는 0부터 시작된다.

```python
print('{} {}'.format('홍길동', '바보'))
# 인덱스를 지정하지 않은 상태에서 아래와 같이 {}가 출력할 데이터보다 많으면 에러가 발생된다.
# print('{} {} {}'.format('홍길동', '바보'))
# {}가 모자라면 에러는 발생되지 않고 출력되며 남는 데이터는 무시된다.
print('{}'.format('홍길동', '바보'))
```

```python
print('{0} {1}'.format('홍길동', '바보'))
print('{1} {0}'.format('홍길동', '바보'))
# 인덱스를 지정하지 않은 상태에서 아래와 같이 {}가 출력할 데이터보다 많으면 에러가 발생되었지만 {}에 출력할 데이터의 인덱스를 지정하면 {}와
# 출력할 데이터의 개수가 다르더라도 인덱스 범위만 벗어나지 않으면 정상적으로 출력된다.
print('{0} {1} {0}'.format('홍길동', '바보'))
# print('{0} {1} {2}'.format('홍길동', '바보')) # 에러
# {}에 인덱스를 부분적으로 지정하면 에러가 발생된다.
# print('{0} {}'.format('홍길동', '바보')) # 에러
```

{인덱스:출력서식} => 인덱스는 생략할 수 있지만 ':'은 생략하면 안되고 출력 서식에 '%'는 사용하지 않는다.

```python
print('%-10s는(은) %7d원 입니다.' % ('메론', 15000))
print('{}는(은) {}원 입니다.'.format('메론', 15000))
print('{0}는(은) {1}원 입니다.'.format('메론', 15000))
print('{0:10s}는(은) {1:7d}원 입니다.'.format('메론', 15000))
print('{:10s}는(은) {:7d}원 입니다.'.format('메론', 15000))
print('{:10s}는(은) {:7,d}원 입니다.'.format('메론', 15000))
print('{:10s}는(은) {:.2f}원 입니다.'.format('메론', 15000))
print('{:10s}는(은) {:,.2%}원 입니다.'.format('메론', 15000))
```

'<', '^', '>'를 사용해서 문자열 정렬 방식을 지정할 수 있다 => '<'가 기본값으로 사용된다.

```python
print('{:10s}는(은) {:7d}원 입니다.'.format('메론', 15000))
print('{:<10s}는(은) {:7d}원 입니다.'.format('메론', 15000)) # 왼쪽 정렬
print('{:>10s}는(은) {:7d}원 입니다.'.format('메론', 15000)) # 오른쪽 정렬
print('{:^10s}는(은) {:7d}원 입니다.'.format('메론', 15000)) # 가운데 정렬
```

출력시 남는 자리의 공백을 채울 채움 문자를 '<', '^', '>' 앞에 지정할 수 있다.

```python
# print('{:_10s}는(은) {:7d}원 입니다.'.format('메론', 15000))
print('{:_<10s}는(은) {:7d}원 입니다.'.format('메론', 15000))
print('{:*>10s}는(은) {:7d}원 입니다.'.format('메론', 15000))
print('{:#^10s}는(은) {:7d}원 입니다.'.format('메론', 15000))
```
