len() 함수는 인수로 지정된 문자열을 구성하는 문자의 개수를 얻어온다.

```python
string = 'We are the champions, My friends!'
print(len(string))
print(string[-1])
print(string[len(string) - 1])
```

count() 메소드는 문자열에서 인수로 지정한 문자열의 출현 횟수를 얻어온다. => 특정 문자열이 포함되었나 알아낼 수 있다.  
count() 메소드를 실행한 결과가 0이면 불포함, 1이면 포함, 2이상이면 중복해서 포함

```python
string = 'We are the champions, My friends!'
print(string.count('e'))
print(string.count('We'))
print(string.count('we'))
```

index() 메소드는 문자열에서 인수로 지정한 문자열이 최초로 출현하는 인덱스를 얻어온다. => 특정 문자열이 포함되었나 알아낼 수 있다.  
index() 메소드를 실행한 결과가 0이상이면 포함, 에러가 발생되면 불포함

```python
string = 'We are the champions, My friends!'
print(string.index('e'))
print(string.index('We'))
print(string.index('we'))
```

find() 메소드는 문자열에서 인수로 지정한 문자열이 최초로 출현하는 인덱스를 얻어온다. => 특정 문자열이 포함되었나 알아낼 수 있다.  
find() 메소드를 실행한 결과가 0이상이면 포함, -1이면 불포함

```python
string = 'We are the champions, My friends!'
print(string.find('e'))
print(string.find('We'))
print(string.find('we')) # 인수로 지정한 문자열이 없으면 -1을 리턴한다.
```

join() 메소드는 인수로 지정된 문자열 사이사이에 특정 문자열을 삽입한다.

```python
string = 'samsungAI'
print('/'.join(string))
print(' '.join(string))
print('\n'.join(string))
print('!(^^)~'.join(string))
```

upper() 메소드는 문자열을 무조건 대문자로 변환한다.  
lower() 메소드는 문자열을 무조건 소문자로 변환한다.  
title() 메소드는 문자열을 구성하는 각 단어의 첫 문자만 대문자로 변환한다.

```python
string = 'samsungAI good'
print(string.upper())
print(string.lower())
print(string.title())
```

lstrip() 메소드는 문자열 앞(왼쪽)의 불필요한 공백을 제거한다.  
rstrip() 메소드는 문자열 뒤(오른쪽)의 불필요한 공백을 제거한다.  
strip() 메소드는 문자열 앞뒤의 불필요한 공백을 제거한다.

```python
string = '    samsungAI    '
print(string)
print(string.lstrip())
print(string.rstrip())
print(string.strip())
```

replace() 메소드는 특정 문자열을 지정한 문자열로 일괄 치환한다.

```python
string = 'We are the champions, My friends!'
print(string)
print(string.replace('e', '2'))
```

문자열 메소드 응용

strip() 메소드가 꼭 공백만 제거하는 것은 아니다.  
strip() 메소드의 인수로 아무것도 지정하지 않으면 공백이 기본값으로 사용되서 공백이 제거되지만 인수를 지정하면 인수로 지정한 문자가 문자열 양쪽에 나타나지 않을때까지 제거한다.  
인수로 지정하는 문자열의 순서와 제거하는 순서는 아무런 상관이 없다.

```python
print('www.samsungAI.com')
print('    www.samsungAI.com    '.strip()) # 공백 제거
print('    www.samsungAI.com    '.strip(' ')) # 인수로 지정한 문자 제거
print('www.samsungAI.com'.strip('w.com'))
print('www.samsungAI.com'.strip('cm.wo'))
```

문자열 중간의 특정 문자열을 제거하려면 replace() 메소드를 사용하면 된다.

```python
print('www.samsungAI.com'.replace('s', ''))
```

주민등록번호를 입력받아 성별 뒤의 모든 문자가 '*'로 출력되게 한다.

830422-1185600 => 830422-1******

```python
jumin = input('주민등록번호: ')
print(jumin)
print(jumin[0:8])
print(jumin[:8])
print(jumin[:8] + '******')
print(jumin[:8] + '*' * 6)
```

```python
# '-'과 '-' 사이의 문자열을 '*'로 아래의 예와 같이 치환하기 이때 '-'과 '-' 사이의 문자열 개수는 유동적이다.
# find() 메소드는 문자열의 앞에서부터 찾지만 rfind()는 문자열의 뒤에서부터 찾는다.
# 123456-123-12345 => 123456-***-12345
# 123456-1234567-12345 => 123456-*******-12345
# 123456-12-12345 => 123456-**-12345
code = input('코드: ')
print(code, len(code), code.find('-'), code.rfind('-'))
start = code[:code.find('-') + 1]
print(start)
end = code[code.rfind('-'):]
print(end)
print(len(code) - len(start) - len(end))
code = start + '*' * (len(code) - len(start) - len(end)) + end
print(code)
```

```python
code = input('코드: ')
print(code)
code = code.split('-')
print(code)
print(code[0] + '-' + '*' * len(code[1]) + '-' + code[2])
```

```python
code[1] = '*' * len(code[1])
print(code)
print('-'.join(code))
```
