if 명령의 형식

if 조건식:  
&nbsp;&nbsp;&nbsp;&nbsp;조건식이 참일 경우 실행할 문장  
&nbsp;&nbsp;&nbsp;&nbsp;...  
else:  
&nbsp;&nbsp;&nbsp;&nbsp;조건이 거짓일 경우 실행할 문장  
&nbsp;&nbsp;&nbsp;&nbsp;...  
    
조건식을 만족하지 않을때 아무런 일도 할 필요가 없다면 else 이하는 생략한다.

관계 연산자

&nbsp;<&nbsp;&nbsp;: 작다, 미만  
&nbsp;<= : 작거나 같다, 이하  
&nbsp;>&nbsp;&nbsp;: 크다, 초과  
&nbsp;>= : 크거나 같다, 이상  
&nbsp;== : 같다  
&nbsp;!= : 같지 않다

논리 연산자

and : 논리곱, 두 조건이 모두 참일 경우에만 참, ~이고, ~이면서, ~중에서  
or&nbsp;&nbsp;: 논리합, 두 조건 중에서 한 개 이상 참일 경우에 참, ~또는, ~이거나  
not : 논리 부정
