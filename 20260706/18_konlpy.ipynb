형태소 분석(morphological analysis)

형태소 분석은 자연 언어의 문장을 형태소라는 최소 단위로 분할하고 품사를 판별하는 작업이다.  
영문 형태소 분석은 형태소마다 띄어쓰기를 해서 문장을 구성하는 것이 기본이기 때문에 크게 어렵지 않으나 아시아 계열 언어 분석에는 많은 노력이 필요하다.  
대표적으로 문법 규칙에 의한 확률적 언어 모델을 사용하는 방법이 있는데 최근 확률적 언어 모델을 사용한 형태소 분석 많아서 정밀도가 높아졌다.  
어쨋거나 두 가지 품사 사전과 문법 사전을 기반으로 대조하면서 형태소 분석을 한다.

한국어 형태소 분석 라이브러리(konlpy)를 사용하기 위해 라이브러리를 설치하고 import 한다. => java 1.7 이상을 반드시 설치하고 JAVA_HOME 환경변수를 설정한다.  
konlpy는 대한민국 헌법 말뭉치(corpus) kolaw와 국회 법안 말뭉치 kobill을 제공한다.

konlpy 라이브러리가 설채되지 않을 경우 아래의 방법을 사용해서 설치한다.  
pip install jpype<1를 실행해서 다운 그레이드 후 설치한다.

```python
# !pip install konlpy
# konlpy가 제공하는 한글 말뭉치를 불러온다.
from konlpy.corpus import kolaw, kobill
```

```python
# 각 말뭉치가 저장된 텍스트 파일의 이름은 fileids() 메소드로 확인할 수 있다.
print(kolaw.fileids())
print(kobill.fileids())
```

```python
# 말뭉치가 저장된 텍스트 파일을 읽어들이려면 말뭉치가 저장된 텍스트 파일을 open() 메소드로 열고 read() 메소드로 읽는다.
kolaw_corpus = kolaw.open('constitution.txt').read()
print(kolaw_corpus[:40])
```

```python
kobill_corpus = kobill.open('1809890.txt').read()
print(kobill_corpus[:40])
```

konlpy에는 다음과 같은 형태소 분석, 태깅 라이브러리를 파이썬에서 쉽게 사용할 수 있도록 모아놓았다.

Hannanum: 한나눔, KAIST Semantic Web Research Center 개발.  
Kkma: 꼬꼬마, 서울대학교 IDS(Intelligent Data System) 연구실 개발.  
Komoran: 코모란, Shineware에서 개발.  
Mecab: 메카브, 일본어용 형태소 분석기를 한국어를 사용할 수 있도록 수정.  
Open Korean Text: 오픈 소스 한국어 분석기. 과거에 트위터 형태소 분석기.

```python
text = '이것도되나욬ㅋㅋㅋㅋ'
```

```python
print(hannanum.nouns(text))
print(kkma.nouns(text))
print(komoran.nouns(text))
print(okt.nouns(text))
```

```python
print(hannanum.morphs(text))
print(kkma.morphs(text))
print(komoran.morphs(text))
print(okt.morphs(text))
```

```python
print(hannanum.pos(text))
print(kkma.pos(text))
print(komoran.pos(text))
print(okt.pos(text))
```

```python
print(okt.pos(text))
print(okt.pos(text, norm=True)) # 품사 태깅
```

```python
print(okt.pos(text))
print(okt.pos(text, stem=True)) # 원형 글자로 바꿔준다.
```
