앤스콤 4분할 그래프

데이터를 시작화하지 않고 수치만 확인했을때 발생될 수 있는 함정을 보여주는 그래프이다.  
앤스콤 데이터셋은 4개의 그룹으로 구성되어 있으며 4개의 데이터 그룹은 각각 평균, 분산과 같은 수치나 상관관계 회귀선이 모두 같다는 특징이 있다.  
그래서 이런 결과를 보면 4개의 데이터셋 데이터는 모두 같을 것이라고 착각할 수 있다.

앤스콤 데이터셋은 seaborn 라이브러리에 포함되어 있다.  
seaborn 라이브러리의 load_dataset() 메소드의 인수로 'anscombe'을 전달하면 앤스콤 데이터셋을 불러올 수 있다.

```python
anscombe_df = sns.load_dataset('anscombe')
anscombe_df
```

```python
subset1 = anscombe_df[anscombe_df.dataset == 'I'].drop('dataset', axis=1)
subset1.mean()
```

```python
subset2 = anscombe_df[anscombe_df.dataset == 'II'].drop('dataset', axis=1)
subset2.mean()
```

```python
subset3 = anscombe_df[anscombe_df.dataset == 'III'].drop('dataset', axis=1)
subset3.mean()
```

```python
subset4 = anscombe_df[anscombe_df.dataset == 'IV'].drop('dataset', axis=1)
subset4.mean()
```

```python
anscombe_df.groupby('dataset').mean()
```

```python
plt.figure(figsize=[6, 4])
plt.plot(subset1.x, subset1.y, 'ro')
plt.show()
```

```python
plt.figure(figsize=[6, 4])
plt.plot(subset2.x, subset2.y, 'go')
plt.show()
```

```python
plt.figure(figsize=[6, 4])
plt.plot(subset3.x, subset3.y, 'bo')
plt.show()
```

```python
plt.figure(figsize=[6, 4])
plt.plot(subset4.x, subset4.y, 'mo')
plt.show()
```

subplot

```python
# figure() 메소드로 그래프 전체가 그려질 기본 틀을 만든다.
fig = plt.figure(figsize=[20, 4])
# add_subplot() 메소드로 그래프의 격자를 만든다. add_subplot(격자의 행 개수, 격자의 열 개수, 그래프 출력 위치)
axes1 = fig.add_subplot(1, 4, 1)
axes2 = fig.add_subplot(1, 4, 2)
axes3 = fig.add_subplot(1, 4, 3)
axes4 = fig.add_subplot(1, 4, 4)
# plot() 메소드로 각 격자에 데이터를 전달해서 그래프를 그린다.
axes1.plot(subset1.x, subset1.y, 'ro')
axes2.plot(subset2.x, subset2.y, 'go')
axes3.plot(subset3.x, subset3.y, 'bo')
axes4.plot(subset4.x, subset4.y, 'mo')
# set_title() 메소드로 각각의 그래프에 제목을 추가한다.
axes1.set_title('데이터셋 I')
axes2.set_title('데이터셋 II')
axes3.set_title('데이터셋 III')
axes4.set_title('데이터셋 IV')
# suptitle() 메소드로 그래프 전체 제목을 추가한다.
fig.suptitle('앤스콤 4분할 그래프')
# 그래프의 제목과 숫자가 겹쳐보이면 tigth_layout() 메소드를 실행하면 그래프의 레이아웃이 겹치지 않도록 조정된다.
fig.tight_layout()
plt.show()
```
