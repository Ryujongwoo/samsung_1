```python
import warnings
warnings.filterwarnings(action='ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib as mpl
import matplotlib.font_manager as fm
mpl.rcParams['axes.unicode_minus'] = False
```

```python
data = np.random.randint(-100, 100, 50).cumsum()
print(data)
```

② matplotlib 라이브러리의 rcParams[]로 전역 글꼴을 설정한다. => 그래프에 폰트 설정이 필요한 파일마다 설정한다.

matplotlib 라이브러리의 폰트 기본 설정 정보 읽기

```python
print('기본으로 설정된 폰트 이름: {}'.format(plt.rcParams['font.family']))
print('기본으로 설정된 폰트 크기: {}'.format(plt.rcParams['font.size']))
```

```python
plt.rcParams['font.family'] = 'NanumGothicCoding'
plt.rcParams['font.size'] = 15
print('기본으로 설정된 폰트 이름: {}'.format(plt.rcParams['font.family']))
print('기본으로 설정된 폰트 크기: {}'.format(plt.rcParams['font.size']))
```

```python
plt.figure(figsize=[8, 6])
plt.plot(range(1, 51), data, 'ro--')
plt.title('시간별 가격 추이')
plt.xlabel('시간')
plt.ylabel('가격')
plt.show()
```
