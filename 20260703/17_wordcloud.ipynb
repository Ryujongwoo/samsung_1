wordcloud를 사용하기 위해서 라이브러리를 설치하고 import 한다.

```python
# !pip install wordcloud
from wordcloud import WordCloud
```

```python
text = open('./data/wordData.txt').read()
print(type(text))
print(text)
```

pip install wordcloud를 실행해서 wordcloud를 설치하면 1.9.x 버전이 설치된다.  
1.9.x 버전에서 generate() 메소드를 실행했을 때 ValueError가 발생되며 정상적으로 동작하지 않으면 wordcloud를 1.8.0버전으로 pip install wordcloud==1.8.0를 실행해서 다운그레이드해서 설치해야 한다.

```python
# generate() 메소드로 단어별 출현 빈도수를 비율로 계산한다.
wordcloud = WordCloud().generate(text)
# 단어별 출현 빈도수 비율은 words_ 속성으로 확인할 수 있다.
wordcloud.words_
```

```python
plt.figure(figsize=[12, 8])
# imshow() 메소드로 단어별 출현 빈도수를 이미지로 출력한다.
plt.imshow(wordcloud, interpolation='bilinear')
# 워드클라우드 이미지에 나타나는 x, y축의 눈금을 표시하지 않는다.
plt.axis('off')
plt.show()
```

WordCloud()의 인수로 max_font_size 옵션을 지정하면 최고로 출현 빈도가 높은 단어의 크기를 지정할 수 있다.  
나머지 단어는 최고로 출현 빈도가 높은 단어의 크기를 지정하면 단어 출현 빈도수 비율에 맞춰서 나머지 단어도 크기가 변경된다.

```python
wordcloud = WordCloud(max_font_size=50).generate(text)
plt.figure(figsize=[12, 8])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

WordCloud()의 인수로 max_words 옵션을 지정하면 워드클라우드에 표시될 단어의 개수를 지정한다.

```python
wordcloud = WordCloud(max_font_size=50, max_words=50).generate(text)
plt.figure(figsize=[12, 8])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

불용어(stopword) 처리

불용어란 자주 등장하지만 의미가 없는 즉, 워드클라우드로 작성할 단어 목록에서 제외시킬 단어를 말한다.

```python
# 불용어 처리를 하기 위해 불용어 데이터셋을 import 한다.
from wordcloud import STOPWORDS
```

```python
# 불용어 데이터셋을 불러온다.
stopwords = set(STOPWORDS)
print(type(stopwords))
print(len(stopwords))
stopwords
```

```python
# 불용어 데이터셋에 불용어를 추가한다.
stopwords.add('Alice')
stopwords
```

```python
# 불용어 데이터셋에서 불용어를 제거한다.
stopwords.remove('Alice')
stopwords
```

```python
# 워드클라우드에 표시할 최대 단어는 200개, 배경색은 흰색, 불용어 적용, 마스크 이미지는 alice_mask(alice_mask.png)를 사용한다.
wordcloud = WordCloud(
    max_words=200,
    background_color='white',
    stopwords=stopwords,
    mask=alice_mask
).generate(text)

plt.figure(figsize=[12, 8])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
