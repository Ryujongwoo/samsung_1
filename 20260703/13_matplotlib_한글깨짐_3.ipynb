```python
import warnings
warnings.filterwarnings(action='ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib as mpl
import matplotlib.font_manager as fm
mpl.rcParams['axes.unicode_minus'] = False
```

```python
data = np.random.randint(-100, 100, 50).cumsum()
print(data)
```

③ 폰트 설정 정보를 matplotlibrc 설정 파일에 직접 넣어준다. => 그래프에 폰트 설정이 필요한 컴퓨터마다 설정한다.  
mpl.matplotlib_fname()를 실행하면 출력되는 경로의 matplotlibrc 파일을 수정한다.  
matplotlibrc 파일을 수정하면 jupyter nbclassic을 실행할 때 마다 별도의 설정없이 matplotlibrc 파일의 설정을 확인한다.  

matplotlibrc 파일에서 아래의 항목을 찾아서 주석(#)을 제거하고 수정한다.  
font.family:  sans-serif => NanumGothicCoding로 수정한다.  
font.size:    10.0 => 15.0으로 수정한다.

```python
plt.figure(figsize=[8, 6])
plt.plot(range(1, 51), data, 'ro--')
plt.title('시간별 가격 추이')
plt.xlabel('시간')
plt.ylabel('가격')
plt.show()
```
