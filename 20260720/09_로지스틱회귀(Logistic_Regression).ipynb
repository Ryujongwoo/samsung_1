로지스틱 회귀

공부시간, 과외시간과 시험점수 사이의 관계는 좌표로 나타냈을 때 형태가 직선으로 해결되는 선형회귀를 사용하기에 적합했었다.  
공부시간, 과외시간에 따른 시험점수가 합격여부로 발표되는 시험이 있을 경우 직선으로 해결하기에는 적합하지 못한 문제가 발생된다. 이럴때 사용하는 로지스틱 회귀는 참과 거짓 중에 하나를 내놓는 과정으로 참과 거짓을 구분한 'S'자를 눕혀놓은 형태의 선을 그어주는 작업이다.

```python
# [공부시간, 합격여부]
data = [[2, 0], [4, 0], [6, 0], [8, 1], [10, 1], [12, 1], [14, 1]]
x = [d[0] for d in data] # 공부시간
y = [d[1] for d in data] # 합격여부, 실제값
print(f'공부시간: {x}, 합격여부: {y}')
```

```python
a = tf.Variable(tf.random_uniform([1], dtype=tf.float64)) # 공부시간 가중치
b = tf.Variable(tf.random_uniform([1], dtype=tf.float64)) # 바이어스
sess = tf.Session()
sess.run(tf.global_variables_initializer())
print(f'a = {sess.run(a)}, b = {sess.run(b)}')
```

시그모이드 방정식

```python
# np.e: numpy에서 제공하는 지수값(2.718281828459045)을 의미하는 상수이다.
# print(np.e)
Y = 1 / (1 + np.e ** -(a * x + b))
```

시그모이드 방정식의 오차를 계산하는 수식을 만든다. => 시그모이드 함수의 특성은 예측값(Y)이 항상 0아니면 1이다.

```python
# -tf.reduce_mean(np.array(실제값) * tf.log(예측값) + (1 - np.array(실제값)) * tf.log(1 - 예측값))
loss = -tf.reduce_mean(np.array(y) * tf.log(Y) + (1 - np.array(y)) * tf.log(1 - Y))
```

경사 하강법 알고리즘을 이용해서 오차를 최소로하는 값을 찾는다.

```python
gradient_descent = tf.train.GradientDescentOptimizer(0.1).minimize(loss)
```

학습 시킨다.

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for epoch in range(300001):
    '''
    sess.run(gradient_descent)
    if epoch % 5000 == 0:
        print('epoch: {:6d}, loss: {:7.4f}, a: {:7.4f}, b: {:7.4f}'.format(epoch, sess.run(loss), sess.run(a)[0], sess.run(b)[0]))
    '''
    # run() 메소드로 여러개를 한 번에 실행하려면 run() 메소드로 실행할 목록을 리스트로 묶어서 전달한다.
    _, _loss, _a, _b = sess.run([gradient_descent, loss, a, b])
    if epoch % 5000 == 0:
        print('epoch: {}, loss: {:7.4f}, a: {:7.4f}, b: {:7.4f}'.format(epoch, _loss, _a[0], _b[0]))
```
