```python
import warnings
warnings.filterwarnings("ignore")
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
# [공부시간, 과외시간, 시험점수]
data = [[2, 0, 81], [4, 4, 93], [6, 2, 91], [8, 3, 97]]
x1 = [d[0] for d in data] # 공부시간
x2 = [d[1] for d in data] # 과외시간
y = [d[2] for d in data] # 시험점수, 실제값
```

```python
a1 = tf.Variable(tf.random_uniform([1], 0, 10)) # 공부시간 가중치
a2 = tf.Variable(tf.random_uniform([1], 0, 10)) # 과외시간 가중치
b = tf.Variable(tf.random_uniform([1], 0, 100)) # 바이어스
sess = tf.Session()
sess.run(tf.global_variables_initializer())
print(f'a1 = {sess.run(a1)}, a2 = {sess.run(a2)}, b = {sess.run(b)}')
```

```python
Y = a1 * x1 + a2 * x2 + b
rmse = tf.sqrt(tf.reduce_mean(tf.square(Y - y)))
gradient_descent = tf.train.GradientDescentOptimizer(0.1).minimize(rmse)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for epoch in range(3001):
    sess.run(gradient_descent)
    if epoch % 200 == 0:
        print('epoch: {:4d}, rmse: {:7.4f}, a1: {:7.4f}, a2: {:7.4f}, b: {:7.4f}'.format(
            epoch, sess.run(rmse), sess.run(a1)[0], sess.run(a2)[0], sess.run(b)[0]
        ))
```
