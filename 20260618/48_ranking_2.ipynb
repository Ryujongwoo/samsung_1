선택 정렬 알고리즘의 반복 알고리즘을 이용해서 석차를 계산한다.

<img src="ranking.png" width="1100" align="left" />

```python
score = [70, 100, 80, 100, 90]
# rank = [1] * len(score)
rank = [1 for _ in range(len(score))]
# print(rank)

for i in range(len(score) - 1): # 회전수, 선택대상
    for j in range(i + 1, len(score)): # 비교대상
        # i번째 점수가 크면 j번째 석차를 증가시키고 j번째 점수가 크면 i번째 석차를 증가시킨다.
        # 부등호를 반대로 수정하면 오름차순 석차가 계산된다.
        if score[i] > score[j]:
            rank[j] += 1
        elif score[i] < score[j]:
            rank[i] += 1
        # ===== if
    # ===== for j
# ===== for i

for s, r in zip(score, rank):
    print('{:3d}점은 {}등 입니다.'.format(s, r), end=' ')
    # 10점당 '*'을 1개씩 출력한다.
    '''
    for i in range(s // 10):
        print('*', end='')
    print()
    '''
    print('*' * (s // 10))
```
