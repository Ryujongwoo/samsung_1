insertion sort(삽입 정렬, 삽입법)  

삽입 정렬은 이미 정렬된 데이터에 새로운(삽입할) 데이터를 위치에 맞게 삽입한다.  
정렬할 데이터가 n개일 경우 회전수는 n-1번이 된다.

<img src="insertionSort.png" width="1200" align="left" />

```python
n = 5
for i in range(1, n):
    for j in range(i - 1, -1, -1):
        print(f'[i = {i}, j = {j}]', end=' ')
    print()
```

```python
data = [8, 3, 4, 9, 1]
for i in range(1, len(data)): # 삽입할 데이터 위치, 회전수
    # 이미 정렬된 데이터에 새 데이터가 삽입될 위치를 찾기위해서 새 데이터와 새 데이터의 인덱스를 각각 key, index에 저장한다.
    key = data[i]
    index = i
    # 새 데이터가 삽입될 위치를 찾는다.
    for j in range(i - 1, -1, -1): # 삽입할 데이터 위치를 찾기위해 비교하는 데이터의 위치
        if data[j] > key:
            # 이미 정렬된 데이터가 크기때문에 정렬된 데이터를 인덱스가 1큰 위치로 이동시킨다.
            data[j + 1] = data[j]
            # index에 저장된 삽입할 데이터의 인덱스 값을 1감소시킨다.
            index -= 1
        else:
            # 삽입할 데이터가 이미 정렬된 데이터보다 크거나 같다면 앞에 위치한 데이터들은 더 이상 비교할 필요가 없으므로 반복을 탈출한다.
            break
        # ===== if
    # ===== for j => 회전 종료
    # key에 저장된 삽입할 데이터를 삽입할 인덱스 위치에 맞게 삽입한다.
    data[index] = key
    print(f'{i}회전 결과: {data}')
# ===== for i
print(f'최종 정렬 결과: {data}')
```
