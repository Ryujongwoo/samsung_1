bubble sort(교환법)

선택 정렬 => 특정 위치(i번째)의 데이터를 선택해서 나머지 데이터(j번째)와 비교하며 정렬한다.  
버블 정렬 => 인접한 데이터(j번째, j+1번째)를 비교하며 정렬한다.  
버블 정렬은 선택 정렬과 같이 정렬할 데이터가 n개일 경우 회전수는 n-1번이 된다.

```python
n = 5
for i in range(n-1):
    for j in range(n-1-i):
        print(f'[i = {i}, j = {j}]', end=' ')
    print()
```

```python
# data = [8, 3, 4, 9, 1]
data = [9, 1, 3, 4, 8]
for i in range(len(data) - 1):
    for j in range(len(data) - 1 - i):
        # 오름차순 정렬 => 앞(j번째)의 데이터가 뒤(j+1번째)의 데이터보다 크면 데이터를 교환한다.
        # 내림차순 정렬 => 앞(j번째)의 데이터가 뒤(j+1번째)의 데이터보다 작으면 데이터를 교환한다.
        # 부등호를 '>'에서 '<'로 변경하면 내림차순으로 정렬된다.
        if data[j] > data[j + 1]:
            data[j], data[j + 1] = data[j + 1], data[j]
        # ===== if
    # ===== for j => 회전 종료
    print(f'{i + 1}회전 결과: {data}')
# ===== for i => 정렬 종료
print(f'최종 정렬 결과: {data}')
```
