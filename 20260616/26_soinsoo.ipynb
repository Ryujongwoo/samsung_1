```python
# 소인수 분해가 불가능한 숫자가 입력될 때 까지 실행해야 하기때문에 무한루프를 돌려서 처리한다.
while True:
    n = int(input('소인수 분해할 숫자를 입력하세요: '))
    if n < 2:
        print('소인수 분해가 불가능한 숫자가 입력되서 프로그램을 종료합니다.')
        break
    # ===== if
    # 2이상인 숫자가 입력되었으므로 소인수 분해를 시작한다.
    s = [] # 분해된 소인수를 기억할 빈 리스트를 선언한다.
    c = 0 # 소인수의 개수를 기억할 변수를 선언하고 초기화시킨다.
    number = n # 소인수 분해 결과를 출력하기위해서 입력받은 숫자를 보관한다.
    
    # 소인수 분해를 시작한다.
    # 소인수 분해를 하기위해서 입력한 숫자(n)의 크기에 따라서 반복횟가 달라지므로 무한루프를 돌려서 처리한다.
    while True:
        k = 2 # 소인수 분해를 시작할 수를 초기화시킨다. 1은 모든 수를 나눠서 떨어뜨릴 수 있으므로 2부터 시작한다.
        while True:
            r = n % k
            if r == 0: # 소인수 분해가 되었는가?
                break
            # ===== if
            k += 1 # 소인수 분해가 안되면 k값을 1증가시킨다.
        # ===== while
        c += 1 # 소인수 분해가 되었으면 소인수의 개수를 1증가시킨다. => 소인수의 개수가 1개면 n은 소수이다.
        s.append(k) #  발생된 소인수를 리스트에 저장한다.
        # 다음 소인수를 계산하기위해 소인수 분해하고 남은 숫자(몫)을 다시 n에 저장한다.
        n //= k
        if n == 1: # 소인수 분해가 끝났는가?
            break
        # ===== if
    # ===== while
    
    # 소인수 분해를 하기위해서 입력한 숫자가 소수인가 비교한다.
    if c == 1: # 소수인가?
        print(f'입력: {number} => 출력: 소수')
    else:
        print(f'입력: {number} => 출력: ', end='')
        for y in range(c - 1):
            print(s[y], end=' * ')
        # ===== for
        print(s[-1])
    # ===== if
# ===== while
print(f'입력: {n} => 출력: 프로그램 종료')
```
