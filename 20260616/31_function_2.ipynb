전달받을 인수가 없는 함수 => 처리할 데이터가 없는 함수

```python
def func1():
    print('func1() 함수가 실행됩니다. - 1')
    return 'func1() 함수가 실행됩니다. - 2'
```

```python
print(func1())
```

결과(return) 값이 없는 함수

```python
def func2(a, b):
    print(f'{a} + {b} = {a + b}')

# 파이썬은 함수의 오버로딩을 지원하지 않는다. => 오버로딩 효과를 내려면 가변 인자를 사용하는 함수를 만들면 된다.
# 오버로딩 => 같은 이름으로 여러개의 함수를 만들어 사용하는 것을 말한다.
# 함수 오버로딩은 이름이 같은 함수가 여러개 있을 경우 인수의 개수가 다르면 서로 다른 함수로 취급하고 만약에 인수의 개수도 같다면 인수의 데이터 타입으로 함수를
# 식별한다.
# 파이썬은 함수의 오버로딩이 지원되지 않기때문에 같은 이름으로 여러개의 함수를 정의하면 맨 마지막에 정의 함수만 유효한 함수가 되고 이전에 정의한 함수는 뒤에서
# 정의한 함수로 덮어쓰기 된다.
    
def func2(a, b, c):
    print(f'{a} + {b} + {c} = {a + b + c}')
```

```python
print(func2(5, 8))
```

```python
print(func2(5, 8, 3))
```
