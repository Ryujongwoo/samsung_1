```python
# 데이터가 넘어오면 1부터 넘어온 데이터까지 합계를 계산하고 데이터가 넘어오지 않으면 100까지 합계를 계산하는 함수
def func5(n = 100):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
```

```python
print(func5(10))
```

```python
print(func5())
```
