함수 실행 순서

프로그램에서 함수가 호출되면 호출된 함수로 제어가 넘어간다. => 함수가 실행된 후 복귀할 주소는 스택에 저장된다.  
호출된 함수가 정의된 기능을 모두 실행하거나 return 명령을 만나면 함수가 종료된다. => 스택에 저장된 복귀할 주소를 참조한다. => 마지막에 호출된 함수가 먼저 종료된다.  
return이 있으면 return 뒤의 값을 가지고 함수가 호출된곳으로 돌아가고 return이 없거나 return만 있으면 None을 가지고 돌아간다.

```python
# 가장 일반적으로 사용되는 함수의 모양
# a, b 2개의 데이터를 인수로 넘겨받아 덧셈을 실행한 결과를 리턴하는 함수
def add(a, b): # 호출당하는 함수의 인수를 가인수라 부른다.
    print('add(a, b) 함수가 실행됩니다.')
    result = a + b
    return result
```

실인수와 가인수의 이름은 반드시 같을 필요는 없지만 인수의 개수는 반드시 같아야 한다.  
실인수와 가인수는 서로 독립적인 메모리 공간을 사용하기 때문에 가인수의 값이 변경되더라도 실인수의 값은 변경되지 않는다.

```python
a, b = map(int, input('숫자 2개를 입력하세요: ').split())
print(f'함수 실행 결과: {add(a, b)}') # 함수를 호출하는 부분의 인수를 실인수라고 부른다.
```
