분산, 표준편차

n개의 데이터(999가 입력되기 전 까지)를 입력받아 최대값과 최소값을 제외한 나머지 데이터의 표준편차를 계산하는 프로그램  

표준편차 공식

$$\sqrt{\frac{\sum_{i=1}^{n-2}(x_i-\bar{x})^2}{N-2}}$$

999가 입력되기 전까지 데이터 입력받기

```python
data = []
while True:
    n = int(input('n: '))
    if n == 999:
        break
    data.append(n)
print(data)
```

최대값, 최소값 판단하기 => 최대값을 기억할 변수는 아주 작은 값을, 최소값을 기억할 변수는 아주 큰 값을 초기치로 넣어주고 비교한다.

```python
# data = [35, 91, 75, 62, 84]
data = [35, 62, 75, 84, 91]
big = 0
small = 100
for score in data:
    if big < score:
        big = score
    if small > score:
        small = score
print(f'최대값: {big}, 최소값: {small}')
```

```python
data = [35, 91, 75, 62, 84]
big = 0
small = 100
for i in range(len(data)):
    if big < data[i]:
        big = data[i]
    if small > data[i]:
        small = data[i]
print(f'최대값: {big}, 최소값: {small}')
```

최대값, 최소값 판단하기 => 최대값과 최소값을 기억할 변수는 각각 리스트의 0번째 인덱스의 값을 초기치로 넣어주고 비교한다.

```python
# data = [35, 91, 75, 62, 84]
data = [35, 62, 75, 84, 91]
big = data[0]
small = data[0]
for score in data[1:]:
    if big < score:
        big = score
    elif small > score:
        small = score
print(f'최대값: {big}, 최소값: {small}')
```

```python
data = [35, 91, 75, 62, 84]
big = data[0]
small = data[0]
for i in range(1, len(data)):
    if big < data[i]:
        big = data[i]
    elif small > data[i]:
        small = data[i]
print(f'최대값: {big}, 최소값: {small}')
```

max(), min() 함수로 최대값, 최소값 판단하기

```python
data = [35, 91, 75, 62, 84]
print(f'최대값: {max(data)}, 최소값: {min(data)}')
```

데이터를 정렬시켜서 최대값, 최소값 판단하기

```python
data = [35, 91, 75, 62, 84]
data.sort() # 오름차순 정렬
print(data)
print(f'최대값: {data[-1]}, 최소값: {data[0]}')
data.sort(reverse=True) # 내림차순 정렬
print(data)
print(f'최대값: {data[0]}, 최소값: {data[-1]}')
```

표준편차를 계산할 데이터를 입력받는다.

```python
# 표준편차를 계산할 데이터를 999가 입력될 때까지 입력받는다.
score = []
# 999가 몇 번째 데이터로 입력될지 모르기때문에 무한루프를 돌린다.
while True:
    data = int(input('{}번째 표준편차를 계산할 데이터를 입력하세요: '.format(len(score) + 1)))
    # 입력받은 데이터가 999면 더 이상 데이터를 입력받을 필요없으므로 무한루프를 탈출한다.
    if data == 999:
        break
    # ===== if
    # 999가 입력되지 않았으므로 리스트에 입력받은 데이터를 추가한다.
    score.append(data)
# ===== while
print(score)
```

최대값과 최소값을 제외한 데이터의 평균을 계산한다.

```python
big = max(score)
small = min(score)
'''
total = 0
for s in score:
    total += s
# total = total - big - small
total -= big + small
'''
total = sum(score) - big - small
print(f'최대값과 최소값을 제외한 데이터의 합계: {total}')
mean = total / (len(score) - 2)
print(f'최대값과 최소값을 제외한 데이터의 평균: {mean}')
```

```python
score.sort()
total = 0
# for i in range(1, len(score) - 1):
    # total += score[i]
# for s in score[1:len(score) - 1]:
    # total += s
total = sum(score[1:-1])
print(f'최대값과 최소값을 제외한 데이터의 합계: {total}')
# mean = total / (len(score) - 2)
mean = total / len(score[1:-1])
print(f'최대값과 최소값을 제외한 데이터의 평균: {mean}')
```

최대값과 최소값을 제외한 나머지 데이터와 산술평균의 편차의 제곱의 합계를 계산한다.

```python
import math
```

```python
total = 0
for s in score[1:-1]:
    # total += (s - mean) * (s - mean)
    # total += (s - mean) ** 2
    total += math.pow(s - mean, 2)
print(f'최대값과 최소값을 제외한 데이터와 산술평균의 편차의 제곱의 합계: {total}')
```

최대값과 최소값을 제외한 데이터의 분산과 표준편차를 계산한다.

```python
var = total / len(score[1:-1])
print(f'최대값과 최소값을 제외한 데이터의 분산: {var}')
# std = var ** 0.5
std = math.sqrt(var)
print(f'최대값과 최소값을 제외한 데이터의 표준편차: {std}')
```
