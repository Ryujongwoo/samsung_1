numpy 파일 입출력

numpy는 배열 객체를 바이너리(2진) 파일 또는 텍스트 파일로 저장하고 불러오는 기능을 제공한다.

```python
a = np.random.randint(1, 11, (2, 3))
print(a)
b = np.random.randint(1, 11, (2, 3))
print(b)
```

save(), savez() 메소드로 numpy 배열 객체를 바이너리 형태의 파일로 저장할 수 있다.  
save() 메소드는 numpy 배열 1개를 저장한다. 확장명 => *.npy  
savez() 메소드는 numpy 배열 여러개를 저장한다. 확장명 => *.npz  

```python
# a라는 numpy 배열을 바이너리 형태로 저장한다.
# save('경로', 배열명)
np.save('./output/my_array1', a)

# a, b 배열을 바이너리 형태로 저장한다.
# savez('경로', 배열명1, 배열명2, ...)
np.savez('./output/my_array2', a, b)
```

load() 메소드로 npy, npz 파일에서 데이터를 읽어올 수 있다.  
save(), savez() 메소드는 메소드에 따라 정해진 확장명이 있으므로 확장명을 입력하지 않아도 되지만 load() 메소드는 반드시 확장명까지 입력해야 한다.  
load('경로')

```python
# npy 파일 읽기
print(np.load('./output/my_array1.npy'))
print('=' * 100)

# npz 파일 읽기
print(np.load('./output/my_array2.npz'))
npzFile = np.load('./output/my_array2.npz')
print(npzFile['arr_0'])
print(npzFile['arr_1'])
print('=' * 100)

for arr in npzFile:
    print(arr)
    print(npzFile[arr])
```

savetxt() 메소드로 numpy 배열 객체를 텍스트 형태의 파일로 저장할 수 있다.  
여기서 말하는 텍스트 파일은 확장명이 txt인 파일을 말하는 것이 아니고 csv 또는 tsv 파일을 의미한다.  
savetxt('경로', 배열명, delimiter='구분자') 구분자를 생략하면 공백이 구분자로 사용된다.

```python
np.savetxt('./output/saved.csv', a, delimiter=',') # csv 파일은 데이터를 ','로 구분한다.
np.savetxt('./output/saved.tsv', a, delimiter='\t') # tsv 파일은 데이터를 tab('\t')으로 구분한다.
```

브로드캐스팅

shape이 같은 두 배열의 이항 연산은 같은 배열 요소별로 연산이 실행된다.  
두 배열의 shape이 다를 경우 두 배열간의 형상(모양)을 맞추는 아래 그림과 같은 브로드캐스팅 과정을 거쳐 실행된다.

<img src="numpyImage05.jpg" width="900" align="left" />

shape이 같은 두 배열의 연산

```python
a = np.arange(1, 25).reshape(4, 6)
print(a)
b = np.arange(25, 49).reshape(4, 6)
print(b)
```

```python
# shape이 같은 배열을 이항 연산할 경우 위치가 같은 요소 단위로 실행된다.
print(a + b)
```

shape이 다른 두 배열의 연산

shape이 다른 두 배열 사이의 이항 연산에서 브로드캐스팅 발생시 두 배열을 같은 shape으로 만든 후 연산을 실행한다.

```python
# 배열과 스칼라(단일값)의 연산
# 배열과 스칼라 사이의 이항 연산시 스칼라를 배열로 변환한다.
print(a)
print(a + 100)
```

```python
# a + 100은 다음과 같은 과정을 거쳐서 실행된다.
new_arr = np.full_like(a, fill_value=100)
print(new_arr)
print(a + new_arr)
```

```python
# shape이 다른 배열의 연산
a = np.arange(5).reshape(1, 5)
print(a)
b = np.arange(5).reshape(5, 1)
print(b)
print(a + b)
```

배열 복사

numpy 배열 객체에 대한 인덱싱, 슬라이싱이 반환하는 numpy 배열은 새로은 객체가 아닌 기존 배열의 view이다.  
반환된 배열의 값을 변경하면 원본 배열에 변경한 값이 반영된다. => 얕은 복사(주소 복사)

```python
a = np.random.randint(0, 9, (3, 3))
print(a)
b = a[0] # b에는 a배열의 0번째 인덱스에 해당된 데이터가 저장된 주소가 전달된다.
print(b)
b[1] = 777 # a[0]와 b는 같은 주소를 기억하기때문에 b[1]를 수정하면 a[0][1]도 같이 수정된다.
print(b)
print(a)
```

```python
a = np.random.randint(0, 9, (3, 3))
print(a)
b = a[:, 0] # 행은 상관없이(전체행) 0번째 열만 슬라이싱 한다. 한 열 인덱싱
print(b)
b[1] = 777
print(b)
print(a)
```

기존 배열로부터 새로운 배열을 생성하기 위해 copy() 메소드를 사용한다.  
copy() 메소드로 복사된 배열은 원본 배열과 완전히 다른 별도의 객체가 된다. => 깊은 복사  
넘파이배열.copy(), np.copy(넘파이배열)

```python
a = np.random.randint(0, 9, (3, 3))
print(a)
# b = a[0] # 얕은 복사
b = a.copy()[0] # 깊은 복사
print(b)
b[1] = 777
print(b)
print(a)
```

```python
a = np.random.randint(0, 9, (3, 3))
print(a)
# b = a[:, 0] # 얕은 복사
b = np.copy(a)[:, 0] # 깊은 복사
print(b)
b[1] = 777
print(b)
print(a)
```

배열 정렬

numpy는 axis를 기준으로 요소들을 정렬하는 sort() 메소드를 제공한다.  
넘파이배열.sort(), np.sort(넘파이배열)  
sort() 메소드의 axis를 생략했을때 기본값은 None이 아닌 -1이다. -1은 넘파이 배열의 마지막 axis를 의미한다.  
아래 a 배열은 2차원 배열이므로 axis가 0과 1이 존재하고 이때, 마지막 axis는 1이다.

```python
a = np.random.randint(1, 101, (3, 3))
print(a)
```

```python
b = a.copy() # 정렬 작업을 위해 깊은 복사로 복사본을 만든다.
print(b)
# 넘파이배열.sort() 형식으로 sort() 메소드를 실행하면 정렬 결과가 리턴되는 것이 아니고 넘파이 배열에 즉시(바로) 반영된다.
c = b.sort() # b.sort(axis=-1) => 기본값, b.sort(axis=1) => 마지막 2차원 배열의 axis
print(c)
print(b)
```

```python
b = a.copy()
print(b)
c = b.sort(axis=0)
print(c)
print(b)
```

```python
b = a.copy()
print(b)
# np.sort(넘파이배열) 형식으로 sort() 메소드를 실행하면 정렬 결과를 리턴하고 원본은 변형되지 않는다.
c = np.sort(b) # np.sort(b, axis=-1), np.sort(b, axis=1)
print(c)
print(b)
```

```python
b = a.copy()
print(b)
c = np.sort(b, axis=0)
print(c)
print(b)
```
