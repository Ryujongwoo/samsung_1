인덱싱

numpy 배열의 각 요소는 axis 인덱스 배열로 참조할 수 있다.  
1차원 배열은 1개의 인덱스, 2차원 배열은 2개의 인덱스, 3차원 배열은 3개의 인덱스로 요소를 지정할 수 있다.  
인덱싱 작업시 [행][열]과 같이 인덱스를 지정하거나 [행, 열]과 같이 인덱스를 지정할 수 있다.

```python
# 1차원 배열 요소 참조 및 변경
a = np.arange(24) # 1차원 배열
print(a)
print(a[5])
a[5] = 10000
print(a)
```

```python
# 2차원 배열 요소 참조 및 변경
b = np.arange(24).reshape(4, 6) # 2차원 배열
print(b)
print(b[1][2])
b[1, 2] = 10000
print(b)
```

```python
# 3차원 배열 요소 참조 및 변경
c = np.arange(24).reshape(2, 4, 3) # 3차원 배열
print(c)
print(c[1][3][2])
c[1, 3, 2] = 10000
print(c)
```

불린 인덱싱

numpy의 불린 인덱싱은 배열의 각 요소의 선택 여부를 True, False로 지정하는 방식이다. => True가 선택된다.

```python
a = np.arange(24).reshape(4, 6)
print(a)
```

불린 인덱싱을 이용해서 a 배열에서 요소의 값이 짝수인 요소의 전체 합계

```python
even_arr1 = a % 2 # 브로드캐스팅
print(even_arr1)
print('=' * 100)
even_arr2 = even_arr1 == 0 # 브로드캐스팅
print(even_arr2)
print('=' * 100)
even_arr3 = a % 2 == 0 # 브로드캐스팅
print(even_arr3)
```

```python
# 불린 인덱싱은 브로드캐스팅 결과가 True인 요소들만 추출한다.
print(a[even_arr3])
print(a[a % 2 == 0])
```

```python
print(a.sum())
print(np.sum(a))
print(a[a % 2 == 0].sum())
print(np.sum(a[a % 2 == 0]))
print(a[a % 2 == 1].sum())
print(np.sum(a[a % 2 == 1]))
```

배열 변환

배열 변환 방법은 전치, shape 변경, 크기 변경, 추가, 삽입, 삭제, 결합, 분리 등이 있다.

전치는 행렬의 인덱스가 바뀌는 변환이다.

```python
# numpy에서 행렬을 전치하기위해서 넘파이배열.T 속성을 반드시 대문자로 사용한다.
a = np.random.randint(1, 10, (3, 2))
print(a)
print(a.T)
```

numpy는 배열의 형태를 변경하는 reshape() 메소드 raval() 메소드를 제공한다.  
reshape() 메소드는 지정한 shape으로 배열의 형태를 변환하지만 raval() 메소드는 배열의 shape을 무조건 1차원으로 변환한다.

```python
a = np.random.randint(1, 10, (3, 2))
print(a)
# ravel() 메소드는 원본 배열의 view를 반환한다. => 얕은 복사를 한다.
print(a.ravel())
print(a)
```

```python
b = a.ravel()
print(b)
# ravel() 메소드가 리턴한 배열을 수정하면 원본 배열의 내용이 같이 수정된다.
b[4] = 999
print(b)
print(a)
```

```python
a = np.random.randint(1, 10, (3, 2))
print(a)
# reshape() 메소드는 원본 배열의 view를 반환한다. => 얕은 복사를 한다.
print(a.reshape(3, 2, 1))
print(a)
```

```python
b = a.reshape(3, 2, 1)
print(b)
# reshape() 메소드가 리턴한 배열을 수정하면 원본 배열의 내용이 같이 수정된다.
b[1][1][0] = 999
print(b)
print(a)
```

resize() 메소드로 새로운 shape을 지정해서 크기를 변경할 수 있다.  
넘파이배열.resize(new_shape), np.resize(넘파이배열, new_shape)

```python
a = np.random.randint(1, 10, (2, 6))
print(a)
# 넘파이배열.resize(new_shape) 형식은 크기가 변경된 배열을 리턴하지 않는다.
print(a.resize(6, 2)) # 원본 배열 자체의 shape을 변경한다.
print(a)
```

```python
a = np.random.randint(1, 10, (2, 6))
print(a)
# np.resize(넘파이배열, new_shape) 형식은 크기가 변경된 배열을 깊은 복사를 해서 리턴한다.
print(np.resize(a, (6, 2))) # 원본 배열 자체의 shape을 변경하지 않는다.
print(a)
```

```python
b = np.resize(a, (6, 2))
print(b)
b[2, 1] = 999
print(b)
print(a)
```

```python
# reshape() 메소드와 resize() 메소드는 shape을 변경하는 부분에서는 동일하다.
a = np.random.randint(1, 10, (2, 6))
print(a)
# reshape() 메소드는 배열을 구성하는 요소의 개수를 변경하지 못하는 반면에 resize() 메소드는 배열을 구성하는 요소의 개수를 변경할 수 있다.
a.reshape(2, 10) # reshape() 원본과 요소의 개수를 다르게하면 에러가 발생된다.
```

```python
# 배열 요소의 개수가 12개에서 20개로 늘어나고 늘어난 요소에 0이 채워진다.
a.resize(2, 10)
print(a)
# 배열 요소의 개수가 20개에서 8개로 줄어들고 줄어든 요소의 데이터는 분실된다.
a.resize(2, 4)
print(a)
```

```python
a = np.random.randint(1, 10, (2, 6))
print(a)
# 넘파이배열.resize(new_shape) 형식과는 달리 np.resize(넘파이배열, new_shape) 형식는 배열 요소의 개수가 12개에서 20개로 늘어나고 늘어난 요소에 원본 배열의
# 데이터가 반복해서 채워진다.
b = np.resize(a, (2, 10))
print(b)
c = np.resize(a, (2, 4))
print(c)
```

append() 메소드는 axis로 데이터가 추가되는 방향을 지정해서 배열 끝에 데이터(배열)를 추가한다.  
np.append(넘파이배열, 추가할넘파이배열[, axis])

```python
a = np.random.randint(1, 10, (3, 3))
print(a)
b = np.random.randint(10, 19, (3, 3))
print(b)
```

```python
# axis를 생략하면 None이 기본값으로 사용되고 1차원 배열로 변환되서 결합된다.
print(np.append(a, b))
result = np.append(a, b, axis=None)
print(result)
```

```python
# axis=0인 경우 열방향(아래쪽)으로 배열을 추가한다.
print(np.append(a, b, axis=0))
```

```python
# axis=0인 경우 두 배열의 행의 개수는 달라도 상관없지만 열의 개수는 반드시 같아야 한다.
c = np.random.randint(20, 29, (4, 3)) # a 배열과 행의 개수는 다르지만 열의 개수가 같은 배열
print(c)
print(np.append(a, c, axis=0))
print('=' * 100)

d = np.random.randint(20, 29, (3, 4)) # a 배열과 행의 개수는 같지만 열의 개수가 다른 배열
print(d)
# print(np.append(a, d, axis=0)) # a 배열과 d 배열의 열의 개수가 달라서 에러가 발생된다.
```

```python
# axis=1인 경우 행방향(오른쪽)으로 배열을 추가한다.
print(np.append(a, b, axis=1))
```

```python
# axis=1인 경우 두 배열의 열의 개수는 달라도 상관없지만 행의 개수는 반드시 같아야 한다.
c = np.random.randint(20, 29, (4, 3)) # a 배열과 행의 개수는 다르지만 열의 개수가 같은 배열
print(c)
# print(np.append(a, c, axis=1)) # a 배열과 c 배열의 행의 개수가 달라서 에러가 발생된다.
print('=' * 100)

d = np.random.randint(20, 29, (3, 4)) # a 배열과 행의 개수는 같지만 열의 개수가 다른 배열
print(d)
print(np.append(a, d, axis=1))
```

np.concatenate((넘파이배열1, 넘파이배열2, ...)[, axis]), np.concatenate([넘파이배열1, 넘파이배열2, ...][, axis])

```python
# axis=0(기본값)을 지정하면 열방향(아래쪽)으로 배열을 결합한다. 결합하려는 열의 개수가 다르면 에러가 발행한다.
print(np.concatenate([a, b]))
print(np.concatenate([a, b], axis=0))
```

```python
# axis=1을 지정하면 행방향(오른쪽)으로 배열을 결합한다. 결합하려는 행의 개수가 다르면 에러가 발행한다.
print(np.concatenate([a, b], axis=1))
```

배열 분리

vsplit() 메소드는 배열을 열(수직) 방향으로 분할한다.  
np.vsplit(넘파이배열, 분할할개수)

```python
a = np.arange(1, 25).reshape(4, 6)
print(a)
```

```python
result = np.vsplit(a, 2)
print(type(result)) # vsplit() 메소드의 실행 결과는 numpy 배열이 저장된 파이썬 list 형태이다.
print(result)
print(type(result[0]))
print(result[0])
print(type(result[1]))
print(result[1])
```

```python
# vsplit() 메소드로 배열을 분리하려면 기본적으로 전체 행의 개수의 약수개로 분리가 가능하다.
result = np.vsplit(a, 4)
print(result)
```

```python
# vsplit() 메소드의 2번째 인수로 분할 기준이되는 인덱스를 튜플로 넘겨주면 튜플에 지정된 행 인덱스에 따라 구분된다.
result = np.vsplit(a, (1, 3))
print(result)
```

hsplit() 메소드는 배열을 행(수직) 방향으로 분할한다.  
np.hsplit(넘파이배열, 분할할개수)

```python
result = np.hsplit(a, 2)
print(type(result)) # hsplit() 메소드의 실행 결과는 numpy 배열이 저장된 파이썬 list 형태이다.
print(result)
print(type(result[0]))
print(result[0])
print(type(result[1]))
print(result[1])
```

```python
# hsplit() 메소드로 배열을 분리하려면 기본적으로 전체 열의 개수의 약수개로 분리가 가능하다.
result = np.hsplit(a, 3)
print(result)
```

```python
result = np.hsplit(a, 6)
print(result)
```

```python
# hsplit() 메소드의 2번째 인수로 분할 기준이되는 인덱스를 튜플로 넘겨주면 튜플에 지정된 열 인덱스에 따라 구분된다.
result = np.hsplit(a, (1, 3, 5))
print(result)
```
