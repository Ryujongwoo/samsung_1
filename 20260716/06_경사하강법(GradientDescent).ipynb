```python
import warnings
warnings.filterwarnings("ignore")
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

경사 하강법(Gradient Descent)

가중치를 너무 크게잡아도 너무 작게잡아도 오차가 커지기때문에 평균 제곱근 오차를 계산하면 'U'형태의 2차 곡선이 만들어진다.  
이렇게 만들어진 2차 곡선의 임의의 한 점을 선택해서 미분을 실행하면 임의의 점에 대한 기울기가 계산되고 기울기를 0에 최대한 가깝게 이동하면서 최적의 기울기를 계산한다.

<img src="./gradient_descent.png" width="1100" align="left" />

데이터를 준비한다.

```python
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
xData = [d[0] for d in data]
yData = [d[1] for d in data]

for i in range(len(xData)):
    print(f'공부 시간: {xData[i]}, 공부 시간에 따른 시험 점수: {yData[i]}')
```

난수를 이용해서 가중치와 바이어스를 정한다.

```python
a = tf.Variable(tf.random_uniform([1], 0, 10))
b = tf.Variable(tf.random_uniform([1], 0, 100))

sess = tf.Session()
sess.run(tf.global_variables_initializer())
print('a = {}, b = {}'.format(sess.run(a), sess.run(b)))
```

예측 성적을 얻기위한 가설(수식)을 정의한다.

```python
Y = a * xData + b
```

RMSE에서 사용할 수식(오차 함수)을 만든다.

```python
# np.sqrt(((predict_result - yData) ** 2).mean()) => numpy 사용
rmse = tf.sqrt(tf.reduce_mean(tf.square(Y - yData)))
```

경사 하강법 알고리즘을 이용해서 RMSE를 최소로하는 값을 찾는다.

```python
learning_rate = 0.1
gradient_descent = tf.train.GradientDescentOptimizer(learning_rate).minimize(rmse)
```

학습시킨다.

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for epoch in range(3001):
    sess.run(gradient_descent)
    if epoch % 200 == 0:
        print('epoch: {:4d}, rmse: {:10.7f}, a: {:10.7f}, b: {:10.7f}'.format(epoch, sess.run(rmse), sess.run(a)[0], sess.run(b)[0]))
```
