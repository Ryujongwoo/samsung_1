tensorflow를 사용하기위해 라이브러리를 설치하고 import 한다.

상수, 변수, 수식을 만든다. => 세션에 넣는다. 변수는 반드시 초기화 시킨다. => 실행한다.

tensorflow 상수

```python
# constant() 메소드로 tensorflow에서 사용할 상수를 정의한다.
a = tf.constant(100) # 상수 정의
print('a = {}'.format(a))
b = tf.constant(200)
print('b = {}'.format(b))

# tensorflow 1.x 버전은 반드시 세션이라는 작업 영역을 만든 후 상수, 변수, 수식을 대입한 후 실행한다.
# Session() 클래스로 tensorflow에서 사용할 세션을 정의한다.
sess = tf.Session() # tensorflow 세션을 만든다.
# run() 메소드로 tensorflow 세션의 내용을 실행한다.
print('a = {}'.format(sess.run(a)))
print('b = {}'.format(sess.run(b)))
```

tensorflow 변수

```python
# Variable() 클래스로 tensorflow에서 사용할 변수를 정의한다.
c = tf.Variable(300) # 변수 정의
print('c = {}'.format(c))
d = tf.Variable(400)
print('d = {}'.format(d))

sess = tf.Session()

# tensorflow 변수 초기화
# Variable() 클래스로 생성한 tensorflow 변수는 사용하기 전에 반드시 초기화시켜야 한다.
# tensorflow 변수 초기화는 tensorflow 변수를 사용하기 전에만 하면된다.
# global_variables_initializer() 메소드로 tensorflow 변수를 초기화시킨다.

init = tf.global_variables_initializer()
sess.run(init)
# 위의 2줄을 사용한 tensorflow 변수 초기화는 아래와 같이 1줄로 줄여서 사용할 수 있다.
# sess.run(tf.global_variables_initializer())

print('c = {}'.format(sess.run(c)))
print('d = {}'.format(sess.run(d)))

e = tf.Variable(500)
# tensorflow 변수 초기화를 실행한 후 정의한 변수는 다시 초기화 시켜야 한다.
sess.run(tf.global_variables_initializer())
print('e = {}'.format(sess.run(e)))
```

tensorflow 수식

```python
a = tf.constant(100)
b = tf.constant(200)
c = tf.Variable(300)
d = tf.Variable(400)

e = a + b # 수식 정의
print('e = {}'.format(e))
f = tf.add(a, b)
print('f = {}'.format(f))

g = c * d
print('g = {}'.format(g))
h = tf.multiply(c, d)
print('h = {}'.format(h))

sess = tf.Session()
sess.run(tf.global_variables_initializer())

print('e = {}'.format(sess.run(e)))
print('f = {}'.format(sess.run(f)))
print('g = {}'.format(sess.run(g)))
print('h = {}'.format(sess.run(h)))
```
